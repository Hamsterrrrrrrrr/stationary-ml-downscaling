```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
import seaborn as sns
from scipy import stats

# Set style
sns.set_style("white")

# Load evaluation results
eval_dir = Path("evaluation_results")
scenarios = ['historical', 'ssp126', 'ssp245', 'ssp585']

# Variables configuration
variables = [
    {'name': 'tas', 'label': 'Temperature', 'unit': 'C', 'decimals': 2},
    {'name': 'pr', 'label': 'Precipitation', 'unit': 'mm/day', 'decimals': 3},
    {'name': 'hurs', 'label': 'Relative Humidity', 'unit': '%', 'decimals': 2},
    {'name': 'sfcWind', 'label': 'Wind Speed', 'unit': 'm/s', 'decimals': 4}
]

# Load all data
all_data = {}
for var in variables:
    all_data[var['name']] = {}
    for scenario in scenarios:
        ds = xr.open_dataset(eval_dir / f"{var['name']}_evaluation_{scenario}.nc")
        all_data[var['name']][scenario] = ds

# Selected normalization methods only
norm_methods = ['minmax_pixel', 'zscore_global', 'zscore_pixel', 'instance_zscore']

# Model labels for display
model_labels = {
    'minmax_pixel': 'MinMax-P',
    'zscore_global': 'Z-score-G',
    'zscore_pixel': 'Z-score-P',
    'instance_zscore': 'Instance-Z'
}

# Colors for each model
model_colors = {
    'minmax_pixel': '#0277BD',   # Dark Blue
    'zscore_global': '#00ACC1',  # CYAN
    'zscore_pixel': '#C62828',   # Dark Red
    'instance_zscore': '#F57C00' # Dark Orange
}

# Scenario labels
scenario_labels = {
    'historical': 'Historical',
    'ssp126': 'SSP1-2.6',
    'ssp245': 'SSP2-4.5',
    'ssp585': 'SSP5-8.5'
}

# Scenario colors
scenario_colors = {
    'historical': '#000000',
    'ssp126': '#2E86AB',
    'ssp245': '#F77F00',
    'ssp585': '#C1121F'
}

def calc_global_mean(data):
    """Calculate global mean time series"""
    if isinstance(data, xr.DataArray):
        return data.mean(dim=['lat', 'lon']).values
    else:
        return np.mean(data, axis=(1, 2))

def calculate_trend(time_series, time_coords):
    """Calculate linear trend using least squares regression."""
    # Remove NaN values
    mask = ~np.isnan(time_series)
    x = time_coords[mask]
    y = time_series[mask]
    
    # Linear regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
    
    return slope, intercept, r_value, p_value

# Create 4x4 figure
fig, axes = plt.subplots(4, 4, figsize=(20, 16))

# Plot each scenario (row) and variable (column)
for row_idx, scenario in enumerate(scenarios):
    for col_idx, var in enumerate(variables):
        var_name = var['name']
        var_label = var['label']
        var_unit = var['unit']
        decimals = var['decimals']
        
        ax = axes[row_idx, col_idx]
        
        ds = all_data[var_name][scenario]
        gt_data = ds['groundtruth'].values
        time = ds.time
        
        # Convert time to decimal years for trend analysis
        years = time.dt.year.values + time.dt.month.values / 12.0
        
        # Get ground truth global mean time series
        gt_mean = calc_global_mean(gt_data)
        
        # Calculate ground truth trend
        gt_slope, gt_intercept, gt_r, gt_p = calculate_trend(gt_mean, years)
        
        # Calculate 12-month rolling mean for smoother visualization
        window = 12
        if len(gt_mean) >= window:
            gt_mean_smooth = np.convolve(gt_mean, np.ones(window)/window, mode='valid')
            years_smooth = years[window-1:]
        else:
            gt_mean_smooth = gt_mean
            years_smooth = years
        
        # Plot ground truth as LINE
        ax.plot(years_smooth, gt_mean_smooth, color='black', linewidth=1.5, 
                alpha=0.6, label='Ground Truth', zorder=18)
        
        # Plot ground truth trend line
        gt_trend_line = gt_slope * years + gt_intercept
        trend_value = round(gt_slope * 10, decimals)
        trend_label = f'GT: {trend_value} {var_unit}/decade'
        ax.plot(years, gt_trend_line, 'k-', linewidth=2.5, alpha=0.95,
                label=trend_label, zorder=19)
        
        # Plot predictions with trend lines
        for norm_method in norm_methods:
            pred_key = f'pred_{norm_method}'
            if pred_key in ds:
                pred_data = ds[pred_key].values
                pred_mean = calc_global_mean(pred_data)
                
                # Calculate trend
                pred_slope, pred_intercept, pred_r, pred_p = calculate_trend(pred_mean, years)
                
                # Plot trend line only (no data points)
                pred_trend_line = pred_slope * years + pred_intercept
                trend_value = round(pred_slope * 10, decimals)
                trend_label = f'{model_labels[norm_method]}: {trend_value}'
                ax.plot(years, pred_trend_line, 
                       color=model_colors[norm_method],
                       linewidth=2, alpha=0.85,
                       linestyle='--',
                       label=trend_label)
        
        # Formatting
        ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.8)
        ax.tick_params(labelsize=10)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        
        # X-axis labels only on bottom row
        if row_idx == 3:
            ax.set_xlabel('Year', fontsize=11, fontweight='bold')
        else:
            ax.set_xlabel('')
        
        # Y-axis labels on all columns (different units)
        ax.set_ylabel(f'{var_unit}', fontsize=11, fontweight='bold')
        
        # Column titles on top row
        if row_idx == 0:
            ax.set_title(var_label, fontsize=13, fontweight='bold', pad=12)
        
        # Scenario label on leftmost column
        if col_idx == 0:
            ax.text(-0.25, 0.5, scenario_labels[scenario], 
                   transform=ax.transAxes,
                   fontsize=12, fontweight='bold',
                   rotation=90, va='center', ha='center',
                   color=scenario_colors[scenario])
        
        # Add legend only on top-left panel
        if row_idx == 0 and col_idx == 0:
            ax.legend(loc='upper left', fontsize=7, framealpha=0.98,
                     edgecolor='black', fancybox=False)

# Adjust layout
plt.tight_layout(rect=[0.05, 0, 1, 1.0])


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
import seaborn as sns
from scipy import stats

# Set style
sns.set_style("white")

# Load evaluation results
eval_dir = Path("evaluation_results")
scenarios = ['historical', 'ssp126', 'ssp245', 'ssp585']

# Only temperature
var_name = 'tas'
var_label = 'Temperature'
var_unit = 'C'

# Load all data for temperature
all_data = {}
for scenario in scenarios:
    ds = xr.open_dataset(eval_dir / f"{var_name}_evaluation_{scenario}.nc")
    all_data[scenario] = ds

# Selected normalization methods only
norm_methods = ['minmax_pixel', 'zscore_global', 'zscore_pixel', 'instance_zscore']

# Model labels for display
model_labels = {
    'minmax_pixel': 'MinMax-P',
    'zscore_global': 'Z-score-G',
    'zscore_pixel': 'Z-score-P',
    'instance_zscore': 'Instance-Z'
}

# Colors for each model
model_colors = {
    'minmax_pixel': '#0277BD',   # Dark Blue
    'zscore_global': '#00ACC1',  # CYAN
    'zscore_pixel': '#C62828',   # Dark Red
    'instance_zscore': '#F57C00' # Dark Orange
}

# Scenario labels
scenario_labels = {
    'historical': 'Historical',
    'ssp126': 'SSP1-2.6',
    'ssp245': 'SSP2-4.5',
    'ssp585': 'SSP5-8.5'
}

# Scenario colors
scenario_colors = {
    'historical': '#000000',
    'ssp126': '#2E86AB',
    'ssp245': '#F77F00',
    'ssp585': '#C1121F'
}

def calc_global_mean(data):
    """Calculate global mean time series"""
    if isinstance(data, xr.DataArray):
        return data.mean(dim=['lat', 'lon']).values
    else:
        return np.mean(data, axis=(1, 2))

def calculate_autocorrelation(time_series, max_lag=100):
    """Calculate temporal autocorrelation function."""
    n = len(time_series)
    mean = np.mean(time_series)
    var = np.var(time_series)
    
    acf = np.zeros(max_lag + 1)
    for lag in range(max_lag + 1):
        if lag == 0:
            acf[lag] = 1.0
        else:
            c0 = np.sum((time_series[:n-lag] - mean) * (time_series[lag:] - mean))
            acf[lag] = c0 / (var * (n - lag))
    
    lags = np.arange(max_lag + 1)
    return lags, acf

def calculate_trend(time_series, time_coords):
    """Calculate linear trend using least squares regression."""
    # Remove NaN values
    mask = ~np.isnan(time_series)
    x = time_coords[mask]
    y = time_series[mask]
    
    # Linear regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
    
    return slope, intercept, r_value, p_value

# Create 4x2 figure (4 scenarios  2 columns: Autocorrelation + Trend)
fig = plt.figure(figsize=(16, 16))
gs = fig.add_gridspec(4, 2, hspace=0.3, wspace=0.3)

# Plot each scenario (row)
for row_idx, scenario in enumerate(scenarios):
    ds = all_data[scenario]
    gt_data = ds['groundtruth'].values
    time = ds.time
    
    # Convert time to decimal years for trend analysis
    years = time.dt.year.values + time.dt.month.values / 12.0
    
    # Get ground truth global mean time series
    gt_mean = calc_global_mean(gt_data)
    
    # ========== Left column: Autocorrelation ==========
    ax_acf = fig.add_subplot(gs[row_idx, 0])
    
    # Calculate ground truth autocorrelation
    lags, gt_acf = calculate_autocorrelation(gt_mean, max_lag=100)
    
    # Calculate confidence interval (95%)
    n = len(gt_mean)
    confidence_interval = 1.96 / np.sqrt(n)
    
    # Plot confidence interval as shaded region
    ax_acf.fill_between(lags, -confidence_interval, confidence_interval,
                        color='lightblue', alpha=0.3, label='95% CI')
    
    # Plot ground truth ACF with filled area (dark painted)
    ax_acf.fill_between(lags, 0, gt_acf, where=(gt_acf >= 0),
                        color='black', alpha=0.7, interpolate=True)
    ax_acf.fill_between(lags, 0, gt_acf, where=(gt_acf < 0),
                        color='black', alpha=0.7, interpolate=True)
    ax_acf.plot(lags, gt_acf, 'o', color='blue', markersize=4, 
               label='Ground Truth', zorder=20)
    
    # Plot predictions ACF as colored lines
    for norm_method in norm_methods:
        pred_key = f'pred_{norm_method}'
        if pred_key in ds:
            pred_data = ds[pred_key].values
            pred_mean = calc_global_mean(pred_data)
            
            # Calculate autocorrelation
            _, pred_acf = calculate_autocorrelation(pred_mean, max_lag=100)
            
            ax_acf.plot(lags, pred_acf, 
                       color=model_colors[norm_method],
                       linewidth=2.5, alpha=0.9,
                       linestyle='--',
                       label=model_labels[norm_method])
    
    # Format ACF plot
    ax_acf.set_xlabel('Lag (months)', fontsize=12, fontweight='bold')
    ax_acf.set_ylabel('Autocorrelation', fontsize=12, fontweight='bold')
    ax_acf.set_title(f'{scenario_labels[scenario]}', 
                    fontsize=13, fontweight='bold', color=scenario_colors[scenario], 
                    pad=12, loc='left')
    ax_acf.grid(True, alpha=0.3, linestyle='--', linewidth=0.8)
    ax_acf.set_ylim(-1.0, 1.05)
    ax_acf.axhline(y=0, color='gray', linestyle='-', linewidth=1, alpha=0.5)
    ax_acf.set_xlim(0, 100)
    
    # Add legend only on top panel
    if row_idx == 0:
        ax_acf.legend(loc='upper right', fontsize=9, framealpha=0.98, 
                     edgecolor='black', fancybox=False)
    
    ax_acf.tick_params(labelsize=11)
    ax_acf.spines['top'].set_visible(False)
    ax_acf.spines['right'].set_visible(False)
    
    # Column header
    if row_idx == 0:
        ax_acf.text(0.5, 1.15, 'Autocorrelation', 
                   transform=ax_acf.transAxes,
                   fontsize=14, fontweight='bold', ha='center')
    
    # ========== Right column: Trend ==========
    ax_trend = fig.add_subplot(gs[row_idx, 1])
    
    # Calculate ground truth trend
    gt_slope, gt_intercept, gt_r, gt_p = calculate_trend(gt_mean, years)
    
    # Calculate 12-month rolling mean for smoother visualization
    window = 12
    if len(gt_mean) >= window:
        gt_mean_smooth = np.convolve(gt_mean, np.ones(window)/window, mode='valid')
        years_smooth = years[window-1:]
    else:
        gt_mean_smooth = gt_mean
        years_smooth = years
    
    # Plot ground truth as LINE
    ax_trend.plot(years_smooth, gt_mean_smooth, color='black', linewidth=1.5, 
                 alpha=0.6, label='Ground Truth', zorder=18)
    
    # Plot ground truth trend line
    gt_trend_line = gt_slope * years + gt_intercept
    trend_value = round(gt_slope * 10, 2)
    trend_label = f'GT: {trend_value} {var_unit}/decade'
    ax_trend.plot(years, gt_trend_line, 'k-', linewidth=2.5, alpha=0.95,
                 label=trend_label, zorder=19)
    
    # Plot predictions with trend lines
    for norm_method in norm_methods:
        pred_key = f'pred_{norm_method}'
        if pred_key in ds:
            pred_data = ds[pred_key].values
            pred_mean = calc_global_mean(pred_data)
            
            # Calculate trend
            pred_slope, pred_intercept, pred_r, pred_p = calculate_trend(pred_mean, years)
            
            # Plot trend line only (no data points)
            pred_trend_line = pred_slope * years + pred_intercept
            trend_value = round(pred_slope * 10, 2)
            trend_label = f'{model_labels[norm_method]}: {trend_value}'
            ax_trend.plot(years, pred_trend_line, 
                         color=model_colors[norm_method],
                         linewidth=2.5, alpha=0.85,
                         linestyle='--',
                         label=trend_label)
    
    # Format trend plot
    ax_trend.set_xlabel('Year', fontsize=12, fontweight='bold')
    ax_trend.set_ylabel(f'Global Mean Temperature ({var_unit})', fontsize=12, fontweight='bold')
    ax_trend.set_title(f'{scenario_labels[scenario]}', 
                      fontsize=13, fontweight='bold', color=scenario_colors[scenario], 
                      pad=12, loc='left')
    ax_trend.grid(True, alpha=0.3, linestyle='--', linewidth=0.8)
    
    # Add legend only on top panel
    if row_idx == 0:
        ax_trend.legend(loc='upper left', fontsize=9, framealpha=0.98,
                       edgecolor='black', fancybox=False)
    
    ax_trend.tick_params(labelsize=11)
    ax_trend.spines['top'].set_visible(False)
    ax_trend.spines['right'].set_visible(False)
    
    # Column header
    if row_idx == 0:
        ax_trend.text(0.5, 1.15, 'Temperature Trend', 
                     transform=ax_trend.transAxes,
                     fontsize=14, fontweight='bold', ha='center')

# Adjust layout
plt.tight_layout()

plt.show()
```
