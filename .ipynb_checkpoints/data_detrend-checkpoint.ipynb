```python
import xarray as xr
import numpy as np
import pickle
from scipy import stats

# Load all datasets
data_dir = "data"
ds_hist = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_historical_r1i1p1f1_1850_2014_allvars.nc")
ds_ssp126 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp126_r1i1p1f1_2015_2100_allvars.nc")
ds_ssp245 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp245_r1i1p1f1_2015_2100_allvars.nc")
ds_ssp585 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp585_r1i1p1f1_2015_2100_allvars.nc")

# Define periods for detrending methods 2 and 3
detrend_periods = {
    'historical': [(1850, 1980), (1981, 2000), (2001, 2014)],
    'ssp': [(2015, 2100)]
}

# Detrending Method 1: Remove global mean anomaly (shared baseline)
def compute_climatology(data, baseline_period=(1850, 1900)):
    """Compute monthly climatology from baseline period"""
    start_year, end_year = baseline_period
    clim_data = data.sel(time=slice(f'{start_year}', f'{end_year}'))
    climatology = clim_data.groupby('time.month').mean(dim='time')
    return climatology

def detrend_global_mean_anomaly(data, hist_data=None):
    """Remove global mean anomaly using historical baseline"""
    # Use historical data for baseline if provided, otherwise use data itself
    baseline_data = hist_data if hist_data is not None else data
    climatology = compute_climatology(baseline_data, baseline_period=(1850, 1900))
    
    anomaly = data.groupby('time.month') - climatology
    global_mean_anomaly = anomaly.mean(dim=['lat', 'lon'])
    
    global_mean_broadcast = xr.DataArray(
        np.broadcast_to(global_mean_anomaly.values[:, None, None], data.shape),
        coords=data.coords,
        dims=data.dims
    )
    return data - global_mean_broadcast

# Detrending Method 2: Linear detrend at each grid point by period
def detrend_grid_point_by_period(data, periods):
    """Linear detrending at each grid point within specified periods"""
    data_np = data.values
    n_time, n_lat, n_lon = data_np.shape
    years = data.time.dt.year.values
    detrended = np.zeros_like(data_np)
    
    for start_year, end_year in periods:
        mask = (years >= start_year) & (years <= end_year)
        if not mask.any():
            continue
            
        period_data = data_np[mask]
        n_time_period = period_data.shape[0]
        
        t = np.arange(n_time_period).astype(float)
        t_mean = np.mean(t)
        t_centered = t - t_mean
        denominator = np.sum(t_centered**2)
        
        data_reshaped = period_data.reshape(n_time_period, -1)
        data_mean = np.mean(data_reshaped, axis=0)
        data_centered = data_reshaped - data_mean
        
        slopes = np.sum(t_centered[:, np.newaxis] * data_centered, axis=0) / denominator
        intercepts = data_mean - slopes * t_mean
        trends = intercepts[np.newaxis, :] + slopes[np.newaxis, :] * t[:, np.newaxis]
        
        detrended_reshaped = data_reshaped - trends
        detrended[mask] = detrended_reshaped.reshape(n_time_period, n_lat, n_lon)
    
    return xr.DataArray(detrended, coords=data.coords, dims=data.dims)

# Detrending Method 3: Remove global mean linear trend by period
def detrend_global_mean_trend_by_period(data, periods):
    """Fit linear trend to global mean within periods, then remove"""
    data_np = data.values
    n_time, n_lat, n_lon = data_np.shape
    years = data.time.dt.year.values
    detrended = np.zeros_like(data_np)
    
    for start_year, end_year in periods:
        mask = (years >= start_year) & (years <= end_year)
        if not mask.any():
            continue
            
        period_data = data_np[mask]
        global_mean = np.mean(period_data, axis=(1, 2))
        n_time_period = len(global_mean)
        
        t = np.arange(n_time_period).astype(float)
        slope, intercept, _, _, _ = stats.linregress(t, global_mean)
        global_trend = intercept + slope * t
        
        global_trend_broadcast = global_trend[:, None, None]
        detrended[mask] = period_data - global_trend_broadcast
    
    return xr.DataArray(detrended, coords=data.coords, dims=data.dims)

# Process each dataset
def process_dataset(ds, ds_name, ds_hist_for_baseline=None, periods=None):
    """Process dataset to add residuals and detrended variants"""
    ds_new = xr.Dataset()
    variables = ['pr', 'tas']
    
    for var in variables:
        var_hr = f"{var}_hr"
        var_lr = f"{var}_lr_interp"
        
        if var_hr not in ds or var_lr not in ds:
            continue
        
        # Copy original variables
        ds_new[var_hr] = ds[var_hr]
        ds_new[var_lr] = ds[var_lr]
        
        # Add residual
        ds_new[f"{var}_residual"] = ds[var_hr] - ds[var_lr]
        
        # Detrending Method 1: Global mean anomaly (use hist baseline for SSPs)
        if ds_hist_for_baseline is not None:
            ds_new[f"{var}_lr_detrend_gma"] = detrend_global_mean_anomaly(
                ds[var_lr], hist_data=ds_hist_for_baseline[var_lr])
        else:
            ds_new[f"{var}_lr_detrend_gma"] = detrend_global_mean_anomaly(ds[var_lr])
        
        # Detrending Methods 2 & 3: Period-specific detrending
        if periods:
            ds_new[f"{var}_lr_detrend_grid"] = detrend_grid_point_by_period(ds[var_lr], periods)
            ds_new[f"{var}_lr_detrend_gmt"] = detrend_global_mean_trend_by_period(ds[var_lr], periods)
    
    # Copy coordinates
    ds_new = ds_new.assign_coords(ds.coords)
    
    return ds_new

# Process historical dataset
print("Processing historical dataset...")
ds_hist_new = process_dataset(ds_hist, 'historical', periods=detrend_periods['historical'])
ds_hist_new.to_netcdf(f"{data_dir}/MPI-ESM1-2-HR-LR_historical_residual_detrended.nc")

# Process SSP datasets
print("Processing SSP126 dataset...")
ds_ssp126_new = process_dataset(ds_ssp126, 'ssp126', ds_hist_for_baseline=ds_hist, periods=detrend_periods['ssp'])
ds_ssp126_new.to_netcdf(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp126_residual_detrended.nc")

print("Processing SSP245 dataset...")
ds_ssp245_new = process_dataset(ds_ssp245, 'ssp245', ds_hist_for_baseline=ds_hist, periods=detrend_periods['ssp'])
ds_ssp245_new.to_netcdf(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp245_residual_detrended.nc")

print("Processing SSP585 dataset...")
ds_ssp585_new = process_dataset(ds_ssp585, 'ssp585', ds_hist_for_baseline=ds_hist, periods=detrend_periods['ssp'])
ds_ssp585_new.to_netcdf(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp585_residual_detrended.nc")

# Compute z-score pixel statistics for training period only
print("Computing z-score pixel statistics...")
train_start = '1850'
train_end = '1980'

norm_stats_zscore_pixel = {}
variables = ['pr', 'tas']

for var in variables:
    norm_stats_zscore_pixel[var] = {}
    
    variants = {
        'hr': f"{var}_hr",
        'lr_interp': f"{var}_lr_interp",
        'residual': f"{var}_residual",
        'lr_detrend_gma': f"{var}_lr_detrend_gma",
        'lr_detrend_grid': f"{var}_lr_detrend_grid",
        'lr_detrend_gmt': f"{var}_lr_detrend_gmt",
    }
    
    for key, var_name in variants.items():
        if var_name in ds_hist_new:
            data = ds_hist_new[var_name].sel(time=slice(train_start, train_end)).values
            
            # Only store pixel mean and std for z-score normalization
            norm_stats_zscore_pixel[var][key] = {
                'pixel_mean': np.mean(data, axis=0),
                'pixel_std': np.std(data, axis=0)
            }

# Save statistics
print("Saving z-score pixel statistics...")
with open(f"{data_dir}/norm_stats_zscore_pixel_residual_detrended.pkl", 'wb') as f:
    pickle.dump(norm_stats_zscore_pixel, f)

print("Done!")
```
