# Concatenate along time

```python
import xarray as xr
import glob
import os
from pathlib import Path

# Set base directory
base_dir = "."
output_dir = "./concatenated"
os.makedirs(output_dir, exist_ok=True)

# Define all variables, scenarios, and resolutions
variables = ['pr', 'tas']
scenarios = ['historical', 'ssp126', 'ssp245', 'ssp585', 'G6sulfur']
resolutions = ['HR', 'LR']
ensemble_member = 'r1i1p1f1'  

# Track successful and failed concatenations
successful = []
failed = []

# Process each combination
for variable in variables:
    for resolution in resolutions:
        for scenario in scenarios:
            # Create the pattern to match files with specified ensemble member
            pattern = f"{variable}_Amon_MPI-ESM1-2-{resolution}_{scenario}_{ensemble_member}_gn_*.nc"
            
            # Find all matching files
            files = sorted(glob.glob(os.path.join(base_dir, pattern)))
            
            if len(files) == 0:
                print(f"‚ö† No files found for {variable}_{resolution}_{scenario}_{ensemble_member}")
                failed.append(f"{variable}_{resolution}_{scenario}_{ensemble_member}")
                continue
            
            print(f"\n{'='*70}")
            print(f"Processing: {variable} | {resolution} | {scenario} | {ensemble_member}")
            print(f"Found {len(files)} files")
            print(f"{'='*70}")
            
            try:
                # Determine time range from filenames
                first_file = os.path.basename(files[0])
                last_file = os.path.basename(files[-1])
                
                # Extract start year from first file and end year from last file
                start_year = first_file.split('_')[-1].split('-')[0][:4]
                end_year = last_file.split('_')[-1].split('-')[1][:4]
                
                # Create output filename with ensemble member
                output_name = f"{variable}_Amon_MPI-ESM1-2-{resolution}_{scenario}_{ensemble_member}_{start_year}_{end_year}.nc"
                output_file = os.path.join(output_dir, output_name)
                
                # Load first file
                print(f"  [1/{len(files)}] Loading: {os.path.basename(files[0])}")
                combined_ds = xr.open_dataset(files[0])
                
                # Concatenate remaining files
                for i, file_path in enumerate(files[1:], 2):
                    print(f"  [{i}/{len(files)}] Concatenating: {os.path.basename(file_path)}")
                    current_ds = xr.open_dataset(file_path)
                    combined_ds = xr.concat([combined_ds, current_ds], dim='time')
                    current_ds.close()
                    del current_ds
                
                # Save concatenated dataset
                print(f"  üíæ Saving to: {output_name}")
                combined_ds.to_netcdf(output_file)
                print(f"  ‚úÖ Success! Time range: {combined_ds.time.values[0]} to {combined_ds.time.values[-1]}")
                
                successful.append(output_name)
                
                # Clean up
                combined_ds.close()
                del combined_ds
                
            except Exception as e:
                print(f"  ‚ùå Error: {str(e)}")
                failed.append(f"{variable}_{resolution}_{scenario}_{ensemble_member}")

# Print summary
print("\n" + "="*70)
print("CONCATENATION SUMMARY")
print("="*70)
print(f"Ensemble member: {ensemble_member}")
print(f"\n‚úÖ Successfully concatenated: {len(successful)} datasets")
if successful:
    for name in successful:
        print(f"   - {name}")

if failed:
    print(f"\n‚ùå Failed or skipped: {len(failed)} datasets")
    for name in failed:
        print(f"   - {name}")

print(f"\nüìÅ Output directory: {output_dir}/")
print("="*70)
```

# Concatenate along variable

```python
import xarray as xr
import os

# Set directories
input_dir = "./concatenated"
output_dir = "./merged"
os.makedirs(output_dir, exist_ok=True)

# Define scenarios and resolutions
scenarios = ['historical', 'ssp126', 'ssp245', 'ssp585', 'G6sulfur']
resolutions = ['HR', 'LR']
variables = ['pr', 'tas']

# Track progress
successful = []
failed = []

print("="*70)
print("MERGING VARIABLES INTO SINGLE FILES")
print("="*70)

# Process each scenario-resolution combination
for scenario in scenarios:
    for resolution in resolutions:
        print(f"\n{'='*70}")
        print(f"Processing: {scenario} | {resolution}")
        print(f"{'='*70}")

        try:
            # Determine time range based on scenario and resolution
            if scenario == 'historical':
                time_range = '1850_2014'
            elif scenario == 'G6sulfur':
                if resolution == 'HR':
                    time_range = '2020_2099'
                else:  # LR
                    time_range = '2015_2099'
            else:  # ssp126, ssp245, ssp585
                time_range = '2015_2100'
            
            # Dictionary to store data arrays
            data_arrays = {}
            
            # Load each variable
            for var in variables:
                filename = f"{var}_Amon_MPI-ESM1-2-{resolution}_{scenario}_r1i1p1f1_{time_range}.nc"
                filepath = os.path.join(input_dir, filename)
                
                if os.path.exists(filepath):
                    print(f"  Loading: {filename}")
                    ds = xr.open_dataset(filepath)
                    
                    # Extract the data variable as a DataArray
                    if var in ds.data_vars:
                        da = ds[var]
                        
                        # Check and drop height coordinate if it exists
                        if 'height' in da.coords:
                            height_val = da.coords['height'].values
                            print(f"    - Dropping height coordinate (value: {height_val})")
                            da = da.drop_vars('height')
                        
                        # Store the DataArray
                        data_arrays[var] = da
                        print(f"    ‚úì Extracted {var} as DataArray")
                    
                    ds.close()
                else:
                    print(f"  ‚ö† Warning: {filename} not found!")
            
            if len(data_arrays) == 0:
                print(f"  ‚ùå No datasets found for {scenario}_{resolution}")
                failed.append(f"{scenario}_{resolution}")
                continue
            
            # Merge all DataArrays into one Dataset
            print(f"\n  Merging {len(data_arrays)} variables...")
            merged_ds = xr.Dataset(data_arrays)
            
            # Create output filename
            output_name = f"MPI-ESM1-2-{resolution}_{scenario}_r1i1p1f1_{time_range}_allvars.nc"
            output_file = os.path.join(output_dir, output_name)
            
            # Save merged dataset
            print(f"  üíæ Saving to: {output_name}")
            merged_ds.to_netcdf(output_file)
            
            # Print info about the merged file
            print(f"\n  ‚úÖ Success!")
            print(f"     Variables: {list(merged_ds.data_vars.keys())}")
            print(f"     Dimensions: {dict(merged_ds.dims)}")
            print(f"     Coordinates: {list(merged_ds.coords.keys())}")
            print(f"     Time steps: {len(merged_ds.time)}")
            print(f"     Time range: {merged_ds.time.values[0]} to {merged_ds.time.values[-1]}")
            
            successful.append(output_name)
            
            # Clean up
            merged_ds.close()
            del data_arrays, merged_ds
            
        except Exception as e:
            print(f"  ‚ùå Error: {str(e)}")
            import traceback
            traceback.print_exc()
            failed.append(f"{scenario}_{resolution}")

# Print summary
print("\n" + "="*70)
print("MERGE SUMMARY")
print("="*70)
print(f"\n‚úÖ Successfully merged: {len(successful)} files")
if successful:
    for name in successful:
        print(f"   - {name}")

if failed:
    print(f"\n‚ùå Failed: {len(failed)} combinations")
    for name in failed:
        print(f"   - {name}")

print(f"\nüìÅ Output directory: {output_dir}/")
print(f"\nüìä Final result: 8 merged files")
print(f"   - 4 scenarios √ó 2 resolutions")
print(f"   - Each file contains all 4 variables: hurs, pr, sfcWind, tas")
print("="*70)
```

```python
import xarray as xr
import numpy as np
import os

# Set directory
input_dir = "./merged"

# Define scenarios and resolutions
scenarios = ['historical', 'ssp126', 'ssp245', 'ssp585', 'G6sulfur']
resolutions = ['HR', 'LR']
variables = ['pr', 'tas']

print("="*80)
print("NaN CHECK FOR MERGED MULTI-VARIABLE DATASETS")
print("="*80)

# Store results
all_results = []
files_with_nans = []

for scenario in scenarios:
    for resolution in resolutions:
        # Determine time range
        if scenario == 'historical':
            time_range = '1850_2014'
        elif scenario == 'G6sulfur':
            if resolution == 'HR':
                time_range = '2020_2099'
            else:  # LR
                time_range = '2015_2099'
        else:  # ssp126, ssp245, ssp585
            time_range = '2015_2100'

        filename = f"MPI-ESM1-2-{resolution}_{scenario}_r1i1p1f1_{time_range}_allvars.nc"
        filepath = os.path.join(input_dir, filename)
        
        if not os.path.exists(filepath):
            print(f"\n‚ùå MISSING: {filename}")
            continue
        
        print(f"\n{'='*80}")
        print(f"File: {filename}")
        print(f"Scenario: {scenario} | Resolution: {resolution}")
        print(f"{'='*80}")
        
        try:
            # Open dataset
            ds = xr.open_dataset(filepath)
            
            print(f"\nDataset info:")
            print(f"  Variables: {list(ds.data_vars.keys())}")
            print(f"  Dimensions: {dict(ds.dims)}")
            print(f"  Time range: {ds.time.values[0]} to {ds.time.values[-1]}")
            
            # Check each variable for NaNs
            print(f"\n{'‚îÄ'*80}")
            print("NaN Analysis by Variable:")
            print(f"{'‚îÄ'*80}")
            
            file_has_nans = False
            
            for var in variables:
                if var in ds.data_vars:
                    data = ds[var].values
                    
                    # Count NaNs
                    nan_count = np.isnan(data).sum()
                    total_count = data.size
                    nan_percentage = (nan_count / total_count) * 100
                    
                    # Get data statistics
                    if nan_count < total_count:  # If not all NaN
                        data_min = np.nanmin(data)
                        data_max = np.nanmax(data)
                        data_mean = np.nanmean(data)
                    else:
                        data_min = data_max = data_mean = np.nan
                    
                    print(f"\n  Variable: {var}")
                    print(f"    Shape: {ds[var].shape}")
                    print(f"    Total values: {total_count:,}")
                    print(f"    NaN count: {nan_count:,}")
                    print(f"    NaN percentage: {nan_percentage:.6f}%")
                    
                    if nan_count > 0:
                        print(f"    ‚ö†Ô∏è  WARNING: Contains NaN values!")
                        file_has_nans = True
                        
                        # Find where NaNs are located
                        nan_mask = np.isnan(data)
                        nan_time_indices = np.where(nan_mask.any(axis=(1,2)))[0]
                        
                        if len(nan_time_indices) > 0:
                            print(f"    NaN present in {len(nan_time_indices)} time steps")
                            print(f"    First few time indices with NaNs: {nan_time_indices[:5]}")
                            print(f"    Corresponding dates: {ds.time.values[nan_time_indices[:5]]}")
                    else:
                        print(f"    ‚úÖ No NaN values")
                    
                    print(f"    Data range: [{data_min:.6f}, {data_max:.6f}]")
                    print(f"    Mean: {data_mean:.6f}")
                    
                    # Store results
                    all_results.append({
                        'file': filename,
                        'scenario': scenario,
                        'resolution': resolution,
                        'variable': var,
                        'nan_count': nan_count,
                        'nan_percentage': nan_percentage,
                        'total_values': total_count,
                        'has_nans': nan_count > 0
                    })
            
            if file_has_nans:
                files_with_nans.append(filename)
                print(f"\n  üìä FILE STATUS: Contains NaN values in one or more variables")
            else:
                print(f"\n  ‚úÖ FILE STATUS: No NaN values in any variable")
            
            ds.close()
            
        except Exception as e:
            print(f"\n‚ùå ERROR: {str(e)}")
            import traceback
            traceback.print_exc()

# Print summary
print("\n" + "="*80)
print("SUMMARY")
print("="*80)

total_files = len(scenarios) * len(resolutions)
checked_files = len([r for r in all_results if r['variable'] == 'hurs'])  # Count unique files

print(f"\nüìÅ Files checked: {checked_files}/{total_files}")
print(f"‚ö†Ô∏è  Files with NaN values: {len(files_with_nans)}")

if files_with_nans:
    print(f"\nFiles containing NaN values:")
    for fname in files_with_nans:
        print(f"  - {fname}")

# Summary by variable
print(f"\n{'‚îÄ'*80}")
print("NaN Summary by Variable:")
print(f"{'‚îÄ'*80}")

for var in variables:
    var_results = [r for r in all_results if r['variable'] == var]
    files_with_nan = [r for r in var_results if r['has_nans']]
    
    print(f"\n  {var}:")
    print(f"    Files with NaN: {len(files_with_nan)}/{len(var_results)}")
    
    if files_with_nan:
        for r in files_with_nan:
            print(f"      - {r['scenario']}_{r['resolution']}: {r['nan_percentage']:.6f}% NaN")

# Overall statistics
total_checks = len(all_results)
total_with_nans = len([r for r in all_results if r['has_nans']])

print(f"\n{'‚îÄ'*80}")
print(f"Overall: {total_with_nans}/{total_checks} variable-file combinations contain NaN values")
print("="*80)
```

```python
import xarray as xr
import os

# Set directories
input_dir = "./merged"
output_dir = "./merged_converted"
os.makedirs(output_dir, exist_ok=True)

# Define scenarios and resolutions
scenarios = ['historical', 'ssp126', 'ssp245', 'ssp585', 'G6sulfur']
resolutions = ['HR', 'LR']

print("="*80)
print("CONVERTING UNITS: pr ‚Üí mm/day, tas ‚Üí ¬∞C")
print("="*80)

successful = []
failed = []

for scenario in scenarios:
    for resolution in resolutions:
        # Determine time range
        if scenario == 'historical':
            time_range = '1850_2014'
        elif scenario == 'G6sulfur':
            if resolution == 'HR':
                time_range = '2020_2099'
            else:  # LR
                time_range = '2015_2099'
        else:  # ssp126, ssp245, ssp585
            time_range = '2015_2100'

        filename = f"MPI-ESM1-2-{resolution}_{scenario}_r1i1p1f1_{time_range}_allvars.nc"
        filepath = os.path.join(input_dir, filename)
        
        if not os.path.exists(filepath):
            print(f"\n‚ùå MISSING: {filename}")
            failed.append(filename)
            continue
        
        print(f"\n{'='*80}")
        print(f"Processing: {filename}")
        print(f"{'='*80}")
        
        try:
            # Open dataset
            ds = xr.open_dataset(filepath)
            
            # 1. CONVERT PRECIPITATION: kg m-2 s-1 ‚Üí mm/day
            if 'pr' in ds.data_vars:
                print(f"\n  Converting pr: kg m-2 s-1 ‚Üí mm/day")
                
                # Original values
                pr_original = ds['pr']
                print(f"    Original range: [{pr_original.min().values:.6e}, {pr_original.max().values:.6e}] kg m-2 s-1")
                
                # Convert: kg m-2 s-1 * seconds_per_day = mm/day
                # (1 kg/m2 = 1 mm of water)
                seconds_per_day = 86400
                ds['pr'] = pr_original * seconds_per_day
                
                print(f"    Converted range: [{ds['pr'].min().values:.6f}, {ds['pr'].max().values:.6f}] mm/day")
                
                # Update attributes
                ds['pr'].attrs['units'] = 'mm day-1'
                ds['pr'].attrs['long_name'] = 'Precipitation'
                if 'standard_name' in ds['pr'].attrs:
                    ds['pr'].attrs['standard_name'] = 'precipitation_flux'
                ds['pr'].attrs['conversion_note'] = 'Converted from kg m-2 s-1 to mm day-1'
                
                print(f"    ‚úÖ pr converted successfully")
            
            # 2. CONVERT TEMPERATURE: Kelvin ‚Üí Celsius
            if 'tas' in ds.data_vars:
                print(f"\n  Converting tas: K ‚Üí ¬∞C")
                
                # Original values
                tas_original = ds['tas']
                print(f"    Original range: [{tas_original.min().values:.2f}, {tas_original.max().values:.2f}] K")
                
                # Convert: Celsius = Kelvin - 273.15
                ds['tas'] = tas_original - 273.15
                
                print(f"    Converted range: [{ds['tas'].min().values:.2f}, {ds['tas'].max().values:.2f}] ¬∞C")
                
                # Update attributes
                ds['tas'].attrs['units'] = 'degC'
                ds['tas'].attrs['long_name'] = 'Near-Surface Air Temperature'
                if 'standard_name' in ds['tas'].attrs:
                    ds['tas'].attrs['standard_name'] = 'air_temperature'
                ds['tas'].attrs['conversion_note'] = 'Converted from K to degC'
                
                print(f"    ‚úÖ tas converted successfully")
            
            # Save converted dataset
            output_file = os.path.join(output_dir, filename)
            print(f"\n  üíæ Saving to: {output_dir}/{filename}")
            ds.to_netcdf(output_file)
            
            print(f"\n  ‚úÖ Successfully converted and saved!")
            successful.append(filename)
            
            ds.close()
            
        except Exception as e:
            print(f"\n  ‚ùå ERROR: {str(e)}")
            import traceback
            traceback.print_exc()
            failed.append(filename)

# Print summary
print("\n" + "="*80)
print("CONVERSION SUMMARY")
print("="*80)

print(f"\n‚úÖ Successfully converted: {len(successful)}/{len(scenarios) * len(resolutions)} files")
if successful:
    for fname in successful:
        print(f"   - {fname}")

if failed:
    print(f"\n‚ùå Failed: {len(failed)} files")
    for fname in failed:
        print(f"   - {fname}")

print(f"\nüìÅ Output directory: {output_dir}/")
print(f"\nUnit conversions applied:")
print(f"  ‚Ä¢ pr: kg m-2 s-1 ‚Üí mm/day (monthly mean precipitation rate)")
print(f"  ‚Ä¢ tas: K ‚Üí ¬∞C (subtracted 273.15)")
print(f"  ‚Ä¢ hurs: % (unchanged)")
print(f"  ‚Ä¢ sfcWind: m s-1 (unchanged)")
print("="*80)

# Verify one file to show the changes
print("\n" + "="*80)
print("VERIFICATION - Checking one converted file:")
print("="*80)

verify_file = f"MPI-ESM1-2-HR_historical_r1i1p1f1_1850_2014_allvars.nc"
verify_path = os.path.join(output_dir, verify_file)

if os.path.exists(verify_path):
    ds_verify = xr.open_dataset(verify_path)
    
    print(f"\nFile: {verify_file}")
    print(f"\nVariable units:")
    for var in ['hurs', 'pr', 'sfcWind', 'tas']:
        if var in ds_verify.data_vars:
            units = ds_verify[var].attrs.get('units', 'N/A')
            print(f"  {var}: {units}")
    
    ds_verify.close()
```

```python
import xarray as xr
import os

# Set directory
data_dir = "./merged_converted"
output_dir = "./merged_converted_hr_grid"
os.makedirs(output_dir, exist_ok=True)

# Open all datasets
print("="*80)
print("LOADING DATASETS")
print("="*80)

# Historical datasets
ds_HR_historical = xr.open_dataset(os.path.join(data_dir, "MPI-ESM1-2-HR_historical_r1i1p1f1_1850_2014_allvars.nc"))
ds_LR_historical = xr.open_dataset(os.path.join(data_dir, "MPI-ESM1-2-LR_historical_r1i1p1f1_1850_2014_allvars.nc"))

# SSP126 datasets
ds_HR_ssp126 = xr.open_dataset(os.path.join(data_dir, "MPI-ESM1-2-HR_ssp126_r1i1p1f1_2015_2100_allvars.nc"))
ds_LR_ssp126 = xr.open_dataset(os.path.join(data_dir, "MPI-ESM1-2-LR_ssp126_r1i1p1f1_2015_2100_allvars.nc"))

# SSP245 datasets
ds_HR_ssp245 = xr.open_dataset(os.path.join(data_dir, "MPI-ESM1-2-HR_ssp245_r1i1p1f1_2015_2100_allvars.nc"))
ds_LR_ssp245 = xr.open_dataset(os.path.join(data_dir, "MPI-ESM1-2-LR_ssp245_r1i1p1f1_2015_2100_allvars.nc"))

# SSP585 datasets
ds_HR_ssp585 = xr.open_dataset(os.path.join(data_dir, "MPI-ESM1-2-HR_ssp585_r1i1p1f1_2015_2100_allvars.nc"))
ds_LR_ssp585 = xr.open_dataset(os.path.join(data_dir, "MPI-ESM1-2-LR_ssp585_r1i1p1f1_2015_2100_allvars.nc"))

# G6sulfur datasets
ds_HR_g6sulfur = xr.open_dataset(os.path.join(data_dir, "MPI-ESM1-2-HR_G6sulfur_r1i1p1f1_2020_2099_allvars.nc"))
ds_LR_g6sulfur = xr.open_dataset(os.path.join(data_dir, "MPI-ESM1-2-LR_G6sulfur_r1i1p1f1_2015_2099_allvars.nc"))

print("‚úÖ All datasets loaded\n")

# Define the interpolation function
def interpolate_lr_to_hr_grid(lr_data, hr_lat, hr_lon, clip_negative=False):
    """
    Interpolate LR data to HR grid using bilinear interpolation.
    
    Args:
        lr_data: xarray DataArray (low resolution)
        hr_lat: HR latitude coordinates
        hr_lon: HR longitude coordinates
        clip_negative: If True, clip negative values to 0 (for precipitation)
    
    Returns:
        interpolated_data: LR data interpolated to HR grid
    """
    print(f"    Interpolating {lr_data.shape} to HR grid...")
    interpolated = lr_data.interp(
        lat=hr_lat, 
        lon=hr_lon, 
        method='linear', 
        kwargs={'fill_value': 'extrapolate'}
    )
    
    if clip_negative:
        print(f"      Clipping negative values to 0...")
        interpolated = interpolated.clip(min=0)
    
    print(f"      Result shape: {interpolated.shape}")
    return interpolated

# Get HR grid coordinates (use historical HR as reference)
hr_lat = ds_HR_historical.lat.values
hr_lon = ds_HR_historical.lon.values

print(f"HR grid dimensions: lat={len(hr_lat)}, lon={len(hr_lon)}")
print(f"LR grid dimensions: lat={len(ds_LR_historical.lat)}, lon={len(ds_LR_historical.lon)}\n")

# Define which variables need negative clipping
variables_clip_negative = ['pr']  
variables_no_clip = ['tas']  

# Interpolate all LR datasets to HR grid
lr_datasets = {
    'historical': (ds_LR_historical, '1850_2014'),
    'ssp126': (ds_LR_ssp126, '2015_2100'),
    'ssp245': (ds_LR_ssp245, '2015_2100'),
    'ssp585': (ds_LR_ssp585, '2015_2100'),
    'G6sulfur': (ds_LR_g6sulfur, '2015_2099')  
}

print("="*80)
print("INTERPOLATING LR DATASETS TO HR GRID")
print("="*80)

for scenario_name, (lr_ds, time_range) in lr_datasets.items():
    print(f"\n{'='*80}")
    print(f"Processing: {scenario_name}")
    print(f"{'='*80}")
    
    # Create a new dataset to store interpolated variables
    interpolated_vars = {}
    
    # Get available variables in this dataset
    available_vars = [var for var in ['pr', 'tas'] if var in lr_ds.data_vars]
    
    # Interpolate each available variable
    for var in available_vars:
        print(f"\n  Variable: {var}")
        
        # Determine if we need to clip negative values
        clip_negative = var in variables_clip_negative
        
        # Interpolate
        interpolated_vars[var] = interpolate_lr_to_hr_grid(
            lr_ds[var], 
            hr_lat, 
            hr_lon, 
            clip_negative=clip_negative
        )
        
        # Copy attributes
        interpolated_vars[var].attrs = lr_ds[var].attrs.copy()
        interpolated_vars[var].attrs['interpolation_note'] = 'Interpolated from LR to HR grid using bilinear interpolation'
    
    # Create new dataset with interpolated variables
    ds_interpolated = xr.Dataset(interpolated_vars)
    
    # Copy global attributes
    ds_interpolated.attrs = lr_ds.attrs.copy()
    ds_interpolated.attrs['spatial_resolution'] = 'Interpolated to HR grid'
    ds_interpolated.attrs['original_resolution'] = 'LR'
    
    # Save interpolated dataset
    output_filename = f"MPI-ESM1-2-LR-to-HR_{scenario_name}_r1i1p1f1_{time_range}_allvars.nc"
    output_path = os.path.join(output_dir, output_filename)
    
    print(f"\n  üíæ Saving: {output_filename}")
    ds_interpolated.to_netcdf(output_path)
    print(f"  ‚úÖ Saved successfully!")
    
    # Close the interpolated dataset
    ds_interpolated.close()

print("\n" + "="*80)
print("INTERPOLATION COMPLETE")
print("="*80)
print(f"\nüìÅ Output directory: {output_dir}/")
print(f"\nInterpolated 5 LR datasets to HR grid:")
print(f"  - LR historical ‚Üí HR grid")
print(f"  - LR ssp126 ‚Üí HR grid")
print(f"  - LR ssp245 ‚Üí HR grid")
print(f"  - LR ssp585 ‚Üí HR grid")
print(f"  - LR G6sulfur ‚Üí HR grid")

print(f"\nVariables interpolated:")
print(f"  - pr (clipped to ‚â• 0)")
print(f"  - tas (no clipping)")

# Verification: Compare shapes
print("\n" + "="*80)
print("VERIFICATION - Grid Shape Comparison")
print("="*80)

# Load one interpolated file to verify
verify_file = os.path.join(output_dir, "MPI-ESM1-2-LR-to-HR_historical_r1i1p1f1_1850_2014_allvars.nc")
ds_verify = xr.open_dataset(verify_file)

print(f"\nOriginal LR dimensions: {dict(ds_LR_historical.dims)}")
print(f"Original HR dimensions: {dict(ds_HR_historical.dims)}")
print(f"Interpolated LR dimensions: {dict(ds_verify.dims)}")

print(f"\n‚úÖ Interpolation successful! LR data now matches HR grid dimensions.")

ds_verify.close()

# Close all datasets
print("\nClosing datasets...")
for ds in [ds_HR_historical, ds_LR_historical, ds_HR_ssp126, ds_LR_ssp126,
           ds_HR_ssp245, ds_LR_ssp245, ds_HR_ssp585, ds_LR_ssp585,
           ds_HR_g6sulfur, ds_LR_g6sulfur]:  # Added G6sulfur datasets
    ds.close()

print("‚úÖ Done!")
```

```python
import xarray as xr
import os

# Set directories
hr_dir = "./merged_converted"
lr_interp_dir = "./merged_converted_hr_grid"
output_dir = "./merged_hr_lr_combined"
os.makedirs(output_dir, exist_ok=True)

print("="*80)
print("COMBINING HR AND LR-INTERPOLATED DATASETS")
print("="*80)

# Define scenarios with their time ranges
scenarios = {
    'historical': '1850_2014',
    'ssp126': '2015_2100',
    'ssp245': '2015_2100',
    'ssp585': '2015_2100',
    'G6sulfur': '2020_2099'  # Using shared period only
}

for scenario, output_time_range in scenarios.items():
    print(f"\n{'='*80}")
    print(f"Processing: {scenario}")
    print(f"{'='*80}")
    
    # Determine input time ranges (may differ from output for G6sulfur)
    if scenario == 'G6sulfur':
        hr_time_range = '2020_2099'  # HR is 2020-2099
        lr_time_range = '2015_2099'  # LR is 2015-2099
    else:
        hr_time_range = output_time_range
        lr_time_range = output_time_range
    
    # Load HR dataset
    hr_filename = f"MPI-ESM1-2-HR_{scenario}_r1i1p1f1_{hr_time_range}_allvars.nc"
    hr_path = os.path.join(hr_dir, hr_filename)
    
    # Load LR-interpolated dataset
    lr_filename = f"MPI-ESM1-2-LR-to-HR_{scenario}_r1i1p1f1_{lr_time_range}_allvars.nc"
    lr_path = os.path.join(lr_interp_dir, lr_filename)
    
    print(f"\n  Loading HR dataset: {hr_filename}")
    ds_hr = xr.open_dataset(hr_path)
    
    print(f"  Loading LR-interpolated dataset: {lr_filename}")
    ds_lr_interp = xr.open_dataset(lr_path)
    
    # For G6sulfur, subset LR to match HR period (2020-2099)
    if scenario == 'G6sulfur':
        print(f"\n  Subsetting LR data to shared period: 2020-2099")
        print(f"    Original LR time range: {ds_lr_interp.time.values[0]} to {ds_lr_interp.time.values[-1]}")
        
        # Select only 2020-2099 from LR data
        ds_lr_interp = ds_lr_interp.sel(time=slice('2020-01-01', '2099-12-31'))
        
        print(f"    Subset LR time range: {ds_lr_interp.time.values[0]} to {ds_lr_interp.time.values[-1]}")
    
    # Verify coordinates match
    print(f"\n  Verifying coordinates...")
    print(f"    HR dims: {dict(ds_hr.dims)}")
    print(f"    LR-interp dims: {dict(ds_lr_interp.dims)}")
    
    # Verify time dimensions match for G6sulfur
    if scenario == 'G6sulfur':
        if len(ds_hr.time) != len(ds_lr_interp.time):
            print(f"    ‚ö† WARNING: Time dimensions don't match!")
            print(f"      HR time steps: {len(ds_hr.time)}")
            print(f"      LR time steps: {len(ds_lr_interp.time)}")
        else:
            print(f"    ‚úì Time dimensions match: {len(ds_hr.time)} time steps")
    
    # Create combined dataset with renamed LR variables
    print(f"\n  Combining datasets...")
    
    combined_vars = {}
    
    # Get available variables in HR dataset
    available_vars = [var for var in ['pr', 'tas', 'hurs', 'sfcWind'] if var in ds_hr.data_vars]
    
    # Add HR variables with original names
    for var in available_vars:
        combined_vars[f'{var}_hr'] = ds_hr[var]
        print(f"    Added: {var}_hr")
    
    # Add LR-interpolated variables with lr_interp suffix
    for var in available_vars:
        if var in ds_lr_interp.data_vars:
            combined_vars[f'{var}_lr_interp'] = ds_lr_interp[var]
            print(f"    Added: {var}_lr_interp")
    
    # Create combined dataset
    ds_combined = xr.Dataset(combined_vars)
    
    # Copy global attributes from HR dataset
    ds_combined.attrs = ds_hr.attrs.copy()
    ds_combined.attrs['description'] = 'Combined HR and LR-interpolated data on HR grid'
    ds_combined.attrs['hr_source'] = hr_filename
    ds_combined.attrs['lr_interp_source'] = lr_filename
    if scenario == 'G6sulfur':
        ds_combined.attrs['note'] = 'LR data subset to 2020-2099 to match HR period'
    
    # Save combined dataset
    output_filename = f"MPI-ESM1-2-HR-LR_{scenario}_r1i1p1f1_{output_time_range}_allvars.nc"
    output_path = os.path.join(output_dir, output_filename)
    
    print(f"\n  üíæ Saving combined dataset: {output_filename}")
    ds_combined.to_netcdf(output_path)
    
    print(f"\n  ‚úÖ Successfully combined!")
    print(f"     Variables in combined dataset: {list(ds_combined.data_vars.keys())}")
    print(f"     Time range: {ds_combined.time.values[0]} to {ds_combined.time.values[-1]}")
    
    # Close datasets
    ds_hr.close()
    ds_lr_interp.close()
    ds_combined.close()

print("\n" + "="*80)
print("COMBINATION COMPLETE")
print("="*80)

print(f"\nüìÅ Output directory: {output_dir}/")
print(f"\nCreated 5 combined datasets:")
print(f"  1. MPI-ESM1-2-HR-LR_historical_r1i1p1f1_1850_2014_allvars.nc")
print(f"  2. MPI-ESM1-2-HR-LR_ssp126_r1i1p1f1_2015_2100_allvars.nc")
print(f"  3. MPI-ESM1-2-HR-LR_ssp245_r1i1p1f1_2015_2100_allvars.nc")
print(f"  4. MPI-ESM1-2-HR-LR_ssp585_r1i1p1f1_2015_2100_allvars.nc")
print(f"  5. MPI-ESM1-2-HR-LR_G6sulfur_r1i1p1f1_2020_2099_allvars.nc (shared period only)")

print(f"\nHistorical and SSP scenarios contain:")
print(f"  HR variables: pr_hr, hurs_hr, sfcWind_hr, tas_hr")
print(f"  LR-interpolated variables: pr_lr_interp, hurs_lr_interp, sfcWind_lr_interp, tas_lr_interp")

print(f"\nG6sulfur scenario contains (2020-2099 only):")
print(f"  HR variables: pr_hr, tas_hr")
print(f"  LR-interpolated variables: pr_lr_interp, tas_lr_interp")

# Verification
print("\n" + "="*80)
print("VERIFICATION")
print("="*80)

verify_file = os.path.join(output_dir, "MPI-ESM1-2-HR-LR_historical_r1i1p1f1_1850_2014_allvars.nc")
ds_verify = xr.open_dataset(verify_file)

print(f"\nOpened: MPI-ESM1-2-HR-LR_historical_r1i1p1f1_1850_2014_allvars.nc")
print(f"Dimensions: {dict(ds_verify.dims)}")
print(f"Variables: {list(ds_verify.data_vars.keys())}")

ds_verify.close()

# Verify G6sulfur
verify_file_g6 = os.path.join(output_dir, "MPI-ESM1-2-HR-LR_G6sulfur_r1i1p1f1_2020_2099_allvars.nc")
if os.path.exists(verify_file_g6):
    ds_verify_g6 = xr.open_dataset(verify_file_g6)
    
    print(f"\nOpened: MPI-ESM1-2-HR-LR_G6sulfur_r1i1p1f1_2020_2099_allvars.nc")
    print(f"Dimensions: {dict(ds_verify_g6.dims)}")
    print(f"Variables: {list(ds_verify_g6.data_vars.keys())}")
    print(f"Time range: {ds_verify_g6.time.values[0]} to {ds_verify_g6.time.values[-1]}")
    print(f"Time steps: {len(ds_verify_g6.time)}")
    
    ds_verify_g6.close()

print("\n‚úÖ All done! G6sulfur uses shared period 2020-2099 to avoid artifacts.")
```

## ADF TEST

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import ListedColormap


# ADF TEST FUNCTION
def perform_adf_test(data, data_name, scenario_name):
    """Perform Augmented Dickey-Fuller test on time series data"""
    data_clean = data[~np.isnan(data)]
    result = adfuller(data_clean, autolag='AIC')
    
    return {
        'Data Type': data_name,
        'Scenario': scenario_name,
        'ADF Statistic': result[0],
        'p-value': result[1],
        'Stationary': result[1] < 0.05
    }

def get_global_mean(ds):
    """Extract global mean temperature time series"""
    if isinstance(ds, xr.Dataset):
        return ds['tas'].mean(dim=['lat', 'lon']).values
    else:
        return ds.mean(dim=['lat', 'lon']).values

# ============================================================================
# SPLIT HISTORICAL INTO TRAIN/TEST FOR ALL RESOLUTIONS
# ============================================================================

# Original (LR vs HR)
lr_historical_interp_train = lr_historical_interp.sel(time=slice('1850', '1979'))
lr_historical_interp_test = lr_historical_interp.sel(time=slice('1980', '2014'))
detrended_lr_interp_historical_train = detrended_lr_interp_historical.sel(time=slice('1850', '1979'))
detrended_lr_interp_historical_test = detrended_lr_interp_historical.sel(time=slice('1980', '2014'))
hr_historical_train = hr_historical.sel(time=slice('1850', '1979'))
hr_historical_test = hr_historical.sel(time=slice('1980', '2014'))
res_historical_train = res_historical.sel(time=slice('1850', '1979'))
res_historical_test = res_historical.sel(time=slice('1980', '2014'))

# 4x degraded
hr_historical_4x_interp_train = hr_historical_4x_interp.sel(time=slice('1850', '1979'))
hr_historical_4x_interp_test = hr_historical_4x_interp.sel(time=slice('1980', '2014'))
detrended_4x_interp_historical_train = detrended_4x_interp_historical.sel(time=slice('1850', '1979'))
detrended_4x_interp_historical_test = detrended_4x_interp_historical.sel(time=slice('1980', '2014'))
res_historical_4x_train = res_historical_4x.sel(time=slice('1850', '1979'))
res_historical_4x_test = res_historical_4x.sel(time=slice('1980', '2014'))

# 8x degraded
hr_historical_8x_interp_train = hr_historical_8x_interp.sel(time=slice('1850', '1979'))
hr_historical_8x_interp_test = hr_historical_8x_interp.sel(time=slice('1980', '2014'))
detrended_8x_interp_historical_train = detrended_8x_interp_historical.sel(time=slice('1850', '1979'))
detrended_8x_interp_historical_test = detrended_8x_interp_historical.sel(time=slice('1980', '2014'))
res_historical_8x_train = res_historical_8x.sel(time=slice('1850', '1979'))
res_historical_8x_test = res_historical_8x.sel(time=slice('1980', '2014'))

# ============================================================================
# RUN ADF TESTS FOR ALL THREE CASES
# ============================================================================

# Original case (LR vs HR)
results_original = []
test_cases_original = [
    ('LR Bilinear', 'hist_train', lr_historical_interp_train),
    ('LR Detrended', 'hist_train', detrended_lr_interp_historical_train),
    ('HR Raw', 'hist_train', hr_historical_train),
    ('Residual', 'hist_train', res_historical_train),
    
    ('LR Bilinear', 'hist_test', lr_historical_interp_test),
    ('LR Detrended', 'hist_test', detrended_lr_interp_historical_test),
    ('HR Raw', 'hist_test', hr_historical_test),
    ('Residual', 'hist_test', res_historical_test),
    
    ('LR Bilinear', 'G6sulfur', lr_g6sulfur_interp_aligned),
    ('LR Detrended', 'G6sulfur', detrended_lr_interp_g6sulfur),
    ('HR Raw', 'G6sulfur', hr_g6sulfur_aligned),
    ('Residual', 'G6sulfur', res_g6sulfur),
    
    ('LR Bilinear', 'ssp245', lr_ssp245_interp),
    ('LR Detrended', 'ssp245', detrended_lr_interp_ssp245),
    ('HR Raw', 'ssp245', hr_ssp245),
    ('Residual', 'ssp245', res_ssp245),
    
    ('LR Bilinear', 'ssp585', lr_ssp585_interp),
    ('LR Detrended', 'ssp585', detrended_lr_interp_ssp585),
    ('HR Raw', 'ssp585', hr_ssp585),
    ('Residual', 'ssp585', res_ssp585),
]

for data_name, scenario_name, dataset in test_cases_original:
    global_mean = get_global_mean(dataset)
    result = perform_adf_test(global_mean, data_name, scenario_name)
    results_original.append(result)

# 4x case
results_4x = []
test_cases_4x = [
    ('4x Bilinear', 'hist_train', hr_historical_4x_interp_train),
    ('4x Detrended', 'hist_train', detrended_4x_interp_historical_train),
    ('HR Raw', 'hist_train', hr_historical_train),
    ('Residual', 'hist_train', res_historical_4x_train),
    
    ('4x Bilinear', 'hist_test', hr_historical_4x_interp_test),
    ('4x Detrended', 'hist_test', detrended_4x_interp_historical_test),
    ('HR Raw', 'hist_test', hr_historical_test),
    ('Residual', 'hist_test', res_historical_4x_test),
    
    ('4x Bilinear', 'G6sulfur', hr_g6sulfur_4x_interp),
    ('4x Detrended', 'G6sulfur', detrended_4x_interp_g6sulfur),
    ('HR Raw', 'G6sulfur', hr_g6sulfur_aligned),
    ('Residual', 'G6sulfur', res_g6sulfur_4x),
    
    ('4x Bilinear', 'ssp245', hr_ssp245_4x_interp),
    ('4x Detrended', 'ssp245', detrended_4x_interp_ssp245),
    ('HR Raw', 'ssp245', hr_ssp245),
    ('Residual', 'ssp245', res_ssp245_4x),
    
    ('4x Bilinear', 'ssp585', hr_ssp585_4x_interp),
    ('4x Detrended', 'ssp585', detrended_4x_interp_ssp585),
    ('HR Raw', 'ssp585', hr_ssp585),
    ('Residual', 'ssp585', res_ssp585_4x),
]

for data_name, scenario_name, dataset in test_cases_4x:
    global_mean = get_global_mean(dataset)
    result = perform_adf_test(global_mean, data_name, scenario_name)
    results_4x.append(result)

# 8x case
results_8x = []
test_cases_8x = [
    ('8x Bilinear', 'hist_train', hr_historical_8x_interp_train),
    ('8x Detrended', 'hist_train', detrended_8x_interp_historical_train),
    ('HR Raw', 'hist_train', hr_historical_train),
    ('Residual', 'hist_train', res_historical_8x_train),
    
    ('8x Bilinear', 'hist_test', hr_historical_8x_interp_test),
    ('8x Detrended', 'hist_test', detrended_8x_interp_historical_test),
    ('HR Raw', 'hist_test', hr_historical_test),
    ('Residual', 'hist_test', res_historical_8x_test),
    
    ('8x Bilinear', 'G6sulfur', hr_g6sulfur_8x_interp),
    ('8x Detrended', 'G6sulfur', detrended_8x_interp_g6sulfur),
    ('HR Raw', 'G6sulfur', hr_g6sulfur_aligned),
    ('Residual', 'G6sulfur', res_g6sulfur_8x),
    
    ('8x Bilinear', 'ssp245', hr_ssp245_8x_interp),
    ('8x Detrended', 'ssp245', detrended_8x_interp_ssp245),
    ('HR Raw', 'ssp245', hr_ssp245),
    ('Residual', 'ssp245', res_ssp245_8x),
    
    ('8x Bilinear', 'ssp585', hr_ssp585_8x_interp),
    ('8x Detrended', 'ssp585', detrended_8x_interp_ssp585),
    ('HR Raw', 'ssp585', hr_ssp585),
    ('Residual', 'ssp585', res_ssp585_8x),
]

for data_name, scenario_name, dataset in test_cases_8x:
    global_mean = get_global_mean(dataset)
    result = perform_adf_test(global_mean, data_name, scenario_name)
    results_8x.append(result)

# Create DataFrames
df_results_original = pd.DataFrame(results_original)
df_results_4x = pd.DataFrame(results_4x)
df_results_8x = pd.DataFrame(results_8x)

# ============================================================================
# CREATE 1x3 PUBLICATION-QUALITY HEATMAP
# ============================================================================

scenarios_order = ['hist_train', 'hist_test', 'G6sulfur', 'ssp245', 'ssp585']
data_types_order_original = ['LR Bilinear', 'LR Detrended', 'HR Raw', 'Residual']
data_types_order_4x = ['4x Bilinear', '4x Detrended', 'HR Raw', 'Residual']
data_types_order_8x = ['8x Bilinear', '8x Detrended', 'HR Raw', 'Residual']

# Create pivot tables for all three cases
pivot_pvalue_original = df_results_original.pivot(index='Data Type', columns='Scenario', values='p-value')
pivot_pvalue_original = pivot_pvalue_original.reindex(index=data_types_order_original, columns=scenarios_order)
pivot_stationary_original = df_results_original.pivot(index='Data Type', columns='Scenario', values='Stationary')
pivot_stationary_original = pivot_stationary_original.reindex(index=data_types_order_original, columns=scenarios_order)

pivot_pvalue_4x = df_results_4x.pivot(index='Data Type', columns='Scenario', values='p-value')
pivot_pvalue_4x = pivot_pvalue_4x.reindex(index=data_types_order_4x, columns=scenarios_order)
pivot_stationary_4x = df_results_4x.pivot(index='Data Type', columns='Scenario', values='Stationary')
pivot_stationary_4x = pivot_stationary_4x.reindex(index=data_types_order_4x, columns=scenarios_order)

pivot_pvalue_8x = df_results_8x.pivot(index='Data Type', columns='Scenario', values='p-value')
pivot_pvalue_8x = pivot_pvalue_8x.reindex(index=data_types_order_8x, columns=scenarios_order)
pivot_stationary_8x = df_results_8x.pivot(index='Data Type', columns='Scenario', values='Stationary')
pivot_stationary_8x = pivot_stationary_8x.reindex(index=data_types_order_8x, columns=scenarios_order)

# Set publication quality settings
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans']
plt.rcParams['mathtext.default'] = 'regular'

# Create figure with 1x3 subplots
fig, axes = plt.subplots(1, 3, figsize=(24, 8))

# Publication-quality colormap
cmap = ListedColormap(['#E57373', '#81C784'])

# Panel titles
panel_titles = ['(a) LR Sim vs HR Sim', '(b) 4x Degraded HR vs HR Sim', '(c) 8x Degraded HR vs HR Sim']
pivots = [
    (pivot_pvalue_original, pivot_stationary_original, data_types_order_original),
    (pivot_pvalue_4x, pivot_stationary_4x, data_types_order_4x),
    (pivot_pvalue_8x, pivot_stationary_8x, data_types_order_8x)
]

for idx, (ax, title, (pivot_pvalue, pivot_stationary, data_types)) in enumerate(zip(axes, panel_titles, pivots)):
    # Create color array
    color_data = pivot_stationary.astype(int).values
    
    # Create heatmap
    im = ax.imshow(color_data, cmap=cmap, aspect='auto', alpha=1)
    
    # Add p-values as text
    for i in range(len(data_types)):
        for j in range(len(scenarios_order)):
            p_val = pivot_pvalue.iloc[i, j]
            
            # Format p-value
            if p_val < 0.0001:
                text = '<0.0001'
            elif p_val < 0.001:
                text = f'{p_val:.4f}'
            else:
                text = f'{p_val:.3f}'
            
            # Text color based on background
            text_color = 'white' if color_data[i, j] == 0 else 'black'
            
            ax.text(j, i, text, ha='center', va='center', 
                    fontsize=11, fontweight='normal', color=text_color,
                    family='monospace')
    
    # Set ticks and labels
    ax.set_xticks(range(len(scenarios_order)))
    ax.set_yticks(range(len(data_types)))
    ax.set_xticklabels(scenarios_order, fontsize=11)
    ax.set_yticklabels(data_types, fontsize=11)
    
    # Title
    ax.set_title(title, fontsize=13, fontweight='bold', pad=10)
    
    # Add subtle grid
    ax.set_xticks([x - 0.5 for x in range(1, len(scenarios_order))], minor=True)
    ax.set_yticks([y - 0.5 for y in range(1, len(data_types))], minor=True)
    ax.grid(which='minor', color='white', linestyle='-', linewidth=2.5)
    
    # Remove spines
    for spine in ax.spines.values():
        spine.set_visible(False)

# Add main title
fig.suptitle('Augmented Dickey-Fuller Test for Stationarity', 
             fontsize=16, fontweight='bold', y=1.02)

# Create legend at bottom center
green_patch = mpatches.Patch(color='#81C784', label='Stationary (p < 0.05)')
red_patch = mpatches.Patch(color='#E57373', label='Non-stationary (p ‚â• 0.05)')
fig.legend(handles=[green_patch, red_patch], 
          loc='lower center', 
          bbox_to_anchor=(0.5, -0.1),
          ncol=2,
          fontsize=12, 
          frameon=False,
          fancybox=False,
          shadow=False,
          edgecolor='gray')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from scipy.stats import wasserstein_distance


def compute_wasserstein(data1, data2):
    """Compute Wasserstein distance between two datasets."""
    vals1 = data1['tas'].values.flatten() if 'tas' in data1 else data1.values.flatten()
    vals2 = data2['tas'].values.flatten() if 'tas' in data2 else data2.values.flatten()
    return wasserstein_distance(vals1, vals2)



train_period = slice('1850', '1980')

# ============================================================================
# ORIGINAL (LR vs HR)
# ============================================================================
# Raw inputs
train_input_raw = lr_historical_interp.sel(time=train_period)

# Detrended inputs
train_input_detrend = detrended_lr_interp_historical.sel(time=train_period)

# Raw outputs (HR)
train_output_hr = hr_historical.sel(time=train_period)

# Residual outputs (HR - LR_interp)
train_output_res = res_historical.sel(time=train_period)

# ============================================================================
# 4X DEGRADED
# ============================================================================
# Raw inputs (4x degraded interpolated)
train_input_raw_4x = hr_historical_4x_interp.sel(time=train_period)

# Detrended inputs (4x)
train_input_detrend_4x = detrended_4x_interp_historical.sel(time=train_period)

# Residual outputs (HR - 4x_interp)
train_output_res_4x = res_historical_4x.sel(time=train_period)

# ============================================================================
# 8X DEGRADED
# ============================================================================
# Raw inputs (8x degraded interpolated)
train_input_raw_8x = hr_historical_8x_interp.sel(time=train_period)

# Detrended inputs (8x)
train_input_detrend_8x = detrended_8x_interp_historical.sel(time=train_period)

# Residual outputs (HR - 8x_interp)
train_output_res_8x = res_historical_8x.sel(time=train_period)


def load_test_data_local(scenario, resolution='original'):
    """Load test data for a given scenario and resolution from local variables."""
    test_data = {}
    
    if scenario == 'historical':
        test_period = slice('2001', '2014')
        
        if resolution == 'original':
            # Raw input
            test_data['input_raw'] = lr_historical_interp.sel(time=test_period)
            # Detrended input
            test_data['input_detrend'] = detrended_lr_interp_historical.sel(time=test_period)
            # Raw output (HR)
            test_data['output_hr'] = hr_historical.sel(time=test_period)
            # Residual output
            test_data['output_res'] = res_historical.sel(time=test_period)
            
        elif resolution == '4x':
            test_data['input_raw'] = hr_historical_4x_interp.sel(time=test_period)
            test_data['input_detrend'] = detrended_4x_interp_historical.sel(time=test_period)
            test_data['output_hr'] = hr_historical.sel(time=test_period)
            test_data['output_res'] = res_historical_4x.sel(time=test_period)
            
        elif resolution == '8x':
            test_data['input_raw'] = hr_historical_8x_interp.sel(time=test_period)
            test_data['input_detrend'] = detrended_8x_interp_historical.sel(time=test_period)
            test_data['output_hr'] = hr_historical.sel(time=test_period)
            test_data['output_res'] = res_historical_8x.sel(time=test_period)
        
    elif scenario == 'g6sulfur':
        if resolution == 'original':
            test_data['input_raw'] = lr_g6sulfur_interp_aligned
            test_data['input_detrend'] = detrended_lr_interp_g6sulfur
            test_data['output_hr'] = hr_g6sulfur_aligned
            test_data['output_res'] = res_g6sulfur
            
        elif resolution == '4x':
            test_data['input_raw'] = hr_g6sulfur_4x_interp
            test_data['input_detrend'] = detrended_4x_interp_g6sulfur
            test_data['output_hr'] = hr_g6sulfur_aligned
            test_data['output_res'] = res_g6sulfur_4x
            
        elif resolution == '8x':
            test_data['input_raw'] = hr_g6sulfur_8x_interp
            test_data['input_detrend'] = detrended_8x_interp_g6sulfur
            test_data['output_hr'] = hr_g6sulfur_aligned
            test_data['output_res'] = res_g6sulfur_8x
        
    elif scenario == 'ssp245':
        if resolution == 'original':
            test_data['input_raw'] = lr_ssp245_interp
            test_data['input_detrend'] = detrended_lr_interp_ssp245
            test_data['output_hr'] = hr_ssp245
            test_data['output_res'] = res_ssp245
            
        elif resolution == '4x':
            test_data['input_raw'] = hr_ssp245_4x_interp
            test_data['input_detrend'] = detrended_4x_interp_ssp245
            test_data['output_hr'] = hr_ssp245
            test_data['output_res'] = res_ssp245_4x
            
        elif resolution == '8x':
            test_data['input_raw'] = hr_ssp245_8x_interp
            test_data['input_detrend'] = detrended_8x_interp_ssp245
            test_data['output_hr'] = hr_ssp245
            test_data['output_res'] = res_ssp245_8x
        
    elif scenario == 'ssp585':
        if resolution == 'original':
            test_data['input_raw'] = lr_ssp585_interp
            test_data['input_detrend'] = detrended_lr_interp_ssp585
            test_data['output_hr'] = hr_ssp585
            test_data['output_res'] = res_ssp585
            
        elif resolution == '4x':
            test_data['input_raw'] = hr_ssp585_4x_interp
            test_data['input_detrend'] = detrended_4x_interp_ssp585
            test_data['output_hr'] = hr_ssp585
            test_data['output_res'] = res_ssp585_4x
            
        elif resolution == '8x':
            test_data['input_raw'] = hr_ssp585_8x_interp
            test_data['input_detrend'] = detrended_8x_interp_ssp585
            test_data['output_hr'] = hr_ssp585
            test_data['output_res'] = res_ssp585_8x
    
    return test_data


# ============================================================================
# COMPUTE WASSERSTEIN DISTANCES FOR ALL THREE RESOLUTIONS
# ============================================================================

model_names = ['raw_raw', 'raw_res', 'detrend_res']
scenarios = ['historical', 'g6sulfur', 'ssp245', 'ssp585']
resolutions = ['original', '4x', '8x']

# Define model configurations (input_type, output_type)
model_config = {
    'raw_raw': ('raw', 'hr'),
    'raw_res': ('raw', 'res'),
    'detrend_res': ('detrend', 'res')
}

# Store results for all resolutions
wasserstein_results = {}

for resolution in resolutions:

    wasserstein_data = {model: {'scenarios': [], 'cov_wass': [], 'tgt_wass': []} 
                        for model in model_names}
    
    # Select appropriate training data based on resolution
    if resolution == 'original':
        train_input_raw_res = train_input_raw
        train_input_detrend_res = train_input_detrend
        train_output_hr_res = train_output_hr
        train_output_res_res = train_output_res
    elif resolution == '4x':
        train_input_raw_res = train_input_raw_4x
        train_input_detrend_res = train_input_detrend_4x
        train_output_hr_res = train_output_hr  # HR is always the same
        train_output_res_res = train_output_res_4x
    elif resolution == '8x':
        train_input_raw_res = train_input_raw_8x
        train_input_detrend_res = train_input_detrend_8x
        train_output_hr_res = train_output_hr  # HR is always the same
        train_output_res_res = train_output_res_8x
    
    for model_name in model_names:
        input_type, output_type = model_config[model_name]
        
        # Get appropriate training data
        if input_type == 'raw':
            train_input = train_input_raw_res
        else:
            train_input = train_input_detrend_res
        
        if output_type == 'hr':
            train_output = train_output_hr_res
        else:  # res
            train_output = train_output_res_res
        
        for scenario in scenarios:
            # Load test data
            test_data = load_test_data_local(scenario, resolution=resolution)
            
            # Get test input/output
            test_input = test_data[f'input_{input_type}']
            test_output = test_data[f'output_{output_type}']
            
            # Compute covariate shifts (INPUT)
            cov_wass = compute_wasserstein(train_input, test_input)
            
            # Compute target shifts (OUTPUT)
            tgt_wass = compute_wasserstein(train_output, test_output)
            
            # Store results
            wasserstein_data[model_name]['scenarios'].append(scenario)
            wasserstein_data[model_name]['cov_wass'].append(cov_wass)
            wasserstein_data[model_name]['tgt_wass'].append(tgt_wass)  
    
    wasserstein_results[resolution] = wasserstein_data


# ============================================================================
# PLOT: 3 ROWS √ó 3 COLUMNS (Wasserstein Distance)
# ============================================================================

model_names_subset = ['raw_raw', 'raw_res', 'detrend_res']

# Better, more descriptive titles
model_titles = {
    'raw_raw': 'Direct Prediction',
    'raw_res': 'Residual Learning',
    'detrend_res': 'Stationary Downscaling'
}

row_titles = {
    'original': 'Different Model Resolutions',
    '4x': '4√ó Coarsened Synthetic LR',
    '8x': '8√ó Coarsened Synthetic LR'
}

fig, axes = plt.subplots(3, 3, figsize=(16, 16))

# Scenario styling
scenario_markers = {'historical': 'o', 'g6sulfur': 's', 'ssp245': 'D', 'ssp585': '^'}
scenario_colors = {'historical': '#1f77b4', 'g6sulfur': '#ff7f0e', 'ssp245': '#2ca02c', 'ssp585': '#d62728'}
scenario_labels = {'historical': 'Historical', 'g6sulfur': 'G6sulfur', 'ssp245': 'SSP2-4.5', 'ssp585': 'SSP5-8.5'}

# Plot each resolution (row) and model (column)
for row_idx, resolution in enumerate(resolutions):
    wasserstein_data_res = wasserstein_results[resolution]
    
    for col_idx, model_name in enumerate(model_names_subset):
        ax = axes[row_idx, col_idx]
        model_data = wasserstein_data_res[model_name]
        
        # Plot points for each scenario
        for i, scenario in enumerate(model_data['scenarios']):
            x_val = model_data['cov_wass'][i]
            y_val = model_data['tgt_wass'][i]
            
            ax.scatter(x_val, y_val,
                      s=250, 
                      c=scenario_colors[scenario],
                      marker=scenario_markers[scenario],
                      alpha=1.0,
                      edgecolors='black',
                      linewidth=2.5,
                      label=scenario_labels[scenario] if row_idx == 0 and col_idx == 0 else None)
        
        # Set shared axis limits
        ax.set_xlim(0, 3)
        ax.set_ylim(0, 3)
        
        # Add diagonal reference line (x=y)
        ax.plot([0, 3], [0, 3], 'k--', alpha=0.3, linewidth=2, zorder=1)
        
        # Column title (only for first row)
        if row_idx == 0:
            ax.set_title(model_titles[model_name], fontsize=18, fontweight='bold', pad=30)
        
        # Row label (only for first column)
        if col_idx == 0:
            ax.text(-0.35, 0.5, row_titles[resolution], 
                   transform=ax.transAxes, fontsize=18, fontweight='bold',
                   rotation=90, va='center', ha='center')
        
        # Axis labels
        if row_idx == 2:  # Bottom row
            ax.set_xlabel('Covariate Shift (K)', fontsize=16)
        if col_idx == 0:  # Left column
            ax.set_ylabel('Label Shift (K)', fontsize=16)
        
        # Grid and ticks
        ax.grid(alpha=0.25, linestyle='--', linewidth=1)
        ax.tick_params(labelsize=14)
        
        # Remove top and right spines
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)

# Add legend to bottom center
handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels,
          loc='lower center', 
          bbox_to_anchor=(0.5, -0.05), 
          ncol=4, 
          fontsize=16, 
          frameon=False)

plt.tight_layout()
plt.subplots_adjust(left=0.08, bottom=0.05)  
plt.show()
```
