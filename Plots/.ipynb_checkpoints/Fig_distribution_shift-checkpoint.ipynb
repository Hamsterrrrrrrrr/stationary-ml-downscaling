```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import pickle
from pathlib import Path

# ----------------------------
# Load data
# ----------------------------
data_dir = "../data"
figures_dir = Path("../Figures")

# Original datasets
ds_hist = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_historical_r1i1p1f1_1850_2014_allvars.nc")
ds_ssp245 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp245_r1i1p1f1_2015_2100_allvars.nc")
ds_ssp585 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp585_r1i1p1f1_2015_2100_allvars.nc")

# Detrended datasets
ds_hist_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_historical_residual_detrended.nc")
ds_ssp245_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp245_residual_detrended.nc")
ds_ssp585_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp585_residual_detrended.nc")

# Load normalization statistics
with open(f"{data_dir}/norm_stats.pkl", 'rb') as f:
    norm_stats = pickle.load(f)
    
# Load detrended normalization statistics
with open(f"{data_dir}/norm_stats_zscore_pixel_residual_detrended.pkl", 'rb') as f:
    norm_stats_detrend = pickle.load(f)

# Training period
train_start = '1850'
train_end = '1980'

# ----------------------------
# Histogram function (normalized to area = 1)
# ----------------------------
def histogram_line(values, bins=100):
    """Compute histogram and return line coordinates (area = 1)"""
    values = values[np.isfinite(values)]
    if values.size == 0:
        x = np.linspace(-1, 1, bins)
        return x, np.zeros(bins)
    
    counts, edges = np.histogram(values, bins=bins, density=True)
    centers = 0.5 * (edges[:-1] + edges[1:])
    
    return centers, counts

# ----------------------------
# Normalization function
# ----------------------------
def apply_normalization(data, method, norm_stats, var_base='tas'):
    """Apply normalization method to data"""
    if method == 'none':
        return data
    elif method == 'minmax_global':
        data_min = norm_stats[var_base]['lr_interp']['global_min']
        data_max = norm_stats[var_base]['lr_interp']['global_max']
        return 2 * (data - data_min) / (data_max - data_min + 1e-8) - 1
    elif method == 'minmax_pixel':
        data_min = norm_stats[var_base]['lr_interp']['pixel_min']
        data_max = norm_stats[var_base]['lr_interp']['pixel_max']
        return 2 * (data - data_min) / (data_max - data_min + 1e-8) - 1
    elif method == 'zscore_global':
        data_mean = norm_stats[var_base]['lr_interp']['global_mean']
        data_std = norm_stats[var_base]['lr_interp']['global_std']
        return (data - data_mean) / (data_std + 1e-8)
    elif method == 'zscore_pixel':
        data_mean = norm_stats[var_base]['lr_interp']['pixel_mean']
        data_std = norm_stats[var_base]['lr_interp']['pixel_std']
        return (data - data_mean) / (data_std + 1e-8)
    elif method == 'instance_zscore':
        return (data - data.mean(dim=['lat', 'lon'])) / (data.std(dim=['lat', 'lon']) + 1e-8)
    elif method == 'instance_minmax':
        data_min = data.min(dim=['lat', 'lon'])
        data_max = data.max(dim=['lat', 'lon'])
        return 2 * (data - data_min) / (data_max - data_min + 1e-8) - 1
    else:
        raise ValueError(f"Unknown method: {method}")

# ----------------------------
# Prepare data for each scenario
# ----------------------------
scenarios = {
    'Historical': ds_hist['tas_lr_interp'].sel(time=slice(train_start, train_end)),
    'SSP2-4.5': ds_ssp245['tas_lr_interp'],
    'SSP5-8.5': ds_ssp585['tas_lr_interp']
}

scenarios_detrend = {
    'Historical': ds_hist_detrend['tas_lr_detrend_grid'].sel(time=slice(train_start, train_end)),
    'SSP2-4.5': ds_ssp245_detrend['tas_lr_detrend_grid'],
    'SSP5-8.5': ds_ssp585_detrend['tas_lr_detrend_grid']
}

methods = ['minmax_global', 'minmax_pixel', 'zscore_global', 'zscore_pixel', 'instance_zscore', 'instance_minmax']
method_names = ['MM-G', 'MM-P', 'ZS-G', 'ZS-P', 'Inst-ZS', 'Inst-MM']

colors = {'Historical': '#1f77b4', 'SSP2-4.5': '#ff7f0e', 'SSP5-8.5': '#d62728'}

# X-axis limits for each panel
xlims = [
    [-80, -60, -40, -20, 0, 20, 40, 60],
    [-1.2, -0.6, 0.0, 0.6, 1.2],
    [-2, -1, 0, 1, 2, 3, 4, 5],
    [-4, -3, -2, -1, 0, 1, 2, 3],
    [-8, -4, 0, 4, 8, 12, 16],
    [-4, -3, -2, -1, 0, 1, 2],
    [-1, -0.5, 0, 0.5, 1],
    [-15, -10, -5, 0, 5, 10, 15]
]

# ----------------------------
# Create 2x4 plot with increased spacing
# ----------------------------
fig, axes = plt.subplots(2, 4, figsize=(20, 10))

panel_idx = 0

# Row 1: Original lr_interp
# Panel 1: Raw distributions (3 scenarios) - Already in C
ax = axes[0, 0]
for name, data in scenarios.items():
    vals = data.values.ravel()  # Data already in C
    x, y = histogram_line(vals, bins=100)
    ax.plot(x, y, linewidth=2, label=name, color=colors[name])
ax.set_xlabel('Temperature (C)', fontsize=14, fontweight='bold')
ax.set_ylabel('Histogram', fontsize=14, fontweight='bold')
ax.set_title('None', fontsize=16, fontweight='bold', pad=10)
ax.set_xticks(xlims[panel_idx])
ax.set_xlim(xlims[panel_idx][0], xlims[panel_idx][-1])
ax.legend(frameon=False, fontsize=12)
ax.tick_params(axis='both', which='major', labelsize=12, length=6, width=1.5)
panel_idx += 1

# Panels 2-4: After first 3 normalization methods
for col_idx, (method, method_name) in enumerate(zip(methods[:3], method_names[:3]), start=1):
    ax = axes[0, col_idx]
    for name, data in scenarios.items():
        data_norm = apply_normalization(data, method, norm_stats, 'tas')
        vals = data_norm.values.ravel()
        x, y = histogram_line(vals, bins=100)
        ax.plot(x, y, linewidth=2, label=name, color=colors[name])
    ax.set_xlabel('Normalized Temperature', fontsize=14, fontweight='bold')
    ax.set_ylabel('Histogram', fontsize=14, fontweight='bold')
    ax.set_title(method_name, fontsize=16, fontweight='bold', pad=10)
    ax.set_xticks(xlims[panel_idx])
    ax.set_xlim(xlims[panel_idx][0], xlims[panel_idx][-1])
    ax.legend(frameon=False, fontsize=12)
    ax.tick_params(axis='both', which='major', labelsize=12, length=6, width=1.5)
    panel_idx += 1

# Row 2: Continue with remaining normalizations + detrended
for col_idx, (method, method_name) in enumerate(zip(methods[3:], method_names[3:])):
    ax = axes[1, col_idx]
    for name, data in scenarios.items():
        data_norm = apply_normalization(data, method, norm_stats, 'tas')
        vals = data_norm.values.ravel()
        x, y = histogram_line(vals, bins=100)
        ax.plot(x, y, linewidth=2, label=name, color=colors[name])
    ax.set_xlabel('Normalized Temperature', fontsize=14, fontweight='bold')
    ax.set_ylabel('Histogram', fontsize=14, fontweight='bold')
    ax.set_title(method_name, fontsize=16, fontweight='bold', pad=10)
    ax.set_xticks(xlims[panel_idx])
    ax.set_xlim(xlims[panel_idx][0], xlims[panel_idx][-1])
    ax.legend(frameon=False, fontsize=12)
    ax.tick_params(axis='both', which='major', labelsize=12, length=6, width=1.5)
    panel_idx += 1

# Panel 8: After pointwise linear detrending + Z-score Pixel
ax = axes[1, 3]
for name, data in scenarios_detrend.items():
    # Apply z-score pixel normalization to detrended data
    data_mean = norm_stats_detrend['tas']['lr_detrend_grid']['pixel_mean']
    data_std = norm_stats_detrend['tas']['lr_detrend_grid']['pixel_std']
    data_norm = (data - data_mean) / (data_std + 1e-8)
    
    vals = data_norm.values.ravel()
    x, y = histogram_line(vals, bins=100)
    ax.plot(x, y, linewidth=2, label=name, color=colors[name])
ax.set_xlabel('Normalized Temperature', fontsize=14, fontweight='bold')
ax.set_ylabel('Histogram', fontsize=14, fontweight='bold')
ax.set_title('PLD+ZS-P', fontsize=16, fontweight='bold', pad=10)
ax.set_xticks(xlims[panel_idx])
ax.set_xlim(xlims[panel_idx][0], xlims[panel_idx][-1])
ax.legend(frameon=False, fontsize=12)
ax.tick_params(axis='both', which='major', labelsize=12, length=6, width=1.5)

plt.tight_layout(h_pad=5.0)  # Add vertical padding between rows

output_file = figures_dir / "distribution_shift.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight')

plt.show()
```

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import pickle
from pathlib import Path

# Load data
data_dir = "../data"
figures_dir = Path("../Figures")

# Original datasets
ds_hist = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_historical_r1i1p1f1_1850_2014_allvars.nc")
ds_g6 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_g6sulfur_r1i1p1f1_2020_2099_allvars.nc")
ds_ssp585 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp585_r1i1p1f1_2015_2100_allvars.nc")

# Detrended datasets
ds_hist_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_historical_residual_detrended.nc")
ds_g6_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_g6sulfur_residual_detrended.nc")
ds_ssp585_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp585_residual_detrended.nc")

# Load normalization statistics
with open(f"{data_dir}/norm_stats.pkl", 'rb') as f:
    norm_stats = pickle.load(f)
    
# Load detrended normalization statistics
with open(f"{data_dir}/norm_stats_zscore_pixel_residual_detrended.pkl", 'rb') as f:
    norm_stats_detrend = pickle.load(f)

# Training period
train_start = '1850'
train_end = '1980'

# Prepare data
scenarios_raw = {
    'Historical': ds_hist['tas_lr_interp'].sel(time=slice(train_start, train_end)),
    'G6sulfur': ds_g6['tas_lr_interp'],
    'SSP5-8.5': ds_ssp585['tas_lr_interp']
}

scenarios_gmc = {
    'Historical': ds_hist_detrend['tas_lr_detrend_gma'].sel(time=slice(train_start, train_end)),
    'G6sulfur': ds_g6_detrend['tas_lr_detrend_gma'],
    'SSP5-8.5': ds_ssp585_detrend['tas_lr_detrend_gma']
}

scenarios_gmld = {
    'Historical': ds_hist_detrend['tas_lr_detrend_gmt'].sel(time=slice(train_start, train_end)),
    'G6sulfur': ds_g6_detrend['tas_lr_detrend_gmt'],
    'SSP5-8.5': ds_ssp585_detrend['tas_lr_detrend_gmt']
}

scenarios_pld = {
    'Historical': ds_hist_detrend['tas_lr_detrend_grid'].sel(time=slice(train_start, train_end)),
    'G6sulfur': ds_g6_detrend['tas_lr_detrend_grid'],
    'SSP5-8.5': ds_ssp585_detrend['tas_lr_detrend_grid']
}

colors = {'Historical': '#1f77b4', 'G6sulfur': '#2ca02c', 'SSP5-8.5': '#d62728'}

# Create 2x3 plot
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Panel labels
panel_labels = ['(a) Raw', '(b) ZS-G', '(c) ZS-P', '(d) GMC', '(e) GLD', '(f) PLD']

# Helper function to plot histograms with rectangular bins
def plot_histograms(ax, scenarios_dict, panel_label, show_legend=False):
    # Plot Historical first with filled bars
    data = scenarios_dict['Historical']
    vals = data.values.ravel()
    vals = vals[np.isfinite(vals)]
    ax.hist(vals, bins=100, density=True, alpha=0.5, color=colors['Historical'], 
            label='Historical', edgecolor='none')
    
    # Plot G6sulfur and SSP5-8.5 as histogram outlines (step)
    for name in ['G6sulfur', 'SSP5-8.5']:
        data = scenarios_dict[name]
        vals = data.values.ravel()
        vals = vals[np.isfinite(vals)]
        ax.hist(vals, bins=60, density=True, histtype='step', linewidth=2,
                label=name, color=colors[name])
    
    # Remove y-axis only
    ax.set_yticks([])
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    # Add panel label in upper left corner
    ax.text(0.05, 0.95, panel_label, transform=ax.transAxes,
            fontsize=20, fontweight='bold', va='top', ha='left')
    
    # Only show legend for specified panel
    if show_legend:
        ax.legend(frameon=False, fontsize=16, loc='upper right')
    
    # Keep x-axis ticks and labels visible
    ax.tick_params(axis='x', which='major', labelsize=16)

# Row 1: None, ZS-G, ZS-P
# Panel (0,0): None
ax = axes[0, 0]
plot_histograms(ax, scenarios_raw, panel_labels[0])

# Panel (0,1): ZS-G
ax = axes[0, 1]
scenarios_zsg = {}
for name, data in scenarios_raw.items():
    data_mean = norm_stats['tas']['lr_interp']['global_mean']
    data_std = norm_stats['tas']['lr_interp']['global_std']
    scenarios_zsg[name] = (data - data_mean) / (data_std + 1e-8)
plot_histograms(ax, scenarios_zsg, panel_labels[1])

# Panel (0,2): ZS-P
ax = axes[0, 2]
scenarios_zsp = {}
for name, data in scenarios_raw.items():
    data_mean = norm_stats['tas']['lr_interp']['pixel_mean']
    data_std = norm_stats['tas']['lr_interp']['pixel_std']
    scenarios_zsp[name] = (data - data_mean) / (data_std + 1e-8)
plot_histograms(ax, scenarios_zsp, panel_labels[2])

# Row 2: GMC, GMLD, PLD
# Panel (1,0): GMC
ax = axes[1, 0]
scenarios_gmc_norm = {}
for name, data in scenarios_gmc.items():
    data_mean = norm_stats_detrend['tas']['lr_detrend_gma']['pixel_mean']
    data_std = norm_stats_detrend['tas']['lr_detrend_gma']['pixel_std']
    scenarios_gmc_norm[name] = (data - data_mean) / (data_std + 1e-8)
plot_histograms(ax, scenarios_gmc_norm, panel_labels[3])

# Panel (1,1): GMLD - WITH LEGEND
ax = axes[1, 1]
scenarios_gmld_norm = {}
for name, data in scenarios_gmld.items():
    data_mean = norm_stats_detrend['tas']['lr_detrend_gmt']['pixel_mean']
    data_std = norm_stats_detrend['tas']['lr_detrend_gmt']['pixel_std']
    scenarios_gmld_norm[name] = (data - data_mean) / (data_std + 1e-8)
plot_histograms(ax, scenarios_gmld_norm, panel_labels[4], show_legend=True)

# Panel (1,2): PLD
ax = axes[1, 2]
scenarios_pld_norm = {}
for name, data in scenarios_pld.items():
    data_mean = norm_stats_detrend['tas']['lr_detrend_grid']['pixel_mean']
    data_std = norm_stats_detrend['tas']['lr_detrend_grid']['pixel_std']
    scenarios_pld_norm[name] = (data - data_mean) / (data_std + 1e-8)
plot_histograms(ax, scenarios_pld_norm, panel_labels[5])

plt.tight_layout()
output_file = figures_dir / "distribution_shift.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight')
plt.show()
```
