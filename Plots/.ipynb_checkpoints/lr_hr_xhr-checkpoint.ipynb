```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
import pandas as pd

# ----------------------------
# Configuration
# ----------------------------
figures_dir = Path("../Figures/XHR")
figures_dir.mkdir(parents=True, exist_ok=True)

hr_file = Path("../evaluation_results/tas_evaluation_g6sulfur.nc")
ds_hr = xr.open_dataset(hr_file)

xhr_file = Path("../evaluation_results_xhr/g6sulfur_downscaled_xhr.nc")
ds_xhr = xr.open_dataset(xhr_file)

lr_file = Path("../data/cmip6/g6sulfur_lr_tas_monthly_2015_2100.nc")
ds_lr = xr.open_dataset(lr_file)

hist_file = Path("../data/MPI-ESM1-2-HR-LR_historical_r1i1p1f1_1850_2014_allvars.nc")
ds_hist = xr.open_dataset(hist_file)

# ----------------------------
# Functions (all unweighted for cross-resolution consistency)
# ----------------------------
def calc_baseline(ds, var='tas_hr', start_year=1850, end_year=1900):
    data = ds[var]
    years = ds.time.dt.year
    mask = (years >= start_year) & (years <= end_year)
    baseline_data = data.isel(time=mask)
    return float(baseline_data.mean(dim=['time', 'lat', 'lon']).values)

def calc_global_mean(data):
    return np.mean(data, axis=(1, 2))

def to_annual_mean(data, time_coords):
    years = time_coords.dt.year.values
    df = pd.DataFrame({'year': years, 'value': data})
    annual = df.groupby('year')['value'].mean()
    return annual.index.values, annual.values

def smooth_data(data, window=3):
    return pd.Series(data).rolling(window=window, center=True, min_periods=1).mean().values

# ----------------------------
# Calculate baseline
# ----------------------------
baseline = calc_baseline(ds_hist, 'tas_hr', 1850, 1900)

# ----------------------------
# Process data
# ----------------------------
time_slice = slice('2020', '2099')

# HR G6sulfur
hr_data = ds_hr['groundtruth'].sel(time=time_slice).values
hr_time = ds_hr['groundtruth'].sel(time=time_slice).time
hr_mean = calc_global_mean(hr_data) - baseline
years_hr, hr_annual = to_annual_mean(hr_mean, hr_time)
hr_smooth = smooth_data(hr_annual, window=3)

# XHR G6sulfur (convert K to C)
xhr_data = ds_xhr['tas_downscaled'].sel(time=time_slice).values - 273.15
xhr_time = ds_xhr['tas_downscaled'].sel(time=time_slice).time
xhr_mean = calc_global_mean(xhr_data) - baseline
years_xhr, xhr_annual = to_annual_mean(xhr_mean, xhr_time)
xhr_smooth = smooth_data(xhr_annual, window=3)

# LR G6sulfur (convert K to C)
lr_data = ds_lr['tas'].sel(time=time_slice).values - 273.15
lr_time = ds_lr['tas'].sel(time=time_slice).time
lr_mean = calc_global_mean(lr_data) - baseline
years_lr, lr_annual = to_annual_mean(lr_mean, lr_time)
lr_smooth = smooth_data(lr_annual, window=3)

# ----------------------------
# Plot
# ----------------------------
fig, ax = plt.subplots(figsize=(14, 8))

ax.plot(years_lr, lr_smooth,
        color='blue', linestyle='-', linewidth=4,
        label='LR (CMIP6 Simulation)', alpha=1, zorder=1)
ax.plot(years_hr, hr_smooth,
        color='black', linestyle='-', linewidth=4,
        label='HR (CMIP6 Simulation)', zorder=2)
ax.plot(years_xhr, xhr_smooth,
        color='red', linestyle='-', linewidth=4,
        label='XHR (ML Downscaled)', alpha=1, zorder=3)

ax.set_ylim(0, 6)
ax.set_yticks(np.arange(0, 7, 1))
ax.set_xlim(2020, 2100)
ax.set_xlabel('Year', fontsize=24, fontweight='bold', labelpad=18)
ax.set_ylabel('Temperature Anomaly (C)', fontsize=24, fontweight='bold', labelpad=18)
ax.set_title('G6sulfur Global Mean Temperature: LR vs HR vs XHR', fontsize=28, fontweight='bold', pad=25)
ax.legend(loc='upper left', fontsize=24, frameon=False)
ax.grid(True, alpha=0.5)
ax.spines['top'].set_visible(False)
ax.tick_params(labelsize=24, length=10, width=2)

# Dual y-axis on the right (same values)
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim())
ax2.set_yticks(ax.get_yticks())
ax2.tick_params(labelsize=24, length=10, width=2)
ax2.spines['top'].set_visible(False)

plt.tight_layout()
plt.savefig(figures_dir / "g6sulfur_lr_hr_xhr_comparison.png", dpi=300, bbox_inches='tight')
plt.show()

ds_hr.close()
ds_xhr.close()
ds_lr.close()
ds_hist.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
import pandas as pd

# ----------------------------
# Configuration
# ----------------------------
figures_dir = Path("../Figures/XHR")
figures_dir.mkdir(parents=True, exist_ok=True)

hr_file = Path("../evaluation_results/tas_evaluation_g6sulfur.nc")
ds_hr = xr.open_dataset(hr_file)

xhr_file = Path("../evaluation_results_xhr/g6sulfur_downscaled_xhr.nc")
ds_xhr = xr.open_dataset(xhr_file)

lr_file = Path("../data/cmip6/g6sulfur_lr_tas_monthly_2015_2100.nc")
ds_lr = xr.open_dataset(lr_file)

hist_file = Path("../data/MPI-ESM1-2-HR-LR_historical_r1i1p1f1_1850_2014_allvars.nc")
ds_hist = xr.open_dataset(hist_file)

# ----------------------------
# Functions (all unweighted for cross-resolution consistency)
# ----------------------------
def calc_baseline(ds, var='tas_hr', start_year=1850, end_year=1900):
    data = ds[var]
    years = ds.time.dt.year
    mask = (years >= start_year) & (years <= end_year)
    baseline_data = data.isel(time=mask)
    return float(baseline_data.mean(dim=['time', 'lat', 'lon']).values)

def calc_global_mean(data):
    return np.mean(data, axis=(1, 2))

def to_annual_mean(data, time_coords):
    years = time_coords.dt.year.values
    df = pd.DataFrame({'year': years, 'value': data})
    annual = df.groupby('year')['value'].mean()
    return annual.index.values, annual.values

def smooth_data(data, window=3):
    return pd.Series(data).rolling(window=window, center=True, min_periods=1).mean().values

# ----------------------------
# Calculate baseline
# ----------------------------
baseline = calc_baseline(ds_hist, 'tas_hr', 1850, 1900)

# ----------------------------
# Process data
# ----------------------------
time_slice = slice('2020', '2099')

# HR G6sulfur
hr_data = ds_hr['groundtruth'].sel(time=time_slice).values
hr_time = ds_hr['groundtruth'].sel(time=time_slice).time
hr_mean = calc_global_mean(hr_data) - baseline
years_hr, hr_annual = to_annual_mean(hr_mean, hr_time)
hr_smooth = smooth_data(hr_annual, window=3)

# XHR G6sulfur (convert K to C)
xhr_data = ds_xhr['tas_downscaled'].sel(time=time_slice).values - 273.15
xhr_time = ds_xhr['tas_downscaled'].sel(time=time_slice).time
xhr_mean = calc_global_mean(xhr_data) - baseline
years_xhr, xhr_annual = to_annual_mean(xhr_mean, xhr_time)
xhr_smooth = smooth_data(xhr_annual, window=3)

# LR G6sulfur (convert K to C)
lr_data = ds_lr['tas'].sel(time=time_slice).values - 273.15
lr_time = ds_lr['tas'].sel(time=time_slice).time
lr_mean = calc_global_mean(lr_data) - baseline
years_lr, lr_annual = to_annual_mean(lr_mean, lr_time)
lr_smooth = smooth_data(lr_annual, window=3)

# XHR regridded to HR (convert K to C)
xhr_sel = ds_xhr['tas_downscaled'].sel(time=time_slice)
xhr_regridded = xhr_sel.interp(lat=ds_hr.lat, lon=ds_hr.lon, method='linear')
xhr_rg_data = xhr_regridded.values - 273.15
xhr_rg_time = xhr_regridded.time
xhr_rg_mean = calc_global_mean(xhr_rg_data) - baseline
years_xhr_rg, xhr_rg_annual = to_annual_mean(xhr_rg_mean, xhr_rg_time)
xhr_rg_smooth = smooth_data(xhr_rg_annual, window=3)

# ----------------------------
# Plot
# ----------------------------
fig, ax = plt.subplots(figsize=(14, 8))

ax.plot(years_lr, lr_smooth,
        color='blue', linestyle='-', linewidth=5,
        label='LR (CMIP6 Simulation)', alpha=1, zorder=1)
ax.plot(years_hr, hr_smooth,
        color='black', linestyle='-', linewidth=5,
        label='HR (CMIP6 Simulation)', zorder=2)
ax.plot(years_xhr, xhr_smooth,
        color='red', linestyle='-', linewidth=5,
        label='XHR (ML Downscaled)', alpha=1, zorder=3)
ax.plot(years_xhr_rg, xhr_rg_smooth,
        color='green', linestyle='--', linewidth=5,
        label='XHRHR (ML Downscaled, regridded)', alpha=1, zorder=4)

ax.set_ylim(0, 6)
ax.set_yticks(np.arange(0, 7, 1))
ax.set_xlim(2020, 2100)
ax.set_xlabel('Year', fontsize=24, fontweight='bold', labelpad=18)
ax.set_ylabel('Temperature Anomaly (C)', fontsize=24, fontweight='bold', labelpad=18)
ax.set_title('G6sulfur Global Mean Temperature: LR vs HR vs XHR', fontsize=28, fontweight='bold', pad=25)
ax.legend(loc='upper left', fontsize=20, frameon=False)
ax.grid(True, alpha=0.5)
ax.spines['top'].set_visible(False)
ax.tick_params(labelsize=24, length=10, width=2)

# Dual y-axis on the right (same values)
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim())
ax2.set_yticks(ax.get_yticks())
ax2.tick_params(labelsize=24, length=10, width=2)
ax2.spines['top'].set_visible(False)

plt.tight_layout()
plt.savefig(figures_dir / "g6sulfur_lr_hr_xhr_comparison.png", dpi=300, bbox_inches='tight')
plt.show()

ds_hr.close()
ds_xhr.close()
ds_lr.close()
ds_hist.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib.colors import BoundaryNorm
from scipy import stats

# ----------------------------
# Configuration
# ----------------------------
figures_dir = Path("../Figures/XHR")
figures_dir.mkdir(parents=True, exist_ok=True)

hr_file = Path("../evaluation_results/tas_evaluation_g6sulfur.nc")
ds_hr = xr.open_dataset(hr_file)

xhr_file = Path("../evaluation_results_xhr/g6sulfur_downscaled_xhr.nc")
ds_xhr = xr.open_dataset(xhr_file)

time_slice = slice('2020', '2099')

# ----------------------------
# Load and prepare data
# ----------------------------
# HR groundtruth
hr = ds_hr['groundtruth'].sel(time=time_slice)
lat_hr = hr.lat.values
lon_hr = hr.lon.values

# XHR (convert K to C)
xhr = ds_xhr['tas_downscaled'].sel(time=time_slice) - 273.15
lat_xhr = xhr.lat.values
lon_xhr = xhr.lon.values

# XHR regridded to HR grid
xhr_regridded = xhr.interp(lat=ds_hr.lat, lon=ds_hr.lon, method='linear')

# ----------------------------
# Time-mean maps
# ----------------------------
hr_mean = hr.values.mean(axis=0)
xhr_mean = xhr.values.mean(axis=0)
xhr_rg_mean = xhr_regridded.values.mean(axis=0)

# Bias: XHR_regridded - HR
bias = xhr_rg_mean - hr_mean

# Significance: t-test on per-timestep bias
bias_ts = xhr_regridded.values - hr.values  # (time, lat, lon)
t_stat, p_values = stats.ttest_1samp(bias_ts, 0, axis=0)
significant = p_values < 0.05

# ----------------------------
# Shift longitudes from [0,360] to [-180,180]
# ----------------------------
def shift_lon(data, lon):
    lon_shifted = np.where(lon > 180, lon - 360, lon)
    split_idx = np.searchsorted(lon, 180.0)
    data_shifted = np.roll(data, -split_idx, axis=-1)
    lon_shifted = np.roll(lon_shifted, -split_idx)
    return data_shifted, lon_shifted

hr_mean_s, lon_hr_s = shift_lon(hr_mean, lon_hr)
xhr_mean_s, lon_xhr_s = shift_lon(xhr_mean, lon_xhr)
xhr_rg_mean_s, _ = shift_lon(xhr_rg_mean, lon_hr)
bias_s, _ = shift_lon(bias, lon_hr)
sig_s, _ = shift_lon(significant.astype(float), lon_hr)
sig_s = sig_s.astype(bool)

# ----------------------------
# Plot 2x2
# ----------------------------
fig = plt.figure(figsize=(16, 13))

panels = [
    ('a)', 'HR', hr_mean_s, lat_hr, lon_hr_s, 'RdYlBu_r', None),
    ('b)', 'XHR', xhr_mean_s, lat_xhr, lon_xhr_s, 'RdYlBu_r', None),
    ('c)', 'XHR Regridded', xhr_rg_mean_s, lat_hr, lon_hr_s, 'RdYlBu_r', None),
    ('d)', 'Bias', bias_s, lat_hr, lon_hr_s, 'RdBu_r', 'bias'),
]

for idx, (label, title, data, lat, lon, cmap, ptype) in enumerate(panels):
    ax = fig.add_subplot(2, 2, idx + 1, projection=ccrs.Robinson())
    lon_2d, lat_2d = np.meshgrid(lon, lat)

    if ptype == 'bias':
        vmin, vmax = -3, 3
    else:
        vmin, vmax = -30, 30

    im = ax.pcolormesh(lon_2d, lat_2d, data,
                       transform=ccrs.PlateCarree(),
                       cmap=cmap, vmin=vmin, vmax=vmax, shading='auto')

    if ptype == 'bias':
        lat_sub = lat_hr[::4]
        lon_sub = lon_hr_s[::4]
        sig_sub = sig_s[::4, ::4]
        lon_sub_2d, lat_sub_2d = np.meshgrid(lon_sub, lat_sub)
        ax.scatter(lon_sub_2d[sig_sub], lat_sub_2d[sig_sub],
                   s=0.5, c='black', alpha=0.3,
                   transform=ccrs.PlateCarree(), rasterized=True)

    ax.coastlines(linewidth=0.5, color='black')
    ax.add_feature(cfeature.BORDERS, linewidth=0.3, edgecolor='gray')
    ax.set_global()
    ax.set_title(title, fontsize=32, fontweight='bold', pad=14)
    ax.text(0.02, 0.98, label, transform=ax.transAxes,
            fontsize=28, fontweight='bold', verticalalignment='top')

    cbar = fig.colorbar(im, ax=ax, orientation='horizontal',
                        fraction=0.06, pad=0.04, extend='both')
        
    cbar.ax.tick_params(labelsize=18)
    
    if ptype == 'bias':
        cbar.set_label('Bias (C)', fontsize=24, fontweight='bold')
        cbar.set_ticks(np.arange(-3, 4, 1))
    else:
        cbar.set_label('Temperature (C)', fontsize=24, fontweight='bold')
        cbar.set_ticks(np.arange(-30, 31, 10))

plt.tight_layout(h_pad=2, w_pad=1.5)
plt.savefig(figures_dir / "g6sulfur_xhr_hr_bias_map.png", dpi=300, bbox_inches='tight')
plt.show()

ds_hr.close()
ds_xhr.close()
```
