```python
import torch
import os
import glob

# Directory containing checkpoints
ckpts_dir = "../ckpts"  # Update this path

# Find all tas checkpoint files
tas_files = glob.glob(os.path.join(ckpts_dir, "tas_*.pth"))
tas_files.sort()

print("Epoch numbers for tas models:")
print("-" * 60)

for checkpoint_path in tas_files:
    filename = os.path.basename(checkpoint_path)
    
    try:
        checkpoint = torch.load(checkpoint_path, map_location='cpu')
        epoch = checkpoint.get('epoch', 'Not found')
        
        print(f"{filename:40s} | Epoch: {epoch}")
        
    except Exception as e:
        print(f"{filename:40s} | Error: {str(e)}")

print("-" * 60)
```
