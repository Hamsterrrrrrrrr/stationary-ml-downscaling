```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import pickle
from scipy.stats import wasserstein_distance

# ----------------------------
# Load data
# ----------------------------
data_dir = "../data"

# Original datasets
ds_hist = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_historical_r1i1p1f1_1850_2014_allvars.nc")
ds_g6 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_g6sulfur_r1i1p1f1_2020_2099_allvars.nc")
ds_ssp585 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp585_r1i1p1f1_2015_2100_allvars.nc")
ds_ssp245 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp245_r1i1p1f1_2015_2100_allvars.nc")
ds_ssp126 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp126_r1i1p1f1_2015_2100_allvars.nc")

# Detrended datasets
ds_hist_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_historical_residual_detrended.nc")
ds_g6_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_g6sulfur_residual_detrended.nc")
ds_ssp585_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp585_residual_detrended.nc")
ds_ssp245_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp245_residual_detrended.nc")
ds_ssp126_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp126_residual_detrended.nc")

# Load normalization statistics
with open(f"{data_dir}/norm_stats.pkl", 'rb') as f:
    norm_stats = pickle.load(f)
    
with open(f"{data_dir}/norm_stats_zscore_pixel_residual_detrended.pkl", 'rb') as f:
    norm_stats_detrend = pickle.load(f)

# Training period
train_start = '1850'
train_end = '1980'

# ----------------------------
# Helper function
# ----------------------------
def compute_wasserstein(hist_data, future_data, n_samples=100000):
    """Compute Wasserstein distance between two distributions."""
    hist_vals = hist_data.values.ravel()
    future_vals = future_data.values.ravel()
    
    # Remove NaN
    hist_vals = hist_vals[np.isfinite(hist_vals)]
    future_vals = future_vals[np.isfinite(future_vals)]
    
    # Subsample for computational efficiency
    if len(hist_vals) > n_samples:
        hist_vals = np.random.choice(hist_vals, n_samples, replace=False)
    if len(future_vals) > n_samples:
        future_vals = np.random.choice(future_vals, n_samples, replace=False)
    
    return wasserstein_distance(hist_vals, future_vals)

# ----------------------------
# Organize datasets
# ----------------------------
datasets_raw = {
    'ssp126': ds_ssp126,
    'ssp245': ds_ssp245,
    'ssp585': ds_ssp585,
    'g6sulfur': ds_g6,
}

datasets_detrend = {
    'ssp126': ds_ssp126_detrend,
    'ssp245': ds_ssp245_detrend,
    'ssp585': ds_ssp585_detrend,
    'g6sulfur': ds_g6_detrend,
}

future_scenarios = ['ssp126', 'ssp245', 'ssp585', 'g6sulfur']

# ----------------------------
# Normalization statistics
# ----------------------------
# ZS-G: global z-score
zsg_mean_lr = norm_stats['tas']['lr_interp']['global_mean']
zsg_std_lr = norm_stats['tas']['lr_interp']['global_std']
zsg_mean_hr = norm_stats['tas']['hr']['global_mean']
zsg_std_hr = norm_stats['tas']['hr']['global_std']

# ZS-P: pixel-wise z-score
zsp_mean_lr = norm_stats['tas']['lr_interp']['pixel_mean']
zsp_std_lr = norm_stats['tas']['lr_interp']['pixel_std']
zsp_mean_hr = norm_stats['tas']['hr']['pixel_mean']
zsp_std_hr = norm_stats['tas']['hr']['pixel_std']

# PLD-Res: detrended + pixel-wise z-score
pld_mean_lr = norm_stats_detrend['tas']['lr_detrend_grid']['pixel_mean']
pld_std_lr = norm_stats_detrend['tas']['lr_detrend_grid']['pixel_std']
pld_mean_res = norm_stats_detrend['tas']['residual']['pixel_mean']
pld_std_res = norm_stats_detrend['tas']['residual']['pixel_std']

# ----------------------------
# Historical reference (training period)
# ----------------------------
hist_lr_raw = ds_hist['tas_lr_interp'].sel(time=slice(train_start, train_end))
hist_hr_raw = ds_hist['tas_hr'].sel(time=slice(train_start, train_end))
hist_lr_detrend = ds_hist_detrend['tas_lr_detrend_grid'].sel(time=slice(train_start, train_end))
hist_residual = ds_hist_detrend['tas_residual'].sel(time=slice(train_start, train_end))

# Normalize historical inputs
hist_input_zsg = (hist_lr_raw - zsg_mean_lr) / (zsg_std_lr + 1e-8)
hist_input_zsp = (hist_lr_raw - zsp_mean_lr) / (zsp_std_lr + 1e-8)
hist_input_pld = (hist_lr_detrend - pld_mean_lr) / (pld_std_lr + 1e-8)

# Normalize historical targets
hist_target_zsg = (hist_hr_raw - zsg_mean_hr) / (zsg_std_hr + 1e-8)
hist_target_zsp = (hist_hr_raw - zsp_mean_hr) / (zsp_std_hr + 1e-8)
hist_target_pld = (hist_residual - pld_mean_res) / (pld_std_res + 1e-8)

# ----------------------------
# Compute W distances for each future scenario
# ----------------------------
W_input = {scen: {} for scen in future_scenarios}
W_target = {scen: {} for scen in future_scenarios}

for scen in future_scenarios:
    print(f"Processing {scen}...")
    ds_raw = datasets_raw[scen]
    ds_det = datasets_detrend[scen]
    
    # Future data
    future_lr_raw = ds_raw['tas_lr_interp']
    future_hr_raw = ds_raw['tas_hr']
    future_lr_detrend = ds_det['tas_lr_detrend_grid']
    future_residual = ds_det['tas_residual']
    
    # Normalize future inputs
    future_input_zsg = (future_lr_raw - zsg_mean_lr) / (zsg_std_lr + 1e-8)
    future_input_zsp = (future_lr_raw - zsp_mean_lr) / (zsp_std_lr + 1e-8)
    future_input_pld = (future_lr_detrend - pld_mean_lr) / (pld_std_lr + 1e-8)
    
    # Normalize future targets
    future_target_zsg = (future_hr_raw - zsg_mean_hr) / (zsg_std_hr + 1e-8)
    future_target_zsp = (future_hr_raw - zsp_mean_hr) / (zsp_std_hr + 1e-8)
    future_target_pld = (future_residual - pld_mean_res) / (pld_std_res + 1e-8)
    
    # Compute W_input
    W_input[scen]['ZS-G'] = compute_wasserstein(hist_input_zsg, future_input_zsg)
    W_input[scen]['ZS-P'] = compute_wasserstein(hist_input_zsp, future_input_zsp)
    W_input[scen]['PLD-Res'] = compute_wasserstein(hist_input_pld, future_input_pld)
    
    # Compute W_target
    W_target[scen]['ZS-G'] = compute_wasserstein(hist_target_zsg, future_target_zsg)
    W_target[scen]['ZS-P'] = compute_wasserstein(hist_target_zsp, future_target_zsp)
    W_target[scen]['PLD-Res'] = compute_wasserstein(hist_target_pld, future_target_pld)
    
    print(f"  W_input: ZS-G={W_input[scen]['ZS-G']:.3f}, ZS-P={W_input[scen]['ZS-P']:.3f}, PLD-Res={W_input[scen]['PLD-Res']:.3f}")
    print(f"  W_target: ZS-G={W_target[scen]['ZS-G']:.3f}, ZS-P={W_target[scen]['ZS-P']:.3f}, PLD-Res={W_target[scen]['PLD-Res']:.3f}")

# ----------------------------
# RMSE values (fill in from your results)
# ----------------------------
RMSE = {
    'ssp126': {'ZS-G': None, 'ZS-P': None, 'PLD-Res': None},
    'ssp245': {'ZS-G': None, 'ZS-P': None, 'PLD-Res': None},
    'ssp585': {'ZS-G': None, 'ZS-P': None, 'PLD-Res': None},
    'g6sulfur': {'ZS-G': None, 'ZS-P': None, 'PLD-Res': None},
}

# ----------------------------
# Plot
# ----------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

models = ['ZS-G', 'ZS-P', 'PLD-Res']
model_colors = {'ZS-G': '#ff7f0e', 'ZS-P': '#1f77b4', 'PLD-Res': '#2ca02c'}
model_markers = {'ZS-G': 'o', 'ZS-P': 's', 'PLD-Res': '^'}

# Panel (a): W_input vs RMSE
ax = axes[0]
for model in models:
    x_vals = [W_input[scen][model] for scen in future_scenarios]
    y_vals = [RMSE[scen][model] for scen in future_scenarios]
    ax.scatter(x_vals, y_vals, c=model_colors[model], marker=model_markers[model], 
               s=150, label=model, edgecolors='black', linewidths=0.8, zorder=3)

ax.set_xlabel('Input Distribution Shift\n(Wasserstein Distance)', fontsize=14, fontweight='bold')
ax.set_ylabel('RMSE (C)', fontsize=14, fontweight='bold')
ax.text(0.02, 0.98, '(a)', transform=ax.transAxes, fontsize=18, fontweight='bold', va='top', ha='left')
ax.legend(frameon=False, fontsize=12, loc='lower right')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(axis='both', labelsize=12)

# Panel (b): W_target vs RMSE
ax = axes[1]
for model in models:
    x_vals = [W_target[scen][model] for scen in future_scenarios]
    y_vals = [RMSE[scen][model] for scen in future_scenarios]
    ax.scatter(x_vals, y_vals, c=model_colors[model], marker=model_markers[model], 
               s=150, label=model, edgecolors='black', linewidths=0.8, zorder=3)

ax.set_xlabel('Target Distribution Shift\n(Wasserstein Distance)', fontsize=14, fontweight='bold')
ax.set_ylabel('RMSE (C)', fontsize=14, fontweight='bold')
ax.text(0.02, 0.98, '(b)', transform=ax.transAxes, fontsize=18, fontweight='bold', va='top', ha='left')
ax.legend(frameon=False, fontsize=12, loc='lower right')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(axis='both', labelsize=12)

plt.tight_layout()
plt.show()
```

```python
import xarray as xr
import numpy as np
import pickle
from scipy.stats import wasserstein_distance

# ----------------------------
# Load data
# ----------------------------
data_dir = "../data"

# Original datasets
ds_hist = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_historical_r1i1p1f1_1850_2014_allvars.nc")
ds_g6 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_g6sulfur_r1i1p1f1_2020_2099_allvars.nc")
ds_ssp585 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp585_r1i1p1f1_2015_2100_allvars.nc")
ds_ssp245 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp245_r1i1p1f1_2015_2100_allvars.nc")
ds_ssp126 = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp126_r1i1p1f1_2015_2100_allvars.nc")

# Detrended datasets
ds_hist_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_historical_residual_detrended.nc")
ds_g6_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_g6sulfur_residual_detrended.nc")
ds_ssp585_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp585_residual_detrended.nc")
ds_ssp245_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp245_residual_detrended.nc")
ds_ssp126_detrend = xr.open_dataset(f"{data_dir}/MPI-ESM1-2-HR-LR_ssp126_residual_detrended.nc")

# Load normalization statistics
with open(f"{data_dir}/norm_stats.pkl", 'rb') as f:
    norm_stats = pickle.load(f)
    
with open(f"{data_dir}/norm_stats_zscore_pixel_residual_detrended.pkl", 'rb') as f:
    norm_stats_detrend = pickle.load(f)

# Training period
train_start = '1850'
train_end = '1980'

# ----------------------------
# Helper function
# ----------------------------
def compute_wasserstein(hist_data, future_data, n_samples=100000):
    """Compute Wasserstein distance between two distributions."""
    hist_vals = hist_data.values.ravel()
    future_vals = future_data.values.ravel()
    
    hist_vals = hist_vals[np.isfinite(hist_vals)]
    future_vals = future_vals[np.isfinite(future_vals)]
    
    if len(hist_vals) > n_samples:
        hist_vals = np.random.choice(hist_vals, n_samples, replace=False)
    if len(future_vals) > n_samples:
        future_vals = np.random.choice(future_vals, n_samples, replace=False)
    
    return wasserstein_distance(hist_vals, future_vals)

# ----------------------------
# Setup
# ----------------------------
datasets_raw = {
    'ssp126': ds_ssp126,
    'ssp245': ds_ssp245,
    'ssp585': ds_ssp585,
    'g6sulfur': ds_g6,
}

datasets_detrend = {
    'ssp126': ds_ssp126_detrend,
    'ssp245': ds_ssp245_detrend,
    'ssp585': ds_ssp585_detrend,
    'g6sulfur': ds_g6_detrend,
}

future_scenarios = ['ssp126', 'ssp245', 'ssp585', 'g6sulfur']
models = ['ZS-G', 'ZS-P', 'PLD-Res']

# Normalization statistics
zsg_mean_lr = norm_stats['tas']['lr_interp']['global_mean']
zsg_std_lr = norm_stats['tas']['lr_interp']['global_std']
zsg_mean_hr = norm_stats['tas']['hr']['global_mean']
zsg_std_hr = norm_stats['tas']['hr']['global_std']

zsp_mean_lr = norm_stats['tas']['lr_interp']['pixel_mean']
zsp_std_lr = norm_stats['tas']['lr_interp']['pixel_std']
zsp_mean_hr = norm_stats['tas']['hr']['pixel_mean']
zsp_std_hr = norm_stats['tas']['hr']['pixel_std']

pld_mean_lr = norm_stats_detrend['tas']['lr_detrend_grid']['pixel_mean']
pld_std_lr = norm_stats_detrend['tas']['lr_detrend_grid']['pixel_std']
pld_mean_res = norm_stats_detrend['tas']['residual']['pixel_mean']
pld_std_res = norm_stats_detrend['tas']['residual']['pixel_std']

# Historical reference
hist_lr_raw = ds_hist['tas_lr_interp'].sel(time=slice(train_start, train_end))
hist_hr_raw = ds_hist['tas_hr'].sel(time=slice(train_start, train_end))
hist_lr_detrend = ds_hist_detrend['tas_lr_detrend_grid'].sel(time=slice(train_start, train_end))
hist_residual = ds_hist_detrend['tas_residual'].sel(time=slice(train_start, train_end))

# Normalize historical
hist_input_zsg = (hist_lr_raw - zsg_mean_lr) / (zsg_std_lr + 1e-8)
hist_input_zsp = (hist_lr_raw - zsp_mean_lr) / (zsp_std_lr + 1e-8)
hist_input_pld = (hist_lr_detrend - pld_mean_lr) / (pld_std_lr + 1e-8)

hist_target_zsg = (hist_hr_raw - zsg_mean_hr) / (zsg_std_hr + 1e-8)
hist_target_zsp = (hist_hr_raw - zsp_mean_hr) / (zsp_std_hr + 1e-8)
hist_target_pld = (hist_residual - pld_mean_res) / (pld_std_res + 1e-8)

# ----------------------------
# Compute W distances
# ----------------------------
W_input = {scen: {} for scen in future_scenarios}
W_target = {scen: {} for scen in future_scenarios}

for scen in future_scenarios:
    ds_raw = datasets_raw[scen]
    ds_det = datasets_detrend[scen]
    
    future_lr_raw = ds_raw['tas_lr_interp']
    future_hr_raw = ds_raw['tas_hr']
    future_lr_detrend = ds_det['tas_lr_detrend_grid']
    future_residual = ds_det['tas_residual']
    
    future_input_zsg = (future_lr_raw - zsg_mean_lr) / (zsg_std_lr + 1e-8)
    future_input_zsp = (future_lr_raw - zsp_mean_lr) / (zsp_std_lr + 1e-8)
    future_input_pld = (future_lr_detrend - pld_mean_lr) / (pld_std_lr + 1e-8)
    
    future_target_zsg = (future_hr_raw - zsg_mean_hr) / (zsg_std_hr + 1e-8)
    future_target_zsp = (future_hr_raw - zsp_mean_hr) / (zsp_std_hr + 1e-8)
    future_target_pld = (future_residual - pld_mean_res) / (pld_std_res + 1e-8)
    
    W_input[scen]['ZS-G'] = compute_wasserstein(hist_input_zsg, future_input_zsg)
    W_input[scen]['ZS-P'] = compute_wasserstein(hist_input_zsp, future_input_zsp)
    W_input[scen]['PLD-Res'] = compute_wasserstein(hist_input_pld, future_input_pld)
    
    W_target[scen]['ZS-G'] = compute_wasserstein(hist_target_zsg, future_target_zsg)
    W_target[scen]['ZS-P'] = compute_wasserstein(hist_target_zsp, future_target_zsp)
    W_target[scen]['PLD-Res'] = compute_wasserstein(hist_target_pld, future_target_pld)

# ----------------------------
# RMSE values (fill in your values)
# ----------------------------
RMSE = {
    'ssp126': {'ZS-G': None, 'ZS-P': None, 'PLD-Res': None},
    'ssp245': {'ZS-G': None, 'ZS-P': None, 'PLD-Res': None},
    'ssp585': {'ZS-G': None, 'ZS-P': None, 'PLD-Res': None},
    'g6sulfur': {'ZS-G': None, 'ZS-P': None, 'PLD-Res': None},
}

# ----------------------------
# Print results
# ----------------------------
print("=" * 80)
print("INPUT DISTRIBUTION SHIFT (Wasserstein Distance)")
print("=" * 80)
print(f"{'Scenario':<12} {'ZS-G':>10} {'ZS-P':>10} {'PLD-Res':>10}")
print("-" * 44)
for scen in future_scenarios:
    print(f"{scen:<12} {W_input[scen]['ZS-G']:>10.3f} {W_input[scen]['ZS-P']:>10.3f} {W_input[scen]['PLD-Res']:>10.3f}")

print("\n")
print("=" * 80)
print("TARGET DISTRIBUTION SHIFT (Wasserstein Distance)")
print("=" * 80)
print(f"{'Scenario':<12} {'ZS-G':>10} {'ZS-P':>10} {'PLD-Res':>10}")
print("-" * 44)
for scen in future_scenarios:
    print(f"{scen:<12} {W_target[scen]['ZS-G']:>10.3f} {W_target[scen]['ZS-P']:>10.3f} {W_target[scen]['PLD-Res']:>10.3f}")

print("\n")
print("=" * 80)
print("RMSE (C) - Fill in your values")
print("=" * 80)
print(f"{'Scenario':<12} {'ZS-G':>10} {'ZS-P':>10} {'PLD-Res':>10}")
print("-" * 44)
for scen in future_scenarios:
    zsg = RMSE[scen]['ZS-G'] if RMSE[scen]['ZS-G'] is not None else "N/A"
    zsp = RMSE[scen]['ZS-P'] if RMSE[scen]['ZS-P'] is not None else "N/A"
    pld = RMSE[scen]['PLD-Res'] if RMSE[scen]['PLD-Res'] is not None else "N/A"
    print(f"{scen:<12} {zsg:>10} {zsp:>10} {pld:>10}")
```
