```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# ----------------------------
# Configuration
# ----------------------------
results_residual_dir = Path("../evaluation_results_residual")
figures_dir = Path("../Figures")

var_name = 'tas'
scenarios = ['g6sulfur', 'ssp245']
model_key = 'pred_grid'  # PLD-Res

# Seasons
seasons = {
    'DJF': [12, 1, 2],   # December, January, February
    'MAM': [3, 4, 5],    # March, April, May
    'JJA': [6, 7, 8],    # June, July, August
    'SON': [9, 10, 11]   # September, October, November
}

season_labels = ['DJF', 'MAM', 'JJA', 'SON']

# ----------------------------
# Functions
# ----------------------------
def calculate_seasonal_rmse(pred_data, truth_data, time, season_months):
    """
    Calculate seasonal RMSE (Root Mean Square Error).
    
    Args:
        pred_data: (time, lat, lon) prediction array
        truth_data: (time, lat, lon) ground truth array
        time: time coordinate
        season_months: list of months for the season (e.g., [12, 1, 2] for DJF)
    
    Returns:
        seasonal_rmse: (lat, lon) array of RMSE for the season
    """
    # Get month values
    months = time.dt.month.values
    
    # Create mask for the season
    season_mask = np.isin(months, season_months)
    
    # Calculate squared error
    squared_error = (pred_data - truth_data) ** 2
    
    # Extract seasonal data
    seasonal_squared_error = squared_error[season_mask]
    
    # Take temporal mean and then square root
    seasonal_rmse = np.sqrt(np.mean(seasonal_squared_error, axis=0))
    
    return seasonal_rmse

# ----------------------------
# Load data
# ----------------------------
print("Loading data...")
data = {}

for scenario in scenarios:
    residual_file = results_residual_dir / f"{var_name}_evaluation_{scenario}.nc"
    ds = xr.open_dataset(residual_file)
    
    data[scenario] = {
        'pred': ds[model_key].values,
        'truth': ds['groundtruth'].values,
        'time': ds.time,
        'lat': ds.lat.values,
        'lon': ds.lon.values
    }

# Get coordinates (same for both scenarios)
lat = data['g6sulfur']['lat']
lon = data['g6sulfur']['lon']

# ----------------------------
# Calculate seasonal RMSE
# ----------------------------
seasonal_rmse = {scenario: {} for scenario in scenarios}

for scenario in scenarios:
    for season_name, season_months in seasons.items():
        rmse = calculate_seasonal_rmse(
            data[scenario]['pred'],
            data[scenario]['truth'],
            data[scenario]['time'],
            season_months
        )
        seasonal_rmse[scenario][season_name] = rmse

# Calculate delta RMSE (G6sulfur - SSP245)
delta_rmse = {}
for season_name in season_labels:
    delta_rmse[season_name] = seasonal_rmse['g6sulfur'][season_name] - seasonal_rmse['ssp245'][season_name]

# ----------------------------
# Create 4x3 subplot figure
# ----------------------------
fig = plt.figure(figsize=(12, 10))

# Colormap settings
# For RMSE: use sequential colormap (all positive values)
rmse_cmap = 'YlOrRd'  # Yellow-Orange-Red for RMSE (higher = worse)
delta_cmap = 'RdBu_r'  # Red-Blue reversed for delta (can be positive or negative)

# Color limits for RMSE and delta-RMSE
rmse_vmin, rmse_vmax = 0, 5  # Ticks at 0, 1, 2, 3, 4, 5
delta_vmin, delta_vmax = -2, 2

# Create discrete colormaps
from matplotlib.colors import BoundaryNorm

# RMSE: 2 color grids between each tick (0, 1, 2, 3, 4, 5)
# Levels at: 0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5
rmse_tick_values = np.arange(0, 6, 1)  # Ticks at 0, 1, 2, 3, 4, 5
rmse_levels = np.arange(0, 5.5, 0.5)  # 2 grids between each tick
rmse_norm = BoundaryNorm(rmse_levels, ncolors=256)

# Delta-RMSE: 2 color grids between each tick (-2, -1, 0, 1, 2)
# Levels at: -2, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2
delta_tick_values = np.arange(-2, 3, 1)  # Ticks at -2, -1, 0, 1, 2
delta_levels = np.arange(-2, 2.5, 0.5)  # 2 grids between each tick
delta_norm = BoundaryNorm(delta_levels, ncolors=256)

# Panel labels
panel_labels = [chr(97 + i) + ')' for i in range(12)]  # a) through l)

# Create meshgrid for plotting
lon_2d, lat_2d = np.meshgrid(lon, lat)

# Plot 4 rows (seasons) x 3 columns (G6sulfur, SSP245, Delta)
for row_idx, season_name in enumerate(season_labels):
    
    # Column 1: G6sulfur RMSE
    ax1 = fig.add_subplot(4, 3, row_idx * 3 + 1, projection=ccrs.Robinson())
    rmse_g6 = seasonal_rmse['g6sulfur'][season_name]
    
    im1 = ax1.pcolormesh(lon_2d, lat_2d, rmse_g6,
                        transform=ccrs.PlateCarree(),
                        cmap=rmse_cmap, norm=rmse_norm,
                        shading='auto')
    
    ax1.coastlines(linewidth=0.5, color='black')
    ax1.add_feature(cfeature.BORDERS, linewidth=0.3, edgecolor='gray')
    ax1.set_global()
    
    if row_idx == 0:
        ax1.set_title('G6sulfur', fontsize=16, fontweight='bold', pad=10)
    
    # Add season label on left
    ax1.text(-0.15, 0.5, season_name, 
            transform=ax1.transAxes,
            fontsize=18, fontweight='bold',
            rotation=90, va='center', ha='center')
    
    # Add panel label WITHOUT box
    ax1.text(0.02, 0.98, panel_labels[row_idx * 3], 
            transform=ax1.transAxes,
            fontsize=16, fontweight='bold',
            verticalalignment='top',
            color='black')
    
    # Column 2: SSP245 RMSE
    ax2 = fig.add_subplot(4, 3, row_idx * 3 + 2, projection=ccrs.Robinson())
    rmse_245 = seasonal_rmse['ssp245'][season_name]
    
    im2 = ax2.pcolormesh(lon_2d, lat_2d, rmse_245,
                        transform=ccrs.PlateCarree(),
                        cmap=rmse_cmap, norm=rmse_norm,
                        shading='auto')
    
    ax2.coastlines(linewidth=0.5, color='black')
    ax2.add_feature(cfeature.BORDERS, linewidth=0.3, edgecolor='gray')
    ax2.set_global()
    
    if row_idx == 0:
        ax2.set_title('SSP2-4.5', fontsize=16, fontweight='bold', pad=10)
    
    # Add panel label WITHOUT box
    ax2.text(0.02, 0.98, panel_labels[row_idx * 3 + 1], 
            transform=ax2.transAxes,
            fontsize=16, fontweight='bold',
            verticalalignment='top',
            color='black')
    
    # Column 3: Delta RMSE (G6 - SSP245)
    ax3 = fig.add_subplot(4, 3, row_idx * 3 + 3, projection=ccrs.Robinson())
    delta_rmse_data = delta_rmse[season_name]
    
    im3 = ax3.pcolormesh(lon_2d, lat_2d, delta_rmse_data,
                        transform=ccrs.PlateCarree(),
                        cmap=delta_cmap, norm=delta_norm,
                        shading='auto')
    
    ax3.coastlines(linewidth=0.5, color='black')
    ax3.add_feature(cfeature.BORDERS, linewidth=0.3, edgecolor='gray')
    ax3.set_global()
    
    if row_idx == 0:
        ax3.set_title(r'G6 $-$ SSP245', fontsize=16, fontweight='bold', pad=10)
    
    # Add panel label WITHOUT box
    ax3.text(0.02, 0.98, panel_labels[row_idx * 3 + 2], 
            transform=ax3.transAxes,
            fontsize=16, fontweight='bold',
            verticalalignment='top',
            color='black')

# Add colorbars with tighter spacing
# Colorbar for RMSE (columns 1 and 2) with specific ticks
cbar_ax1 = fig.add_axes([0.172, 0.016, 0.35, 0.012])
cbar1 = fig.colorbar(im1, cax=cbar_ax1, orientation='horizontal', extend='max',
                     ticks=rmse_tick_values)
cbar1.set_label('RMSE (C)', fontsize=14, fontweight='bold')
cbar1.ax.tick_params(labelsize=12)

# Colorbar for delta-RMSE (column 3) with specific ticks
cbar_ax2 = fig.add_axes([0.736, 0.016, 0.2, 0.012])
cbar2 = fig.colorbar(im3, cax=cbar_ax2, orientation='horizontal', extend='both',
                     ticks=delta_tick_values)
cbar2.set_label(r'$\Delta$RMSE (C)', fontsize=14, fontweight='bold')
cbar2.ax.tick_params(labelsize=12)

# Adjust layout with tighter spacing
plt.tight_layout(rect=[0, 0.03, 1.0, 1.0], h_pad=1.0, w_pad=1.0)

# Save figure
output_file = figures_dir / "g6sulfur_ssp245_rmse_map.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight')
```
