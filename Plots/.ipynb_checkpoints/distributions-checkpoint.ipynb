```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
from scipy.ndimage import gaussian_filter1d
from joblib import Parallel, delayed

# ----------------------------
# Configuration
# ----------------------------
results_dir = Path("../evaluation_results")
results_residual_dir = Path("../evaluation_results_residual")
figures_dir = Path("../Figures")

variables = ['tas', 'pr']
scenarios = ['historical', 'ssp126', 'ssp245', 'ssp585', 'g6sulfur']

scenario_labels = {
    'historical': 'Historical',
    'ssp126': 'SSP1-2.6',
    'ssp245': 'SSP2-4.5',
    'g6sulfur': 'G6sulfur',
    'ssp585': 'SSP5-8.5'
}

var_config = {
    'tas': {
        'label': 'Temperature (C)',
        'bins': np.linspace(-30, 50, 161),
        'xlim': (-30, 50)
    },
    'pr': {
        'label': 'Precipitation (mm/day)',
        'bins': np.linspace(0, 25, 126),
        'xlim': (0, 25)
    }
}

models = {
    'groundtruth': 'Truth',
    'bilinear': 'BI',
    'qdm': 'QDM',
    'mld': 'MLd',
    'mls': 'MLs'
}

model_colors = {
    'groundtruth': 'black',
    'bilinear': 'green',
    'qdm': 'blue',
    'mld': 'orange',
    'mls': 'red'
}

model_linewidths = {
    'groundtruth': 4.0,
    'bilinear': 2.0,
    'qdm': 2.0,
    'mld': 2.0,
    'mls': 2.0
}

# ----------------------------
# Latitude weights
# ----------------------------
ds_tmp = xr.open_dataset(results_dir / "tas_evaluation_historical.nc")
lats = ds_tmp['lat'].values
ds_tmp.close()

cos_weights = np.cos(np.deg2rad(lats))  # (192,)

# ----------------------------
# Functions
# ----------------------------
def compute_weighted_histogram_density(data, bins, lats_weights, smooth=True, sigma=1.5):
    """Compute area-weighted normalized histogram density."""
    # data shape: (time, lat, lon)
    weights_3d = np.broadcast_to(lats_weights[np.newaxis, :, np.newaxis], data.shape)
    
    data_flat = data.ravel()
    weights_flat = weights_3d.ravel()
    
    mask = np.isfinite(data_flat)
    data_flat = data_flat[mask]
    weights_flat = weights_flat[mask]
    
    counts, _ = np.histogram(data_flat, bins=bins, weights=weights_flat)
    
    total = counts.sum()
    density = counts / total if total > 0 else counts
    
    if smooth:
        density = gaussian_filter1d(density, sigma=sigma)
    
    return density

def load_and_compute(var, scenario):
    """Load all model data and compute weighted histograms for one (var, scenario)."""
    bins = var_config[var]['bins']
    densities = {}
    
    # Main file (groundtruth, bilinear, MLd)
    main_file = results_dir / f"{var}_evaluation_{scenario}.nc"
    if main_file.exists():
        ds = xr.open_dataset(main_file)
        
        if 'groundtruth' in ds:
            densities['groundtruth'] = compute_weighted_histogram_density(
                ds['groundtruth'].values, bins, cos_weights)
        
        if 'input' in ds:
            densities['bilinear'] = compute_weighted_histogram_density(
                ds['input'].values, bins, cos_weights)
        
        if 'pred_zscore_pixel' in ds:
            densities['mld'] = compute_weighted_histogram_density(
                ds['pred_zscore_pixel'].values, bins, cos_weights)
        
        ds.close()
    
    # QDM
    qdm_file = results_dir / f"{var}_evaluation_{scenario}_qdm.nc"
    if qdm_file.exists():
        ds = xr.open_dataset(qdm_file)
        if 'pred_qdm' in ds:
            densities['qdm'] = compute_weighted_histogram_density(
                ds['pred_qdm'].values, bins, cos_weights)
        ds.close()
    
    # MLs
    res_file = results_residual_dir / f"{var}_evaluation_{scenario}.nc"
    if res_file.exists():
        ds = xr.open_dataset(res_file)
        if 'pred_grid' in ds:
            densities['mls'] = compute_weighted_histogram_density(
                ds['pred_grid'].values, bins, cos_weights)
        ds.close()
    
    return var, scenario, densities

# ----------------------------
# Load and compute in parallel
# ----------------------------
tasks = [(var, scenario) for var in variables for scenario in scenarios]
parallel_results = Parallel(n_jobs=min(len(tasks), 10), verbose=1)(
    delayed(load_and_compute)(var, scenario) for var, scenario in tasks
)

# Organize results
all_densities = {var: {scen: {} for scen in scenarios} for var in variables}
for var, scenario, densities in parallel_results:
    all_densities[var][scenario] = densities

# ----------------------------
# Create 4x5 subplot figure
# ----------------------------
fig, axes = plt.subplots(4, 5, figsize=(25, 20))

for var_idx, var in enumerate(variables):
    config = var_config[var]
    bins = config['bins']
    bin_centers = (bins[:-1] + bins[1:]) / 2
    
    hist_row = var_idx * 2
    error_row = var_idx * 2 + 1
    
    for col, scenario in enumerate(scenarios):
        densities = all_densities[var][scenario]
        
        # --- HISTOGRAM ROW ---
        ax_hist = axes[hist_row, col]
        
        for model_key, model_label in models.items():
            if model_key not in densities:
                continue
            
            density = densities[model_key]
            mask = density > 0
            
            ax_hist.plot(bin_centers[mask], density[mask],
                        color=model_colors[model_key],
                        linestyle='-',
                        linewidth=model_linewidths[model_key],
                        alpha=1.0,
                        label=model_label)
        
        col_letter = chr(97 + col)
        panel_label = f"{col_letter}{hist_row + 1})"
        ax_hist.text(0.02, 0.98, panel_label,
                    transform=ax_hist.transAxes,
                    fontsize=18, fontweight='bold',
                    va='top', ha='left')
        
        ax_hist.set_yscale('log')
        ax_hist.set_xlim(config['xlim'])
        ax_hist.set_ylim(1e-6, 1)
        ax_hist.set_xlabel(config['label'], fontsize=20, fontweight='bold')
        
        if col == 0:
            ax_hist.set_ylabel('Density', fontsize=20, fontweight='bold')
        
        if hist_row == 0:
            ax_hist.set_title(scenario_labels[scenario], fontsize=20, fontweight='bold', pad=16)
        
        if col == 0:
            if var_idx == 0:
                ax_hist.legend(loc='lower left', fontsize=14, frameon=False, handlelength=2)
            else:
                ax_hist.legend(loc='upper right', fontsize=14, frameon=False, handlelength=2)
        
        ax_hist.tick_params(labelsize=16, length=8, width=2)
        ax_hist.spines['top'].set_visible(False)
        ax_hist.spines['right'].set_visible(False)
        ax_hist.spines['left'].set_linewidth(2)
        ax_hist.spines['bottom'].set_linewidth(2)
        
        # --- ERROR ROW ---
        ax_error = axes[error_row, col]
        
        if 'groundtruth' in densities:
            truth_density = densities['groundtruth']
            
            for model_key, model_label in models.items():
                if model_key == 'groundtruth' or model_key not in densities:
                    continue
                
                abs_error = np.abs(densities[model_key] - truth_density)
                
                ax_error.plot(bin_centers, abs_error,
                            color=model_colors[model_key],
                            linestyle='-',
                            linewidth=model_linewidths[model_key],
                            alpha=1.0,
                            label=model_label)
        
        panel_label = f"{col_letter}{error_row + 1})"
        ax_error.text(0.02, 0.98, panel_label,
                     transform=ax_error.transAxes,
                     fontsize=18, fontweight='bold',
                     va='top', ha='left')
        
        ax_error.set_yscale('log')
        ax_error.set_xlim(config['xlim'])
        ax_error.set_ylim(1e-6, 1)
        ax_error.set_xlabel(config['label'], fontsize=20, fontweight='bold')
        
        if col == 0:
            ax_error.set_ylabel('Absolute Error', fontsize=20, fontweight='bold')
        
        ax_error.tick_params(labelsize=16, length=8, width=2)
        ax_error.spines['top'].set_visible(False)
        ax_error.spines['right'].set_visible(False)
        ax_error.spines['left'].set_linewidth(2)
        ax_error.spines['bottom'].set_linewidth(2)

plt.subplots_adjust(hspace=0.4, wspace=0.25)

output_file = figures_dir / "distributions.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
from scipy.ndimage import gaussian_filter1d

# ----------------------------
# Configuration
# ----------------------------
results_dir = Path("../evaluation_results")
results_residual_dir = Path("../evaluation_results_residual")
figures_dir = Path("../Figures")

variables = ['tas', 'pr']
scenarios = ['historical', 'ssp126', 'ssp245', 'ssp585', 'g6sulfur']  

scenario_labels = {
    'historical': 'Historical',
    'ssp126': 'SSP1-2.6',
    'ssp245': 'SSP2-4.5',
    'g6sulfur': 'G6sulfur',
    'ssp585': 'SSP5-8.5'
}

# Variable configuration
var_config = {
    'tas': {
        'label': 'Temperature (C)',
        'bins': np.linspace(-30, 50, 161),
        'xlim': (-30, 50)
    },
    'pr': {
        'label': 'Precipitation (mm/day)',
        'bins': np.linspace(0, 25, 126),
        'xlim': (0, 25)
    }
}

# Models to compare 
models = {
    'groundtruth': 'Truth',
    'bilinear': 'Bilinear',
    'qdm': 'QDM',
    'mld': 'MLd',
    'mls': 'MLs'
}

# Colors for each model
model_colors = {
    'groundtruth': 'black',
    'bilinear': 'green',    
    'qdm': 'blue',         
    'mld': 'orange',         
    'mls': 'red'         
}

# Line styles
model_linestyles = {
    'groundtruth': '-',
    'bilinear': '-',
    'qdm': '-',
    'mld': '-',
    'mls': '-'
}

# Line widths
model_linewidths = {
    'groundtruth': 4.0,  
    'bilinear': 2.0,
    'qdm': 2.0,
    'mld': 2.0,
    'mls': 2.0
}

# Alpha values
model_alphas = {
    'groundtruth': 1.0, 
    'bilinear': 1.0,
    'qdm': 1.0,
    'mld': 1.0,
    'mls': 1.0
}

# ----------------------------
# Functions
# ----------------------------
def compute_histogram_density(data, bins, smooth=True, sigma=1.5):
    """Compute normalized histogram density for given data and bins"""
    data_flat = data.flatten()
    data_flat = data_flat[~np.isnan(data_flat)]
    
    # Compute histogram with density normalization
    counts, _ = np.histogram(data_flat, bins=bins, density=False)
    
    # Normalize to get probability density
    total = counts.sum()
    if total > 0:
        density = counts / total
    else:
        density = counts
    
    # Apply Gaussian smoothing if requested
    if smooth:
        density = gaussian_filter1d(density, sigma=sigma)
    
    return density

def load_model_data(var, scenario):
    """Load all model data for a given variable and scenario"""
    data_dict = {}
    
    # Load main results (groundtruth, bilinear, MLd)
    main_file = results_dir / f"{var}_evaluation_{scenario}.nc"
    if main_file.exists():
        ds_main = xr.open_dataset(main_file)
        
        # Ground truth
        if 'groundtruth' in ds_main:
            data_dict['groundtruth'] = ds_main['groundtruth'].values
        
        # Bilinear (input)
        if 'input' in ds_main:
            data_dict['bilinear'] = ds_main['input'].values
        
        # MLd (zscore_pixel)
        if 'pred_zscore_pixel' in ds_main:
            data_dict['mld'] = ds_main['pred_zscore_pixel'].values
        
        ds_main.close()
    
    # Load QDM results
    qdm_file = results_dir / f"{var}_evaluation_{scenario}_qdm.nc"
    if qdm_file.exists():
        ds_qdm = xr.open_dataset(qdm_file)
        
        if 'pred_qdm' in ds_qdm:
            data_dict['qdm'] = ds_qdm['pred_qdm'].values
        
        ds_qdm.close()
    
    # Load residual results (MLs = PLD-Res)
    residual_file = results_residual_dir / f"{var}_evaluation_{scenario}.nc"
    if residual_file.exists():
        ds_residual = xr.open_dataset(residual_file)
        
        if 'pred_grid' in ds_residual:
            data_dict['mls'] = ds_residual['pred_grid'].values
        
        ds_residual.close()
    
    return data_dict

# ----------------------------
# Create 4x5 subplot figure
# ----------------------------
fig, axes = plt.subplots(4, 5, figsize=(25, 20))

# Panel labels: a1-e1 for tas histogram, a2-e2 for tas error, a3-e3 for pr histogram, a4-e4 for pr error
panel_labels = [
    ['a1)', 'b1)', 'c1)', 'd1)', 'e1)'],  # Row 1: tas histogram
    ['a2)', 'b2)', 'c2)', 'd2)', 'e2)'],  # Row 2: tas error
    ['a3)', 'b3)', 'c3)', 'd3)', 'e3)'],  # Row 3: pr histogram
    ['a4)', 'b4)', 'c4)', 'd4)', 'e4)']   # Row 4: pr error
]

for var_idx, var in enumerate(variables):
    config = var_config[var]
    bins = config['bins']
    bin_centers = (bins[:-1] + bins[1:]) / 2
    
    # Determine which rows to use for this variable
    hist_row = var_idx * 2  # Row 0 for tas, row 2 for pr
    error_row = var_idx * 2 + 1  # Row 1 for tas error, row 3 for pr error
    
    for col, scenario in enumerate(scenarios):
        # --- HISTOGRAM ROW ---
        ax_hist = axes[hist_row, col]
        
        # Load data for this variable and scenario
        data_dict = load_model_data(var, scenario)
        
        # Store densities for error calculation
        densities = {}
        
        # Plot each model
        for model_key, model_label in models.items():
            if model_key not in data_dict:
                continue
            
            # Compute histogram density with smoothing
            density = compute_histogram_density(data_dict[model_key], bins, smooth=True, sigma=1.5)
            densities[model_key] = density
            
            # Filter out zeros for log scale
            mask = density > 0
            
            # Plot
            ax_hist.plot(bin_centers[mask], density[mask],
                   color=model_colors[model_key],
                   linestyle=model_linestyles[model_key],
                   linewidth=model_linewidths[model_key],
                   alpha=model_alphas[model_key],
                   label=model_label)
        
        # Add panel label
        ax_hist.text(0.02, 0.98, panel_labels[hist_row][col],
               transform=ax_hist.transAxes,
               fontsize=16, fontweight='bold',
               va='top', ha='left')
        
        # Format axis
        ax_hist.set_yscale('log')
        ax_hist.set_xlim(config['xlim'])
        ax_hist.set_ylim(1e-6, 1)
        
        # X-axis label for histogram rows
        ax_hist.set_xlabel(config['label'], fontsize=14, fontweight='bold')
        
        # Y-axis label (only on first column)
        if col == 0:
            ax_hist.set_ylabel('Density', fontsize=14, fontweight='bold')
        
        # Title (only on top row) with increased padding
        if hist_row == 0:
            ax_hist.set_title(scenario_labels[scenario], fontsize=18, fontweight='bold', pad=20)
        
        # Legend 
        if col == 0:
            if var_idx == 0:
                # First variable (tas): lower left
                ax_hist.legend(loc='lower left', fontsize=10, frameon=False, handlelength=2)
            else:
                # Second variable (pr): upper right
                ax_hist.legend(loc='upper right', fontsize=10, frameon=False, handlelength=2)
        
        # Formatting
        ax_hist.tick_params(labelsize=12, length=8, width=1.5)
        ax_hist.spines['top'].set_visible(False)
        ax_hist.spines['right'].set_visible(False)
        ax_hist.spines['left'].set_linewidth(2)
        ax_hist.spines['bottom'].set_linewidth(2)
        
        # --- ERROR ROW ---
        ax_error = axes[error_row, col]
        
        # Plot absolute errors relative to Truth
        if 'groundtruth' in densities:
            truth_density = densities['groundtruth']
            
            for model_key, model_label in models.items():
                if model_key == 'groundtruth' or model_key not in densities:
                    continue
                
                # Compute absolute error
                abs_error = np.abs(densities[model_key] - truth_density)
                
                # Plot error
                ax_error.plot(bin_centers, abs_error,
                       color=model_colors[model_key],
                       linestyle=model_linestyles[model_key],
                       linewidth=model_linewidths[model_key],
                       alpha=model_alphas[model_key],
                       label=model_label)
        
        # Add panel label
        ax_error.text(0.02, 0.98, panel_labels[error_row][col],
               transform=ax_error.transAxes,
               fontsize=16, fontweight='bold',
               va='top', ha='left')
        
        # Format axis
        ax_error.set_yscale('log')
        ax_error.set_xlim(config['xlim'])
        ax_error.set_ylim(1e-6, 1)
        
        # X-axis label for error rows
        ax_error.set_xlabel(config['label'], fontsize=14, fontweight='bold')
        
        # Y-axis label (only on first column)
        if col == 0:
            ax_error.set_ylabel('Absolute Error', fontsize=14, fontweight='bold')
            
        # Formatting
        ax_error.tick_params(labelsize=12, length=8, width=1.5)
        ax_error.spines['top'].set_visible(False)
        ax_error.spines['right'].set_visible(False)
        ax_error.spines['left'].set_linewidth(2)
        ax_error.spines['bottom'].set_linewidth(2)

# Increased padding between rows
plt.subplots_adjust(hspace=0.4, wspace=0.25)

# Save figure
output_file = figures_dir / "fig8.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight')
plt.show()
```
