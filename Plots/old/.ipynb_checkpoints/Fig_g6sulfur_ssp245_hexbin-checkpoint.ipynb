```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
import warnings
import matplotlib as mpl

# Suppress warnings
warnings.filterwarnings('ignore', category=RuntimeWarning)

# ----------------------------
# Configuration
# ----------------------------
results_residual_dir = Path("../evaluation_results_residual")
figures_dir = Path("../Figures")
data_dir = Path("../data")

var_name = 'tas'
scenarios = ['g6sulfur', 'ssp245']
model_key = 'pred_grid'

# Seasons
seasons = {
    'DJF': [12, 1, 2],
    'MAM': [3, 4, 5],
    'JJA': [6, 7, 8],
    'SON': [9, 10, 11]
}

season_labels = ['DJF', 'MAM', 'JJA', 'SON']

# ----------------------------
# Helper Functions
# ----------------------------
def load_land_mask(mask_file):
    """Load land-sea mask."""
    ds_mask = xr.open_dataset(mask_file)
    land_mask = ds_mask['sftlf'].values > 50.0
    ds_mask.close()
    return land_mask

def filter_common_years(data_dict, start_year=2020, end_year=2099):
    """
    Filter data to common years across scenarios.
    
    Args:
        data_dict: dictionary with scenario data
        start_year: start year for common period
        end_year: end year for common period
    
    Returns:
        filtered_data_dict: dictionary with filtered data
    """
    filtered_data = {}
    
    for scenario in data_dict.keys():
        time = data_dict[scenario]['time']
        years = time.dt.year.values
        
        # Create mask for common period
        year_mask = (years >= start_year) & (years <= end_year)
        
        filtered_data[scenario] = {
            'pred': data_dict[scenario]['pred'][year_mask],
            'truth': data_dict[scenario]['truth'][year_mask],
            'time': time[year_mask],
            'lat': data_dict[scenario]['lat'],
            'lon': data_dict[scenario]['lon']
        }
        
        print(f"  {scenario}: Filtered from {len(years)} to {np.sum(year_mask)} timesteps "
              f"({start_year}-{end_year})")
    
    return filtered_data

def calculate_seasonal_mean_temperature(truth_data, time, season_months):
    """
    Calculate seasonal mean temperature (ground truth).
    
    Args:
        truth_data: (time, lat, lon) array
        time: time coordinate
        season_months: list of months for the season
    
    Returns:
        seasonal_mean: (lat, lon) array
    """
    months = time.dt.month.values
    season_mask = np.isin(months, season_months)
    
    seasonal_data = truth_data[season_mask]
    seasonal_mean = np.mean(seasonal_data, axis=0)
    
    return seasonal_mean

def calculate_seasonal_bias(pred_data, truth_data, time, season_months):
    """
    Calculate seasonal mean bias.
    
    Args:
        pred_data: (time, lat, lon) array
        truth_data: (time, lat, lon) array
        time: time coordinate
        season_months: list of months for the season
    
    Returns:
        seasonal_bias: (lat, lon) array
    """
    months = time.dt.month.values
    season_mask = np.isin(months, season_months)
    
    bias = pred_data - truth_data
    seasonal_bias_timeseries = bias[season_mask]
    seasonal_mean_bias = np.mean(seasonal_bias_timeseries, axis=0)
    
    return seasonal_mean_bias

def calculate_seasonal_rmse(pred_data, truth_data, time, season_months):
    """
    Calculate seasonal RMSE.
    
    Args:
        pred_data: (time, lat, lon) array
        truth_data: (time, lat, lon) array
        time: time coordinate
        season_months: list of months for the season
    
    Returns:
        seasonal_rmse: (lat, lon) array
    """
    months = time.dt.month.values
    season_mask = np.isin(months, season_months)
    
    squared_error = (pred_data - truth_data) ** 2
    seasonal_squared_error = squared_error[season_mask]
    seasonal_rmse = np.sqrt(np.mean(seasonal_squared_error, axis=0))
    
    return seasonal_rmse

# ----------------------------
# Load data
# ----------------------------
print("Loading data...")
data = {}

for scenario in scenarios:
    residual_file = results_residual_dir / f"{var_name}_evaluation_{scenario}.nc"
    ds = xr.open_dataset(residual_file)
    
    data[scenario] = {
        'pred': ds[model_key].values,
        'truth': ds['groundtruth'].values,
        'time': ds.time,
        'lat': ds.lat.values,
        'lon': ds.lon.values
    }
    
    years = ds.time.dt.year.values
    print(f"  Loaded {scenario}: {years[0]}-{years[-1]} ({len(years)} timesteps)")

# Filter to common years (2020-2099)
print("\nFiltering to common period (2020-2099)...")
data = filter_common_years(data, start_year=2020, end_year=2099)

# Get coordinates
lat = data['g6sulfur']['lat']
lon = data['g6sulfur']['lon']

# Load land mask
print("\nLoading land mask...")
land_mask_file = data_dir / "sftlf_fx_MPI-ESM1-2-HR_ssp585_r1i1p1f1_gn.nc"
land_mask = load_land_mask(land_mask_file)

# ----------------------------
# Calculate scatter plot data for each season
# ----------------------------
print("\nCalculating scatter plot data...")
scatter_data = {}

for season_name, season_months in seasons.items():
    print(f"  Processing {season_name}...")
    
    # Calculate mean temperature (ground truth) for each scenario
    mean_temp_ssp245 = calculate_seasonal_mean_temperature(
        data['ssp245']['truth'],
        data['ssp245']['time'],
        season_months
    )
    
    mean_temp_g6 = calculate_seasonal_mean_temperature(
        data['g6sulfur']['truth'],
        data['g6sulfur']['time'],
        season_months
    )
    
    # X-axis: Mean-state shift (SSP245 - G6)
    mean_state_shift = mean_temp_ssp245 - mean_temp_g6
    
    # Calculate bias for each scenario
    bias_ssp245 = calculate_seasonal_bias(
        data['ssp245']['pred'],
        data['ssp245']['truth'],
        data['ssp245']['time'],
        season_months
    )
    
    bias_g6 = calculate_seasonal_bias(
        data['g6sulfur']['pred'],
        data['g6sulfur']['truth'],
        data['g6sulfur']['time'],
        season_months
    )
    
    # Y-axis for bias: -bias (G6 - SSP245)
    delta_bias = bias_g6 - bias_ssp245
    
    # Calculate RMSE for each scenario
    rmse_ssp245 = calculate_seasonal_rmse(
        data['ssp245']['pred'],
        data['ssp245']['truth'],
        data['ssp245']['time'],
        season_months
    )
    
    rmse_g6 = calculate_seasonal_rmse(
        data['g6sulfur']['pred'],
        data['g6sulfur']['truth'],
        data['g6sulfur']['time'],
        season_months
    )
    
    # Y-axis for RMSE: -RMSE (G6 - SSP245)
    delta_rmse = rmse_g6 - rmse_ssp245
    
    # Flatten and separate by land/ocean
    scatter_data[season_name] = {
        'x_land': mean_state_shift[land_mask],
        'x_ocean': mean_state_shift[~land_mask],
        'y_bias_land': delta_bias[land_mask],
        'y_bias_ocean': delta_bias[~land_mask],
        'y_rmse_land': delta_rmse[land_mask],
        'y_rmse_ocean': delta_rmse[~land_mask]
    }

# ----------------------------
# Create 4x4 subplot figure with hexbin
# ----------------------------
plt.style.use('default')
fig, axes = plt.subplots(4, 4, figsize=(20, 16))

column_titles = [
    'Bias - Land',
    'Bias - Ocean',
    'RMSE - Land',
    'RMSE - Ocean'
]

for row_idx, season_name in enumerate(season_labels):
    
    # Column 0: Bias - Land
    ax = axes[row_idx, 0]
    ax.set_facecolor('black')
    
    hb = ax.hexbin(scatter_data[season_name]['x_land'],
                   scatter_data[season_name]['y_bias_land'],
                   gridsize=50, cmap='inferno', bins='log', mincnt=1,
                   vmin=1, vmax=500)  # 10^0 to 10^2 for land
    
    ax.axhline(y=0, color='white', linestyle='--', linewidth=1.5, alpha=0.8)
    ax.axvline(x=0, color='white', linestyle='--', linewidth=1.5, alpha=0.8)
    
    ax.set_xlim(-5, 5)
    ax.set_ylim(-3, 3)
    ax.tick_params(axis='both', which='major', labelsize=12, 
                  length=8, width=1.5)
    
    if row_idx == 3:
        ax.set_xlabel('Mean-state shift (C)', fontsize=12, fontweight='bold')
    if row_idx == 0:
        ax.set_title(column_titles[0], fontsize=14, fontweight='bold', pad=10)
    
    ax.set_ylabel('-Bias (C)', fontsize=12, fontweight='bold')
    
    # Season label on left
    ax.text(-0.25, 0.5, season_name, 
           transform=ax.transAxes,
           fontsize=16, fontweight='bold',
           rotation=90, va='center', ha='center')
    
    # Panel label
    panel_label = chr(97 + row_idx * 4) + ')'
    ax.text(0.02, 0.98, panel_label, 
           transform=ax.transAxes,
           fontsize=14, fontweight='bold',
           verticalalignment='top', color='white')
    
    # Add colorbar
    cb = plt.colorbar(hb, ax=ax, pad=0.02)
    cb.set_label('log10(N)', fontsize=11, fontweight='bold')
    cb.ax.tick_params(labelsize=10)
    
    # Column 1: Bias - Ocean
    ax = axes[row_idx, 1]
    ax.set_facecolor('black')
    
    hb = ax.hexbin(scatter_data[season_name]['x_ocean'],
                   scatter_data[season_name]['y_bias_ocean'],
                   gridsize=50, cmap='inferno', bins='log', mincnt=1,
                   vmin=1, vmax=1000)  # 10^0 to 10^3 for ocean
    
    ax.axhline(y=0, color='white', linestyle='--', linewidth=1.5, alpha=0.8)
    ax.axvline(x=0, color='white', linestyle='--', linewidth=1.5, alpha=0.8)
    
    ax.set_xlim(-5, 5)
    ax.set_ylim(-3, 3)
    ax.tick_params(axis='both', which='major', labelsize=12, 
                  length=8, width=1.5)
    
    if row_idx == 3:
        ax.set_xlabel('Mean-state shift (C)', fontsize=12, fontweight='bold')
    if row_idx == 0:
        ax.set_title(column_titles[1], fontsize=14, fontweight='bold', pad=10)
    
    ax.set_ylabel('-Bias (C)', fontsize=12, fontweight='bold')
    
    # Panel label
    panel_label = chr(97 + row_idx * 4 + 1) + ')'
    ax.text(0.02, 0.98, panel_label, 
           transform=ax.transAxes,
           fontsize=14, fontweight='bold',
           verticalalignment='top', color='white')
    
    # Add colorbar
    cb = plt.colorbar(hb, ax=ax, pad=0.02)
    cb.set_label('log10(N)', fontsize=11, fontweight='bold')
    cb.ax.tick_params(labelsize=10)
    
    # Column 2: RMSE - Land
    ax = axes[row_idx, 2]
    ax.set_facecolor('black')
    
    hb = ax.hexbin(scatter_data[season_name]['x_land'],
                   scatter_data[season_name]['y_rmse_land'],
                   gridsize=50, cmap='inferno', bins='log', mincnt=1,
                   vmin=1, vmax=500)  # 10^0 to 10^2 for land
    
    ax.axhline(y=0, color='white', linestyle='--', linewidth=1.5, alpha=0.8)
    ax.axvline(x=0, color='white', linestyle='--', linewidth=1.5, alpha=0.8)
    
    ax.set_xlim(-5, 5)
    ax.set_ylim(-2, 2)
    ax.tick_params(axis='both', which='major', labelsize=12, 
                  length=8, width=1.5)
    
    if row_idx == 3:
        ax.set_xlabel('Mean-state shift (C)', fontsize=12, fontweight='bold')
    if row_idx == 0:
        ax.set_title(column_titles[2], fontsize=14, fontweight='bold', pad=10)
    
    ax.set_ylabel('-RMSE (C)', fontsize=12, fontweight='bold')
    
    # Panel label
    panel_label = chr(97 + row_idx * 4 + 2) + ')'
    ax.text(0.02, 0.98, panel_label, 
           transform=ax.transAxes,
           fontsize=14, fontweight='bold',
           verticalalignment='top', color='white')
    
    # Add colorbar
    cb = plt.colorbar(hb, ax=ax, pad=0.02)
    cb.set_label('log10(N)', fontsize=11, fontweight='bold')
    cb.ax.tick_params(labelsize=10)
    
    # Column 3: RMSE - Ocean
    ax = axes[row_idx, 3]
    ax.set_facecolor('black')
    
    hb = ax.hexbin(scatter_data[season_name]['x_ocean'],
                   scatter_data[season_name]['y_rmse_ocean'],
                   gridsize=50, cmap='inferno', bins='log', mincnt=1,
                   vmin=1, vmax=1000)  # 10^0 to 10^3 for ocean
    
    ax.axhline(y=0, color='white', linestyle='--', linewidth=1.5, alpha=0.8)
    ax.axvline(x=0, color='white', linestyle='--', linewidth=1.5, alpha=0.8)
    
    ax.set_xlim(-5, 5)
    ax.set_ylim(-2, 2)
    ax.tick_params(axis='both', which='major', labelsize=12, 
                  length=8, width=1.5)
    
    if row_idx == 3:
        ax.set_xlabel('Mean-state shift (C)', fontsize=12, fontweight='bold')
    if row_idx == 0:
        ax.set_title(column_titles[3], fontsize=14, fontweight='bold', pad=10)
    
    ax.set_ylabel('-RMSE (C)', fontsize=12, fontweight='bold')
    
    # Panel label
    panel_label = chr(97 + row_idx * 4 + 3) + ')'
    ax.text(0.02, 0.98, panel_label, 
           transform=ax.transAxes,
           fontsize=14, fontweight='bold',
           verticalalignment='top', color='white')
    
    # Add colorbar
    cb = plt.colorbar(hb, ax=ax, pad=0.02)
    cb.set_label('log10(N)', fontsize=11, fontweight='bold')
    cb.ax.tick_params(labelsize=10)

plt.tight_layout()
output_file = figures_dir / "g6sulfur_ssp245_hexbin.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight')
plt.show()
```
