# G6sulfur vs SSP245

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
from scipy import stats
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# ----------------------------
# Configuration
# ----------------------------
results_dir = Path("../evaluation_results")
results_residual_dir = Path("../evaluation_results_residual")
figures_dir = Path("../Figures")

var_name = 'tas'
scenarios = ['g6sulfur', 'ssp245']

# ----------------------------
# Functions
# ----------------------------
def calculate_gridpoint_trends_with_pvalue(data, time):
    """
    Calculate linear trend and p-value at each grid point.
    
    Args:
        data: (time, lat, lon) array
        time: time coordinate (datetime64)
    
    Returns:
        trends: (lat, lon) array of slopes in C/year
        pvalues: (lat, lon) array of p-values
    """
    n_time, n_lat, n_lon = data.shape
    
    # Convert time to decimal years
    years = time.dt.year.values + time.dt.month.values / 12.0
    
    # Initialize arrays
    trends = np.zeros((n_lat, n_lon))
    pvalues = np.ones((n_lat, n_lon))  # Default to 1 (not significant)
    
    # Calculate trend at each grid point
    for i in range(n_lat):
        for j in range(n_lon):
            time_series = data[:, i, j]
            
            # Skip if all NaN
            if np.all(np.isnan(time_series)):
                trends[i, j] = np.nan
                pvalues[i, j] = np.nan
                continue
            
            # Remove NaN values
            mask = ~np.isnan(time_series)
            if np.sum(mask) < 10:  # Need at least 10 points
                trends[i, j] = np.nan
                pvalues[i, j] = np.nan
                continue
            
            x = years[mask]
            y = time_series[mask]
            
            # Linear regression
            slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
            
            # Keep as C/year (don't multiply by 10)
            trends[i, j] = slope
            pvalues[i, j] = p_value
    
    return trends, pvalues

# ----------------------------
# Load data and calculate trends
# ----------------------------
trend_data = {}
pvalue_data = {}

for scenario in scenarios:
    print(f"Loading {scenario} data...")
    
    # Load residual results
    residual_file = results_residual_dir / f"{var_name}_evaluation_{scenario}.nc"
    ds_residual = xr.open_dataset(residual_file)
    
    # Get coordinates
    lat = ds_residual.lat.values
    lon = ds_residual.lon.values
    time = ds_residual.time
    
    # Ground truth
    gt_data = ds_residual['groundtruth'].values
    trend_data[f'{scenario}_gt'], pvalue_data[f'{scenario}_gt'] = calculate_gridpoint_trends_with_pvalue(gt_data, time)
    print(f"  Calculated {scenario} ground truth")
    
    # PLD-Res prediction
    pred_data = ds_residual['pred_grid'].values
    trend_data[f'{scenario}_pld'], pvalue_data[f'{scenario}_pld'] = calculate_gridpoint_trends_with_pvalue(pred_data, time)
    print(f"  Calculated {scenario} PLD-Res")

# ----------------------------
# Create 2x2 subplot figure
# ----------------------------
fig = plt.figure(figsize=(12, 8))

# Common colorbar limits
vmin, vmax = -0.06, 0.06
cmap = 'RdYlBu_r'

# Create discrete colormap with 2 color grids between each tick
tick_values = np.arange(-0.06, 0.08, 0.02)
levels = np.arange(-0.06, 0.065, 0.005)  
from matplotlib.colors import BoundaryNorm
norm = BoundaryNorm(levels, ncolors=256)

# Panel labels
panel_labels = ['a)', 'b)', 'c)', 'd)']

# Subplot configuration: (scenario, data_type, title)
subplots = [
    ('g6sulfur', 'gt', 'G6sulfur Truth'),
    ('ssp245', 'gt', 'SSP2-4.5 Truth'),
    ('g6sulfur', 'pld', 'G6sulfur Prediction'),
    ('ssp245', 'pld', 'SSP2-4.5 Prediction')
]

# Create meshgrid for plotting
lon_2d, lat_2d = np.meshgrid(lon, lat)

for idx, (scenario, data_type, title) in enumerate(subplots):
    # Use Robinson projection - 2 rows x 2 columns
    ax = fig.add_subplot(2, 2, idx + 1, projection=ccrs.Robinson())
    
    key = f'{scenario}_{data_type}'
    trends = trend_data[key]
    pvalues = pvalue_data[key]
    
    # Plot the trend map
    im = ax.pcolormesh(lon_2d, lat_2d, trends,
                       transform=ccrs.PlateCarree(),
                       cmap=cmap, norm=norm,
                       shading='auto')
    
    # Add stippling for statistical significance (p < 0.05)
    significance_level = 0.05
    significant = pvalues < significance_level
    
    # Add dots where significant
    step = 2
    ax.plot(lon_2d[::step, ::step][significant[::step, ::step]], 
            lat_2d[::step, ::step][significant[::step, ::step]], 
            'k.', markersize=0.5, transform=ccrs.PlateCarree(), alpha=0.5)
    
    # Add coastlines and features
    ax.coastlines(linewidth=0.5, color='black')
    ax.add_feature(cfeature.BORDERS, linewidth=0.3, edgecolor='gray')
    
    # Add title
    ax.set_title(title, fontsize=16, fontweight='bold', pad=10)
    
    # Add panel letter WITHOUT box
    ax.text(0.02, 0.98, panel_labels[idx], 
            transform=ax.transAxes,
            fontsize=16, fontweight='bold',
            verticalalignment='top',
            color='black')
    
    # Set extent to global
    ax.set_global()

# Add colorbar at the bottom center with specific ticks
cbar_ax = fig.add_axes([0.25, 0.04, 0.5, 0.02])
cbar = fig.colorbar(im, cax=cbar_ax, orientation='horizontal', 
                    extend='both', ticks=tick_values)
cbar.set_label(r'$\Delta$ temperature trend (C/yr)', fontsize=14, fontweight='bold')
cbar.ax.tick_params(labelsize=12)

# Adjust layout
plt.tight_layout(rect=[0, 0.06, 1.0, 1.0], h_pad=1.0, w_pad=1.0)

# Save figure
output_file = figures_dir / "g6sulfur_ssp245_trend_map.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight')
```

# SSP585

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path
from scipy import stats
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# ----------------------------
# Configuration
# ----------------------------
results_dir = Path("../evaluation_results")
results_residual_dir = Path("../evaluation_results_residual")
figures_dir = Path("../Figures")

var_name = 'tas'
scenario = 'ssp585'

# Models to plot - now 3 rows x 3 columns
models = [
    # Row 1: Ground truth + normalization methods
    'groundtruth', 'none', 'minmax_pixel',
    # Row 2: More normalization methods
    'zscore_global', 'zscore_pixel', 'instance_zscore',
    # Row 3: Residual methods
    'raw_residual', 'gmld_residual', 'pld_residual'
]

model_labels = {
    'groundtruth': 'Ground Truth',
    'none': 'None',
    'minmax_pixel': 'MM-P',
    'zscore_global': 'ZS-G',
    'zscore_pixel': 'ZS-P',
    'instance_zscore': 'Inst-ZS',
    'raw_residual': 'Raw-Res',
    'gmld_residual': 'GMLD-Res',
    'pld_residual': 'PLD-Res'
}

# ----------------------------
# Functions
# ----------------------------
def calculate_gridpoint_trends_with_pvalue(data, time):
    """
    Calculate linear trend and p-value at each grid point.
    
    Args:
        data: (time, lat, lon) array
        time: time coordinate (datetime64)
    
    Returns:
        trends: (lat, lon) array of slopes in C/year
        pvalues: (lat, lon) array of p-values
    """
    n_time, n_lat, n_lon = data.shape
    
    # Convert time to decimal years
    years = time.dt.year.values + time.dt.month.values / 12.0
    
    # Initialize arrays
    trends = np.zeros((n_lat, n_lon))
    pvalues = np.ones((n_lat, n_lon))  # Default to 1 (not significant)
    
    # Calculate trend at each grid point
    for i in range(n_lat):
        for j in range(n_lon):
            time_series = data[:, i, j]
            
            # Skip if all NaN
            if np.all(np.isnan(time_series)):
                trends[i, j] = np.nan
                pvalues[i, j] = np.nan
                continue
            
            # Remove NaN values
            mask = ~np.isnan(time_series)
            if np.sum(mask) < 10:  # Need at least 10 points
                trends[i, j] = np.nan
                pvalues[i, j] = np.nan
                continue
            
            x = years[mask]
            y = time_series[mask]
            
            # Linear regression
            slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
            
            # Keep as C/year (don't multiply by 10)
            trends[i, j] = slope
            pvalues[i, j] = p_value
    
    return trends, pvalues

# ----------------------------
# Load data
# ----------------------------
print(f"Loading {scenario} data...")
results_file = results_dir / f"{var_name}_evaluation_{scenario}.nc"
ds = xr.open_dataset(results_file)

# Load residual results
residual_file = results_residual_dir / f"{var_name}_evaluation_{scenario}.nc"
ds_residual = xr.open_dataset(residual_file)

# Get coordinates
lat = ds.lat.values
lon = ds.lon.values
time = ds.time

# ----------------------------
# Calculate trends with p-values
# ----------------------------
trend_data = {}
pvalue_data = {}

print("Calculating trends with p-values...")

# Ground truth
gt_data = ds['groundtruth'].values
trend_data['groundtruth'], pvalue_data['groundtruth'] = calculate_gridpoint_trends_with_pvalue(gt_data, time)

# Normalization methods (from original results)
for model in ['none', 'minmax_pixel', 'zscore_global', 'zscore_pixel', 'instance_zscore']:
    pred_key = f'pred_{model}'
    if pred_key in ds:
        pred_data = ds[pred_key].values
        trend_data[model], pvalue_data[model] = calculate_gridpoint_trends_with_pvalue(pred_data, time)
        print(f"  Calculated {model}")

# Residual methods (from residual results)
residual_mapping = {
    'raw_residual': 'pred_raw',
    'gmld_residual': 'pred_gmt',
    'pld_residual': 'pred_grid'
}

for model_key, data_key in residual_mapping.items():
    if data_key in ds_residual:
        pred_data = ds_residual[data_key].values
        trend_data[model_key], pvalue_data[model_key] = calculate_gridpoint_trends_with_pvalue(pred_data, time)
        print(f"  Calculated {model_key}")

# ----------------------------
# Create figure with Robinson projection - 3x3 grid
# ----------------------------
fig = plt.figure(figsize=(16, 11))

# Common colorbar limits
vmin, vmax = -0.15, 0.15
cmap = 'RdYlBu_r'

# Create discrete colormap with more levels
levels = np.linspace(vmin, vmax, 21)  # Increased from 13 to 21 levels (20 color bands)
from matplotlib.colors import BoundaryNorm
norm = BoundaryNorm(levels, ncolors=256)

# Panel labels
panel_labels = [chr(97 + i) + ')' for i in range(9)]  # a) through i)

for idx, model_key in enumerate(models):
    # Use Robinson projection - 3 rows x 3 columns
    ax = fig.add_subplot(3, 3, idx + 1, projection=ccrs.Robinson())
    
    if model_key not in trend_data:
        continue
    
    trends = trend_data[model_key]
    pvalues = pvalue_data[model_key]
    
    # Create meshgrid for plotting
    lon_2d, lat_2d = np.meshgrid(lon, lat)
    
    # Plot the trend map with discrete levels
    im = ax.pcolormesh(lon_2d, lat_2d, trends,
                       transform=ccrs.PlateCarree(),
                       cmap=cmap, norm=norm,
                       shading='auto')
    
    # Add stippling for statistical significance (p < 0.05)
    significance_level = 0.05
    significant = pvalues < significance_level
    
    # Add dots where significant
    step = 2  # Adjust this to control dot density
    ax.plot(lon_2d[::step, ::step][significant[::step, ::step]], 
            lat_2d[::step, ::step][significant[::step, ::step]], 
            'k.', markersize=0.5, transform=ccrs.PlateCarree(), alpha=0.5)
    
    # Add coastlines and features
    ax.coastlines(linewidth=0.5, color='black')
    ax.add_feature(cfeature.BORDERS, linewidth=0.3, edgecolor='gray')
    
    # Add model label as title (without panel letter)
    ax.set_title(f'{model_labels[model_key]}',
                fontsize=16, fontweight='bold', pad=10)
    
    # Add panel letter in upper left corner WITHOUT box
    ax.text(0.02, 0.98, panel_labels[idx], 
            transform=ax.transAxes,
            fontsize=16, fontweight='bold',
            verticalalignment='top',
            color='black')
    
    # Set extent to global
    ax.set_global()

# Add colorbar at the bottom center (below panels) with tighter spacing
# [left, bottom, width, height]
cbar_ax = fig.add_axes([0.25, 0.02, 0.5, 0.015])
cbar = fig.colorbar(im, cax=cbar_ax, orientation='horizontal', 
                    extend='both')
cbar.set_label(r'$\Delta$ ssp585 temperature trend (C/yr)', fontsize=14, fontweight='bold')
cbar.ax.tick_params(labelsize=12)

# Adjust layout with tighter spacing
plt.tight_layout(rect=[0, 0.035, 1.0, 1.0], h_pad=1.0, w_pad=1.0)

# Save figure
output_file = figures_dir / "ssp585_trend_map.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight')
```
