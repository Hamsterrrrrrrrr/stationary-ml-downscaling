```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path

# ----------------------------
# Configuration
# ----------------------------
results_dir = Path("../evaluation_results")
results_residual_dir = Path("../evaluation_results_residual")
figures_dir = Path("../Figures")

variables = ['tas', 'pr']
scenarios = ['historical', 'ssp126', 'ssp245', 'ssp585', 'g6sulfur']

scenario_labels = {
    'historical': 'Historical',
    'ssp126': 'SSP1-2.6',
    'ssp245': 'SSP2-4.5',
    'g6sulfur': 'G6sulfur',
    'ssp585': 'SSP5-8.5'
}

# Variable configuration
var_config = {
    'tas': {
        'mean_label': 'Mean Temperature (C)',
        'error_label': 'Absolute Error (C)',
    },
    'pr': {
        'mean_label': 'Mean Precipitation (mm/day)',
        'error_label': 'Absolute Error (mm/day)',
    }
}

# Models to compare
models = {
    'groundtruth': 'Truth',
    'bilinear': 'Bilinear',
    'qdm': 'QDM',
    'mld': 'MLd',
    'mls': 'MLs'
}

# Colors for each model
model_colors = {
    'groundtruth': 'black',
    'bilinear': 'green',    
    'qdm': 'blue',         
    'mld': 'orange',         
    'mls': 'red'         
}

# Line styles
model_linestyles = {
    'groundtruth': '-',
    'bilinear': '-',
    'qdm': '-',
    'mld': '-',
    'mls': '-'
}

# Line widths
model_linewidths = {
    'groundtruth': 4.0,
    'bilinear': 2.0,
    'qdm': 2.0,
    'mld': 2.0,
    'mls': 2.0
}

# Alpha values
model_alphas = {
    'groundtruth': 1.0,
    'bilinear': 1.0,
    'qdm': 1.0,
    'mld': 1.0,
    'mls': 1.0
}

# ----------------------------
# Functions
# ----------------------------
def compute_lat_profile(data):
    """Compute latitude profile by averaging over time and longitude"""
    # Average over time (axis 0) and longitude (axis 2)
    lat_profile = np.nanmean(data, axis=(0, 2))
    return lat_profile

def load_model_data(var, scenario):
    """Load all model data for a given variable and scenario"""
    data_dict = {}
    lat_coords = None
    
    # Load main results (groundtruth, bilinear, MLd)
    main_file = results_dir / f"{var}_evaluation_{scenario}.nc"
    if main_file.exists():
        ds_main = xr.open_dataset(main_file)
        
        # Get latitude coordinates
        if lat_coords is None and 'lat' in ds_main.coords:
            lat_coords = ds_main['lat'].values
        
        # Ground truth
        if 'groundtruth' in ds_main:
            data_dict['groundtruth'] = ds_main['groundtruth'].values
        
        # Bilinear (input)
        if 'input' in ds_main:
            data_dict['bilinear'] = ds_main['input'].values
        
        # MLd (zscore_pixel)
        if 'pred_zscore_pixel' in ds_main:
            data_dict['mld'] = ds_main['pred_zscore_pixel'].values
        
        ds_main.close()
    
    # Load QDM results
    qdm_file = results_dir / f"{var}_evaluation_{scenario}_qdm.nc"
    if qdm_file.exists():
        ds_qdm = xr.open_dataset(qdm_file)
        
        if 'pred_qdm' in ds_qdm:
            data_dict['qdm'] = ds_qdm['pred_qdm'].values
        
        ds_qdm.close()
    
    # Load residual results (MLs = PLD-Res)
    residual_file = results_residual_dir / f"{var}_evaluation_{scenario}.nc"
    if residual_file.exists():
        ds_residual = xr.open_dataset(residual_file)
        
        if 'pred_grid' in ds_residual:
            data_dict['mls'] = ds_residual['pred_grid'].values
        
        ds_residual.close()
    
    return data_dict, lat_coords

# ----------------------------
# Create 4x5 subplot figure
# ----------------------------
fig, axes = plt.subplots(4, 5, figsize=(25, 20))

# Panel labels
panel_labels = [
    ['a1)', 'b1)', 'c1)', 'd1)', 'e1)'],  # Row 1: tas mean
    ['a2)', 'b2)', 'c2)', 'd2)', 'e2)'],  # Row 2: tas error
    ['a3)', 'b3)', 'c3)', 'd3)', 'e3)'],  # Row 3: pr mean
    ['a4)', 'b4)', 'c4)', 'd4)', 'e4)']   # Row 4: pr error
]

for var_idx, var in enumerate(variables):
    config = var_config[var]
    
    # Determine which rows to use for this variable
    mean_row = var_idx * 2      # Row 0 for tas, row 2 for pr
    error_row = var_idx * 2 + 1  # Row 1 for tas error, row 3 for pr error
    
    for col, scenario in enumerate(scenarios):
        # --- MEAN PROFILE ROW ---
        ax_mean = axes[mean_row, col]
        
        # Load data for this variable and scenario
        data_dict, lat_coords = load_model_data(var, scenario)
        
        # Store profiles for error calculation
        profiles = {}
        
        # Plot each model
        for model_key, model_label in models.items():
            if model_key not in data_dict:
                continue
            
            # Compute latitude profile
            lat_profile = compute_lat_profile(data_dict[model_key])
            profiles[model_key] = lat_profile
            
            # Plot
            ax_mean.plot(lat_coords, lat_profile,
                        color=model_colors[model_key],
                        linestyle=model_linestyles[model_key],
                        linewidth=model_linewidths[model_key],
                        alpha=model_alphas[model_key],
                        label=model_label)
        
        # Add panel label
        ax_mean.text(0.02, 0.98, panel_labels[mean_row][col],
                    transform=ax_mean.transAxes,
                    fontsize=16, fontweight='bold',
                    va='top', ha='left')
        
        # Format axis
        ax_mean.set_xlim(-90, 90)
        ax_mean.set_xlabel('Latitude (N)', fontsize=14, fontweight='bold')
        
        # Y-axis label (only on first column)
        if col == 0:
            ax_mean.set_ylabel(config['mean_label'], fontsize=14, fontweight='bold')
        
        # Title (only on top row)
        if mean_row == 0:
            ax_mean.set_title(scenario_labels[scenario], fontsize=18, fontweight='bold', pad=20)
        
        # Legend (only on first column)
        if col == 0:
            if var_idx == 0:
                # Temperature: upper right
                ax_mean.legend(loc='upper right', fontsize=10, frameon=False, handlelength=2)
            else:
                # Precipitation: upper right
                ax_mean.legend(loc='upper right', fontsize=10, frameon=False, handlelength=2)
        
        # Formatting
        ax_mean.tick_params(labelsize=12, length=8, width=1.5)
        ax_mean.spines['top'].set_visible(False)
        ax_mean.spines['right'].set_visible(False)
        ax_mean.spines['left'].set_linewidth(2)
        ax_mean.spines['bottom'].set_linewidth(2)
        ax_mean.grid(True, alpha=0.3, linestyle='--', linewidth=0.5)
        
        # --- ERROR PROFILE ROW ---
        ax_error = axes[error_row, col]
        
        # Plot absolute errors relative to Truth
        if 'groundtruth' in profiles:
            truth_profile = profiles['groundtruth']
            
            for model_key, model_label in models.items():
                if model_key == 'groundtruth' or model_key not in profiles:
                    continue
                
                # Compute absolute error
                abs_error = np.abs(profiles[model_key] - truth_profile)
                
                # Plot error
                ax_error.plot(lat_coords, abs_error,
                            color=model_colors[model_key],
                            linestyle=model_linestyles[model_key],
                            linewidth=model_linewidths[model_key],
                            alpha=model_alphas[model_key],
                            label=model_label)
        
        # Add panel label
        ax_error.text(0.02, 0.98, panel_labels[error_row][col],
                     transform=ax_error.transAxes,
                     fontsize=16, fontweight='bold',
                     va='top', ha='left')
        
        # Format axis
        ax_error.set_xlim(-90, 90)
        ax_error.set_xlabel('Latitude (N)', fontsize=14, fontweight='bold')
        
        # Y-axis label (only on first column)
        if col == 0:
            ax_error.set_ylabel(config['error_label'], fontsize=14, fontweight='bold')
        
        # Formatting
        ax_error.tick_params(labelsize=12, length=8, width=1.5)
        ax_error.spines['top'].set_visible(False)
        ax_error.spines['right'].set_visible(False)
        ax_error.spines['left'].set_linewidth(2)
        ax_error.spines['bottom'].set_linewidth(2)
        ax_error.grid(True, alpha=0.3, linestyle='--', linewidth=0.5)

# Adjust spacing
plt.subplots_adjust(hspace=0.4, wspace=0.25)

# Save figure
output_file = figures_dir / "fig8.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight')
plt.show()
```
