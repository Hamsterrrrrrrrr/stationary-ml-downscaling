```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path

# ----------------------------
# Configuration
# ----------------------------
results_residual_dir = Path("../evaluation_results_residual")
figures_dir = Path("../Figures")
data_dir = Path("../data")

var_name = 'tas'
scenarios = ['g6sulfur', 'ssp245']
model_key = 'pred_grid'

# ----------------------------
# Define Regions
# ----------------------------
regions = {
    'Global Land': {
        'lat_range': None,
        'lon_range': None,
        'mask_type': 'land'
    },
    'Global Ocean': {
        'lat_range': None,
        'lon_range': None,
        'mask_type': 'ocean'
    },
    'Arctic': {
        'lat_range': (70, 90),
        'lon_range': None,
        'mask_type': None
    },
    'South Asian Monsoon': {
        'lat_range': (5, 30),
        'lon_range': (70, 100),
        'mask_type': 'land'
    },
    'North Atlantic Storm-Track': {
        'lat_range': (30, 65),
        'lon_range': (-80, 0),
        'mask_type': 'ocean'
    },
    'Amazon': {
        'lat_range': (-20, 5),
        'lon_range': (280, 315),
        'mask_type': 'land'
    }
}

# ----------------------------
# Helper Functions
# ----------------------------
def load_land_mask(mask_file):
    """Load land-sea mask."""
    ds_mask = xr.open_dataset(mask_file)
    land_mask = ds_mask['sftlf'].values > 50.0
    ds_mask.close()
    return land_mask

def create_region_mask(lat, lon, lat_range, lon_range, land_mask=None, mask_type=None):
    """Create a boolean mask for a region."""
    lat_2d, lon_2d = np.meshgrid(lat, lon, indexing='ij')
    mask = np.ones_like(lat_2d, dtype=bool)
    
    if lat_range is not None:
        mask &= (lat_2d >= lat_range[0]) & (lat_2d <= lat_range[1])
    
    if lon_range is not None:
        mask &= (lon_2d >= lon_range[0]) & (lon_2d <= lon_range[1])
    
    if mask_type == 'land':
        mask &= land_mask
    elif mask_type == 'ocean':
        mask &= ~land_mask
    
    return mask

def compute_regional_monthly_bias_groupby(pred_data, truth_data, time, lat, lon, region_mask):
    """
    Compute regional monthly mean bias using groupby (no area weighting).
    
    Args:
        pred_data: (time, lat, lon) array
        truth_data: (time, lat, lon) array
        time: time coordinate
        lat: latitude array
        lon: longitude array
        region_mask: (lat, lon) boolean mask
    
    Returns:
        monthly_bias: array of length 12
    """
    # Create xarray DataArray
    bias_da = xr.DataArray(
        pred_data - truth_data,
        coords={'time': time, 'lat': lat, 'lon': lon},
        dims=['time', 'lat', 'lon']
    )
    
    # Mask to region (set non-region values to NaN)
    bias_da = bias_da.where(region_mask)
    
    # Spatial mean (ignoring NaN), then group by month and average
    bias_spatial = bias_da.mean(dim=['lat', 'lon'])
    monthly_bias = bias_spatial.groupby('time.month').mean().values
    
    return monthly_bias

def bootstrap_delta_bias_ci(pred_data_g6, truth_data_g6, time_g6,
                             pred_data_245, truth_data_245, time_245,
                             region_mask, 
                             n_bootstrap=1000, confidence=90):
    """
    Compute bootstrap confidence intervals for Δ-bias using simple loop (no weights).
    """
    bias_g6 = pred_data_g6 - truth_data_g6
    bias_245 = pred_data_245 - truth_data_245
    
    years_g6 = time_g6.dt.year.values
    years_245 = time_245.dt.year.values
    months_g6 = time_g6.dt.month.values
    months_245 = time_245.dt.month.values
    
    overlapping_years = np.intersect1d(np.unique(years_g6), np.unique(years_245))
    n_years = len(overlapping_years)
    
    if n_years == 0:
        print("  ERROR: No overlapping years!")
        return np.full(12, np.nan), np.full(12, np.nan)
    
    bootstrap_samples = np.zeros((n_bootstrap, 12))
    
    np.random.seed(42)  # For reproducibility
    
    for i in range(n_bootstrap):
        # Resample years with replacement
        resampled_years = np.random.choice(overlapping_years, size=n_years, replace=True)
        
        for month in range(1, 13):
            delta_values = []
            
            for year in resampled_years:
                mask_g6 = (years_g6 == year) & (months_g6 == month)
                mask_245 = (years_245 == year) & (months_245 == month)
                
                if np.sum(mask_g6) > 0 and np.sum(mask_245) > 0:
                    # Simple spatial mean (no weights)
                    bias_g6_masked = bias_g6[mask_g6][:, region_mask]
                    bias_245_masked = bias_245[mask_245][:, region_mask]
                    
                    mean_g6 = np.mean(bias_g6_masked, axis=1)
                    mean_245 = np.mean(bias_245_masked, axis=1)
                    
                    delta_values.extend(mean_g6 - mean_245)
            
            bootstrap_samples[i, month - 1] = np.mean(delta_values) if delta_values else np.nan
    
    # Compute confidence intervals
    alpha = (100 - confidence) / 2
    lower_ci = np.nanpercentile(bootstrap_samples, alpha, axis=0)
    upper_ci = np.nanpercentile(bootstrap_samples, 100 - alpha, axis=0)
    
    return lower_ci, upper_ci

# ----------------------------
# Load data
# ----------------------------
print("Loading data...")
data = {}

for scenario in scenarios:
    residual_file = results_residual_dir / f"{var_name}_evaluation_{scenario}.nc"
    ds = xr.open_dataset(residual_file)
    
    data[scenario] = {
        'pred': ds[model_key].values,
        'truth': ds['groundtruth'].values,
        'time': ds.time,
        'lat': ds.lat.values,
        'lon': ds.lon.values
    }
    print(f"  Loaded {scenario}")

# Get coordinates
lat = data['g6sulfur']['lat']
lon = data['g6sulfur']['lon']

# Load land mask
print("\nLoading land mask...")
land_mask_file = data_dir / "sftlf_fx_MPI-ESM1-2-HR_ssp585_r1i1p1f1_gn.nc"
land_mask = load_land_mask(land_mask_file)

# ----------------------------
# Compute monthly bias for each region and scenario
# ----------------------------
print("\nComputing regional monthly bias...")
results = {}

for region_name, region_def in regions.items():
    print(f"\nProcessing region: {region_name}")
    
    # Create region mask
    region_mask = create_region_mask(
        lat, lon,
        region_def['lat_range'],
        region_def['lon_range'],
        land_mask,
        region_def['mask_type']
    )
    
    print(f"  Region has {np.sum(region_mask)} grid points")
    
    results[region_name] = {}
    
    for scenario in scenarios:
        # Compute monthly bias using groupby (no weights)
        monthly_bias = compute_regional_monthly_bias_groupby(
            data[scenario]['pred'],
            data[scenario]['truth'],
            data[scenario]['time'],
            lat,
            lon,
            region_mask
        )
        
        results[region_name][scenario] = {
            'bias': monthly_bias
        }
        
        print(f"  {scenario}: Annual mean bias = {np.mean(monthly_bias):.3f} K")
    
    # Compute delta-bias
    delta_bias = (results[region_name]['g6sulfur']['bias'] - 
                  results[region_name]['ssp245']['bias'])
    results[region_name]['delta_bias'] = delta_bias
    
    # Compute bootstrap CI for delta-bias
    print("  Computing bootstrap CI for Δ-bias...")
    lower_ci, upper_ci = bootstrap_delta_bias_ci(
        data['g6sulfur']['pred'],
        data['g6sulfur']['truth'],
        data['g6sulfur']['time'],
        data['ssp245']['pred'],
        data['ssp245']['truth'],
        data['ssp245']['time'],
        region_mask,
        n_bootstrap=300
    )
    
    results[region_name]['delta_lower_ci'] = lower_ci
    results[region_name]['delta_upper_ci'] = upper_ci
    
    print(f"  Δ-bias (annual mean) = {np.mean(delta_bias):.3f} K")

# ----------------------------
# Plot regional monthly bias curves
# ----------------------------
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

months = np.arange(1, 13)
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

colors = {
    'ssp245': 'blue',
    'g6sulfur': 'red',
    'delta': 'green'
}

# Y-axis limits for each panel
ylims = [
    [-0.8, -0.4, 0.0, 0.4, 0.8],
    [-1.2, -0.6, 0.0, 0.6, 1.2],
    [-8, -4, 0, 4, 8],
    [-0.8, -0.4, 0.0, 0.4, 0.8],
    [-1.4, -0.7, 0.0, 0.7, 1.4],
    [-0.6, -0.3, 0.0, 0.3, 0.6]
]

for idx, (region_name, region_results) in enumerate(results.items()):
    ax = axes[idx]
    
    # Plot SSP245 bias
    bias_245 = region_results['ssp245']['bias']
    ax.plot(months, bias_245, 's-', color=colors['ssp245'], 
            linewidth=2, label='SSP2-4.5', markersize=10, alpha=1)
    
    # Plot G6sulfur bias
    bias_g6 = region_results['g6sulfur']['bias']
    ax.plot(months, bias_g6, 's-', color=colors['g6sulfur'], 
            linewidth=2, label='G6sulfur', markersize=10, alpha=1)
    
    # Plot delta-bias with CI
    delta_bias = region_results['delta_bias']
    lower_ci = region_results['delta_lower_ci']
    upper_ci = region_results['delta_upper_ci']
    
    ax.plot(months, delta_bias, 's-', color=colors['delta'], 
            linewidth=2.5, label=r'$\Delta$-bias (G6 $-$ SSP245)', markersize=10, alpha=1)
    ax.fill_between(months, lower_ci, upper_ci, 
                     color=colors['delta'], alpha=0.25, 
                     edgecolor='none')
    
    # Add zero line
    ax.axhline(y=0, color='black', linestyle='--', linewidth=2, alpha=1)
    
    # Highlight JJA and DJF
    ax.axvspan(5.5, 8.5, alpha=0.05, color='red')
    ax.axvspan(11.5, 12.5, alpha=0.05, color='blue')
    ax.axvspan(0.5, 2.5, alpha=0.05, color='blue')
    
    # Formatting
    ax.set_ylabel('Bias (°C)', fontsize=12, fontweight='bold')
    ax.set_title(region_name, fontsize=14, fontweight='bold')
    ax.set_xticks(months)
    ax.set_xticklabels(month_labels, rotation=45, ha='right', fontweight='bold')
    ax.set_xlim(0.5, 12.5)
    
    # Set y-axis limits and ticks
    ax.set_yticks(ylims[idx])
    ax.set_ylim(min(ylims[idx]), max(ylims[idx]))
    
    # Increase tick label font size
    ax.tick_params(labelsize=12, length=10, width=2)
    
    # Legend only for second panel (index 1)
    if idx == 1:
        ax.legend(fontsize=13, loc='best', frameon=False)

# Adjust spacing
plt.subplots_adjust(hspace=0.4, wspace=0.3)
output_file = figures_dir / "g6sulfur_ssp245_regional_bias.png"
plt.savefig(output_file, dpi=300, bbox_inches='tight')
plt.show()
```
